```python
print('''
步骤,关键操作,目的
1. 分布,对标准正态分布求平方和 $\sum Z_i^2$,拒绝黑盒：回归定义，用“正态的砖”砌出“卡方的墙”
2. 变量,自由度 $n$ (如 2→30),卡方分布唯一的“变形参数”
3. 计算,统计量 $Y = X_1^2 + \dots + X_n^2$,验证统计性质：期望 $E=n$ 与方差 $D=2n$
4. 判定,模拟 $\chi^2(n_1) + \chi^2(n_2)$,验证“可加性”：两个卡方相加，其结果依然是卡方
5. 视觉,渐变色 KDE + 理论虚线,直观看见分布从“极度右偏”被拉伸为“对称钟形”的过程
''')
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
import os

# ==========================================
# 0. 数据准备 (Data Preparation)
# ==========================================
# [cite_start]核心思想: 不推导公式，用暴力计算(蒙特卡洛)去"看见"规律 [cite: 3]
np.random.seed(2025) 
n_samples = 100000   # 样本量足够大以保证平滑

# --- 场景 A: 验证分布形态随自由度 n 的变化 (定义验证) ---
# 定义: chi^2(n) = sum(Z_i^2), 其中 Z_i ~ N(0,1)
dfs = [2, 5, 10, 30] # 不同的自由度
data_shape = {}

for df in dfs:
    # [cite_start]1. 造怪兽: 生成 df 个标准正态分布，然后平方求和 [cite: 6]
    # shape=(n_samples, df) -> 对每一行(axis=1)求平方和
    norm_samples = np.random.normal(loc=0, scale=1, size=(n_samples, df))
    chi_simulated = np.sum(norm_samples**2, axis=1)
    data_shape[df] = chi_simulated

# --- 场景 B: 验证可加性 (Additivity) ---
# 理论: X~chi2(n1), Y~chi2(n2) -> X+Y ~ chi2(n1+n2)
n1, n2 = 3, 5
# 模拟 n1
chi_n1 = np.sum(np.random.normal(0, 1, (n_samples, n1))**2, axis=1)
# 模拟 n2
chi_n2 = np.sum(np.random.normal(0, 1, (n_samples, n2))**2, axis=1)
# 两个独立卡方相加
chi_sum_actual = chi_n1 + chi_n2 

# ==========================================
# [cite_start]绘图部分：应用“搭台 -> 唱戏 -> 收尾”模式 [cite: 11, 12]
# ==========================================

# 设置全局字体 (Matplotlib 负责基础设施)
# 设置全局字体
plt.rcParams['font.sans-serif'] = ['SimHei']  # Windows/Linux
# plt.rcParams['font.sans-serif'] = ['Arial Unicode MS']  # 如果你是 Mac，请解开这行注释
plt.rcParams['axes.unicode_minus'] = False

# 创建一个 1行2列 的画布
fig, axes = plt.subplots(1, 2, figsize=(16, 6))

# ------------------------------------------
# 子图 1：形态演变与正态逼近 (定义与性质验证)
# ------------------------------------------
ax1 = axes[0]
colors = sns.color_palette("viridis", n_colors=len(dfs)) # 使用渐变色区分自由度

# x轴范围生成 (用于画理论线)
x = np.linspace(0, 60, 1000)

for i, df in enumerate(dfs):
    sim_data = data_shape[df]
    
    # [唱戏] Seaborn 绘制模拟分布 (KDE)
    # [cite_start]fill=True 是 Seaborn 的强项 [cite: 38]
    sns.kdeplot(sim_data, fill=True, color=colors[i], alpha=0.15, linewidth=1.5, 
                label=f'模拟 n={df}', ax=ax1)
    
    # [搭台] Scipy+Matplotlib 绘制理论曲线 (虚线)
    # 验证模拟数据是否完美贴合理论 pdf
    y_theory = stats.chi2.pdf(x, df)
    ax1.plot(x, y_theory, linestyle='--', color=colors[i], alpha=0.8)

# [收尾] 标注 n=30 时的正态近似特性
ax1.set_title("卡方分布形态演变: 从右偏到对称 (定义验证)", fontsize=14)
ax1.set_xlabel("统计量值 (Sum of Squares)", fontsize=12)
ax1.set_ylabel("概率密度", fontsize=12)
ax1.set_xlim(0, 50)
ax1.legend(title="自由度 (实线:模拟 / 虚线:理论)")

# ------------------------------------------
# 子图 2：可加性验证 (Additivity Check)
# ------------------------------------------
ax2 = axes[1]

# 1. [唱戏] 绘制相加后的模拟数据分布
# [cite_start]注意：使用双反斜杠 \\chi 避免转义警告 [cite: 86, 95]
sns.kdeplot(chi_sum_actual, fill=True, color='crimson', alpha=0.3, linewidth=2, 
            label=f'模拟: $\\chi^2({n1}) + \\chi^2({n2})$', ax=ax2)

# 2. [搭台] 绘制理论分布 chi2(n1+n2)
df_total = n1 + n2
x_add = np.linspace(0, 30, 1000)
y_add_theory = stats.chi2.pdf(x_add, df_total)

ax2.plot(x_add, y_add_theory, 'k--', linewidth=2, label=f'理论: $\\chi^2({df_total})$')

# 3. [计算验证] 验证期望与方差 (Annotation)
mean_sim = np.mean(chi_sum_actual)
var_sim = np.var(chi_sum_actual)
# 注意：使用双反斜杠 \\bar 避免转义警告
text_str = (f"理论期望 E = {df_total}\n模拟均值 $\\bar{{x}}$ = {mean_sim:.2f}\n"
            f"----------------\n"
            f"理论方差 D = {2*df_total}\n模拟方差 $S^2$ = {var_sim:.2f}")

# 在图中添加文本框显示统计指标
props = dict(boxstyle='round', facecolor='wheat', alpha=0.3)
ax2.text(0.65, 0.60, text_str, transform=ax2.transAxes, fontsize=11,
        verticalalignment='top', bbox=props)

# [收尾]
ax2.set_title(f"卡方分布可加性验证 ($n_1={n1}, n_2={n2}$)", fontsize=14)
ax2.set_xlabel("统计量值", fontsize=12)
ax2.set_ylabel("概率密度", fontsize=12)
ax2.legend(loc='upper right')

# ==========================================
# 保存与显示
# ==========================================
plt.tight_layout()

# 检查目录是否存在，不存在则创建（可选优化，防止报错）
output_dir = "Mathematical_statistics/images"
if not os.path.exists(output_dir):
    os.makedirs(output_dir)

# [cite_start]保存路径要用正斜杠 [cite: 47, 86]
save_path = f"{output_dir}/chi_squared_simulation.png"
plt.savefig(save_path, dpi=300, bbox_inches='tight') 

print(f"图片已保存: {save_path}") # 增加交互提示
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
import os

# ==========================================
# 0. 数据准备 (Data Preparation)
# ==========================================
np.random.seed(2025) 
n_samples = 50000   # 样本量

# --- 场景 A: 验证分布形态随自由度 n 的变化 ---
dfs = [2, 5, 10, 50] 
data_shape = {}

for df in dfs:
    # 生成标准正态分布并平方求和
    norm_samples = np.random.normal(loc=0, scale=1, size=(n_samples, df))
    chi_simulated = np.sum(norm_samples**2, axis=1)
    data_shape[df] = chi_simulated

# --- 场景 B: 验证可加性 (Additivity) ---
n1, n2 = 3, 5
chi_n1 = np.sum(np.random.normal(0, 1, (n_samples, n1))**2, axis=1)
chi_n2 = np.sum(np.random.normal(0, 1, (n_samples, n2))**2, axis=1)
chi_sum_actual = chi_n1 + chi_n2 

# ==========================================
# 绘图部分：使用 GridSpec 实现复杂布局
# ==========================================

# 设置全局字体
plt.rcParams['font.sans-serif'] = ['SimHei']  # Windows/Linux
# plt.rcParams['font.sans-serif'] = ['Arial Unicode MS']  # Mac
plt.rcParams['axes.unicode_minus'] = False

# 创建画布：使用 GridSpec 将画布分为 上下两部分
fig = plt.figure(figsize=(16, 10))
gs = fig.add_gridspec(2, 4, height_ratios=[1, 1], hspace=0.3)

# ------------------------------------------
# 第一部分：形态演变 (Subplots 1-4)
# ------------------------------------------
colors_hist = '#A6D8E0' # 模拟数据的颜色（浅蓝直方图）
color_theory = '#D93025' # 理论卡方的颜色（深红实线）
color_norm = '#188038'   # 理论正态的颜色（绿色虚线，用于对比）

for i, df in enumerate(dfs):
    ax = fig.add_subplot(gs[0, i])
    sim_data = data_shape[df]
    
    # 1. [模拟数据] 使用直方图 (Histogram)
    sns.histplot(sim_data, bins=50, stat="density", color=colors_hist, 
                 edgecolor='white', alpha=0.6, label='模拟 (Hist)', ax=ax)
    
    # 2. [理论卡方] 绘制 Chi2(n) 理论曲线
    x_min, x_max = ax.get_xlim()
    x = np.linspace(0, x_max, 400)
    y_chi = stats.chi2.pdf(x, df)
    ax.plot(x, y_chi, color=color_theory, linewidth=2.5, label=r'理论 $\chi^2$')
    
    # 3. [核心修改] 添加正态分布对比线
    if df >= 10: 
        mu_norm = df
        sigma_norm = np.sqrt(2 * df)
        y_norm = stats.norm.pdf(x, mu_norm, sigma_norm)
        # 使用虚线绘制正态分布
        ax.plot(x, y_norm, color=color_norm, linestyle='--', linewidth=2, label='正态近似')
        ax.text(0.95, 0.8, "趋于正态", transform=ax.transAxes, ha='right', color=color_norm, fontsize=10)

    ax.set_title(f'自由度 $n={df}$', fontsize=12, fontweight='bold')
    if i == 0:
        ax.set_ylabel("概率密度")
        ax.legend(fontsize=9)
    else:
        ax.set_ylabel("") # 省略其他图的y轴标签
    
    # 移除多余边框
    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)

# ------------------------------------------
# 第二部分：可加性验证 (Subplot 5)
# ------------------------------------------
ax_add = fig.add_subplot(gs[1, :]) # 占据第二行所有列

# 1. 模拟数据 (KDE + Fill)
sns.kdeplot(chi_sum_actual, fill=True, color='orange', alpha=0.3, linewidth=0, 
            label=r'模拟叠加: $\chi^2({n1}) + \chi^2({n2})$', ax=ax_add)

# 2. 理论曲线 (Bold Line)
df_total = n1 + n2
x_add = np.linspace(0, 40, 1000)
y_add_theory = stats.chi2.pdf(x_add, df_total)
ax_add.plot(x_add, y_add_theory, color='#E67C73', linestyle='--', linewidth=3, 
            label=r'理论分布: $\chi^2({df_total})$')

# 统计标注
mean_sim = np.mean(chi_sum_actual)
var_sim = np.var(chi_sum_actual)
text_str = (f"理论期望 E = {df_total} | 模拟均值 = {mean_sim:.2f}\n"
            f"理论方差 D = {2*df_total} | 模拟方差 = {var_sim:.2f}")
props = dict(boxstyle='round', facecolor='white', alpha=0.8, edgecolor='lightgray')
ax_add.text(0.28, 0.90, text_str, transform=ax_add.transAxes, fontsize=11,
            verticalalignment='top', bbox=props)

ax_add.set_title(f"场景 B: 卡方分布的可加性验证 (两个独立卡方之和)", fontsize=14)
ax_add.legend(loc='upper right')
ax_add.grid(True, linestyle=':', alpha=0.4)

# ==========================================
# 保存与输出
# ==========================================

# 添加全局主标题
plt.suptitle('卡方分布形态演变与性质验证', fontsize=18, y=0.98) 

# --- 核心修改：移除所有 tight_layout 调用，使用手动调整 ---
# 手动调整子图间距，消除 UserWarning，并优化视觉效果
plt.subplots_adjust(
    left=0.05,    # 整体左边界
    right=0.98,   # 整体右边界
    bottom=0.08,  # 底部边界 (为底部可能的全局图例或 x 轴标签留出空间)
    top=0.92,     # 顶部边界 (为 plt.suptitle 留出空间)
    wspace=0.25,  # 子图之间宽度间距 (用于第一行4个图)
    hspace=0.4    # 子图之间高度间距 (用于第一行和第二行之间)
)
# -----------------------------------------------------------

output_dir = "Mathematical_statistics/images"
if not os.path.exists(output_dir):
    os.makedirs(output_dir)
# 因为已经手动调整了，所以不需要 bbox_inches='tight'，但保留也没关系。
save_path = f"{output_dir}/chi_squared_evolution_v2.png" 
plt.savefig(save_path, dpi=300, bbox_inches='tight')

print(f"优化后的图片已保存: {save_path}")
plt.show()
```

```python
# ==============================================================================
# 核心笔记：t分布的灵魂与意义 (The Essence of Student's t-distribution)
# ==============================================================================
#
# 1. 核心痛点：上帝视角的缺失 (Missing Sigma)
#    - 理想国 (Z分布)：假设我们拥有上帝视角，知道总体的标准差 σ。
#    - 现实界 (t分布)：我们通常只有小样本 (n < 30)，只能用样本标准差 s 去冒充 σ。
#    - 代价：s 是不稳定的 (它是 σ 的拙劣模仿者)，直接用它套用正态分布会过于自信（低估风险）。
#
# 2. 形态直觉：厚尾 (Fat Tails) 与 矮峰
#    - 视觉特征：相比标准正态分布，t分布中间更低（矮），两边更高（翘）。
#    - "厚尾"的含义：它承认在小样本下，出现“极端值”或“远离均值的数据”的概率，
#      其实比正态分布预测的要大得多。
#
# 3. 哲学意义：无知带来的保守 (Conservative by Design)
#    - 为什么分布要变宽？
#      这是对“不确定性”的妥协。因为样本太少，我们不敢轻易断定某个偏离均值的数据是异常值。
#    - 宽容度的体现：
#      正态分布说：“超过 ±1.96 就是异类” (标准严苛)。
#      t分布说：“因为我样本少，我看不清全貌，哪怕你到了 ±4.0 我也暂且算你正常” (标准宽容)。
#      代价是牺牲了精度(置信区间变宽)，换取了安全性(不犯第一类错误)。
#
# 4. 演变规律：从“诚实”回归“理想”
#    - 当 n -> ∞ (通常 n > 30)：
#      样本标准差 s 越来越接近总体标准差 σ，不确定性消失。
#      t分布会自动卸下伪装，收敛成为标准正态分布 N(0,1)。
#      (这就是为什么在大样本下，t检验和z检验结果几乎一致的原因)。
#
# 5. 一句话总结：
#    正态分布是“盲目自信”的理想派（以为掌控一切波动）；
#    t分布是“谨小慎微”的现实派（承认样本少时会有额外风险，所以把路修得更宽）。
# ==============================================================================
```

```python
print('''
步骤,关键操作,核心逻辑 / 数学直觉
1. 还原,构造 $T = X / \sqrt{Y/n}$,拒绝黑盒：回归定义本源，亲手用“标准正态(X)”和“卡方(Y)”拼装出 t 统计量
2. 对标,引入标准正态 $N(0,1)$ (绿虚线),确立标尺：绿线代表“上帝视角(已知$\sigma$)”，红线代表“现实视角(未知$\sigma$)”，通过对比看见“无知”的代价
3. 洞察,观察 $n=2$ 时的“矮峰厚尾”,看见容忍：红线两端翘起(厚尾)，代表在小样本下，我们被迫提高了对“极端误差”的容忍度
4. 演变,观察 $n=30$ 时的“完美重合”,见证回归：随着信息量($n$)增加，样本标准差 $s$ 逼近真值 $\sigma$，t 分布卸下伪装回归正态
5. 验证,模拟直方图 + 理论密度线,闭环确认：散点堆成的直方图完美填充理论轮廓，证明我们对定义的物理构造是正确的
''')
```

```python
# [自我提醒1] 定义验证：无论 n=2 还是 n=30，蓝色直方图必须永远完美重合红色实线（t分布），这是数学定义的必然。
# [自我提醒2] 性质演变：n=2 时红线（t分布）与绿虚线（正态）的巨大差异才是正常的，这正是“厚尾”特性的直观体现。
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
import os

# ==========================================
# 0. 数据构造 (Construction from Definition)
# ==========================================
np.random.seed(2025)
n_samples = 100000 

# 定义对比场景：极小样本(n=2) vs 大样本(n=30)
scenarios = [2, 30]
sim_data = {}

for n in scenarios:
    # ---------------------------------------------------------
    # 关键步骤：不使用现成t函数，而是用"零件"组装
    # 定义：T = X / sqrt(Y/n)
    # ---------------------------------------------------------
    # 1. 造分子 X ~ N(0, 1)
    X = np.random.normal(loc=0, scale=1, size=n_samples)
    
    # 2. 造分母 Y ~ Chi2(n)
    Y = np.random.chisquare(df=n, size=n_samples)
    
    # 3. 组装 T
    T_simulated = X / np.sqrt(Y / n)
    
    # 数据清洗：为了画图美观，截断极端的离群值 (t分布在n小的时候极易产生巨大值)
    # 我们只关注 -6 到 6 之间的核心形态
    sim_data[n] = T_simulated[(T_simulated >= -6) & (T_simulated <= 6)]

# ==========================================
# 绘图部分 (Visual Verification)
# ==========================================
# 设置字体
plt.rcParams['font.sans-serif'] = ['SimHei'] 
plt.rcParams['axes.unicode_minus'] = False

# 创建 1x2 画布
fig, axes = plt.subplots(1, 2, figsize=(16, 7))

# 颜色定义
c_hist = '#A6D8E0'   # 模拟数据的直方图
c_theo = '#D93025'   # 理论t分布 (红色实线)
c_norm = '#188038'   # 标准正态对比 (绿色虚线)

x_axis = np.linspace(-6, 6, 1000)

for i, n in enumerate(scenarios):
    ax = axes[i]
    data = sim_data[n]
    
    # 1. [模拟层] 直方图 (代表我们手工构造的数据)
    sns.histplot(data, bins=80, stat="density", color=c_hist, 
                 edgecolor='white', alpha=0.5, label=r'模拟构造 $T=X/\sqrt{Y/n}$', ax=ax)
    
    # 2. [理论层] t分布密度曲线 (验证构造是否正确)
    y_t = stats.t.pdf(x_axis, df=n)
    ax.plot(x_axis, y_t, color=c_theo, linewidth=3, label=f'理论 $t({n})$ 分布')
    
    # 3. [参照层] 标准正态分布 (体现"厚尾"特征)
    y_norm = stats.norm.pdf(x_axis)
    ax.plot(x_axis, y_norm, color=c_norm, linestyle='--', linewidth=2, 
            label=r'标准正态 $N(0,1)$')
    
    # 装饰
    ax.set_title(f'场景 {i+1}: 自由度 $n={n}$', fontsize=14, fontweight='bold')
    ax.set_xlabel('统计量 T 值')
    ax.set_xlim(-5, 5)
    ax.set_ylim(0, 0.45) # 固定Y轴以便对比峰值高度
    ax.grid(True, linestyle=':', alpha=0.5)
    
    # --- 特殊标注：针对 n=2 的教学注解 ---
    if n == 2:
        ax.set_ylabel('概率密度')
        # 标注1: 矮峰
        ax.annotate('峰值更低\n(中间被压扁)', xy=(0, stats.t.pdf(0, n)), xytext=(-2.5, 0.38),
                    arrowprops=dict(facecolor='black', arrowstyle='->'), fontsize=11)
        # 标注2: 厚尾
        # 在 x=3 处，t分布明显高于正态
        t_val = stats.t.pdf(3, n)
        norm_val = stats.norm.pdf(3)
        ax.annotate('厚尾 (Fat Tails)\n对极端值的"容忍"', xy=(3, t_val), xytext=(2.2, 0.15),
                    arrowprops=dict(facecolor=c_theo, arrowstyle='->'), 
                    fontsize=11, color=c_theo, fontweight='bold')
        
        ax.legend(loc='upper right', fontsize=10)
        
    # --- 特殊标注：针对 n=30 的教学注解 ---
    if n == 30:
        ax.set_ylabel('')
        ax.text(0.25, 0.60, "当 $n$ 增大时\n$t$ 分布迅速逼近正态分布\n(红色实线盖住了绿色虚线)", 
                transform=ax.transAxes, ha='center', fontsize=12, 
                bbox=dict(boxstyle='round', facecolor='wheat', alpha=0.3))
        ax.legend(loc='upper right', fontsize=10)

# 全局标题
plt.suptitle('t 分布的构造验证与性质：从定义出发', fontsize=18, y=0.98)

# --- 手动布局调整 (避免被遮挡) ---
plt.subplots_adjust(
    left=0.08, right=0.95, 
    top=0.88, bottom=0.12, 
    wspace=0.15
)

# 保存
output_dir = "Mathematical_statistics/images"
if not os.path.exists(output_dir):
    os.makedirs(output_dir)
save_path = f"{output_dir}/t_distribution_definition_check.png"
plt.savefig(save_path, dpi=300)

print(f"图片已保存: {save_path}")
plt.show()
```

```python
# ==============================================================================
# 核心笔记：F分布的灵魂与意义 (The Essence of F-distribution)
# ==============================================================================
#
# 1. 核心直觉：统计学的天平 (The Variance Ratio)
#    - 物理本质：F = (方差1) / (方差2)。它是两个“波动源”的比值。
#    - 判决逻辑：
#      如果 F ≈ 1：说明分子和分母的波动程度相当，“势均力敌”，无显著差异。
#      如果 F >> 1：说明分子的波动（通常代表信号/组间差异）显著压倒了分母的波动（通常代表噪声/误差）。
#
# 2. 形态直觉：不对称的右偏 (Right-Skewed)
#    - 视觉特征：F分布永远从 0 开始（因为平方数之比不可能是负数），且拖着长长的右尾巴。
#    - 为什么长这样？
#      因为它是由两个卡方分布（都是右偏的）相除得到的。
#      这与正态分布或t分布的“钟形对称”完全不同，千万别画错了。
#
# 3. 哲学意义：信号与噪声的博弈 (Signal vs. Noise)
#    - 在方差分析(ANOVA)中：
#      分子 (Numerator) 代表“我们关心的效应”（比如不同药物导致的差异，即信号）。
#      分母 (Denominator) 代表“随机误差”（比如个体间的自然差异，即噪声）。
#    - F检验的本质：就是在问“我看到的差异(信号)，是不是大到足以忽略随机误差(噪声)的干扰？”
#
# 4. 家族关联：数学变形记 (Mathematical Connections)
#    - 倒数律 (Reciprocal)：
#      A是B的 2 倍 <==> B是A的 0.5 倍。
#      所以 F(n1, n2) 的倒数服从 F(n2, n1)。(仅仅是颠倒了视角)。
#    - t分布的影子：
#      t分布比较的是“均值”，F分布比较的是“方差”。
#      但当只比较两组均值时，t检验的平方 就是 F检验 (t^2 ~ F(1, n))。
#      这说明：比较均值差异，本质上也可以看作是比较“差异引起的波动”是否显著。
#
# 5. 一句话总结：
#    t分布是用来“修正误差”的尺子（在小样本下诚实地放宽标准）；
#    F分布是用来“衡量差异”的天平（通过比较波动的大小来判断信号强弱）。
# ==============================================================================
```

```python
print('''
步骤,关键操作,核心逻辑 / 统计直觉
1. 构造,F = (U/n1) / (V/n2),拒绝黑盒：回归定义，用两个"卡方"积木拼搭出标准的 F分布，验证数学公式的正确性
2. 基准,绘制红色理论线 $F(n1, n2)$,确立标尺：这条红线代表"零假设"的世界，即两个比较对象完全没有区别时的理想状态
3. 对比,场景A(左图) F ≈ 1,验证定义：当两组方差确实相等时，模拟数据(蓝)乖乖地躺在红线(理论)的怀抱里，天平平衡
4. 冲击,场景B(右图) F >> 1,看见信号：当一组方差放大(有信号)时，模拟数据(橙)剧烈向右逃逸，与红线彻底分家
5. 判决,红线与色块的分离,假设检验：这种"理论(红)"与"现实(橙)"的错位，正是我们在ANOVA中拒绝零假设的视觉证据
''')
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
import os

# ==============================================================================
# 核心笔记：F分布的灵魂与意义 (The Essence of F-distribution)
# ==============================================================================
# [自我提醒1] 定义验证：当分子分母“势均力敌”时(方差相等)，F值聚集在 1 附近。蓝色直方图应完美填充红色理论线。
# [自我提醒2] 物理直觉：当分子“波动剧烈”时(方差不等)，天平倾斜，F值被剧烈推向右侧。直方图将与“零假设”的红线彻底分离。
#
# 1. 构造直觉：F = (U/n1) / (V/n2)。它本质是两个“卡方”的比值，也就是“方差”的比值。
# 2. 形态直觉：F分布永远从0开始，右偏（长尾巴）。不同于正态分布的对称钟形。
# 3. 检验直觉：F检验的核心就是看 F值是不是“大得离谱”。如果 F >> 1，说明我们捕获了显著的“信号”。
# ==============================================================================

# ==========================================
# 0. 数据构造 (Construction from Definition)
# ==========================================
np.random.seed(2025)
n_samples = 100000 
n1, n2 = 10, 20 # 设定自由度

# --- 场景 A: 零假设成立 (H0: σ1 = σ2) ---
# 此时 F = (S1^2/σ^2) / (S2^2/σ^2) = S1^2 / S2^2
U = np.random.chisquare(df=n1, size=n_samples)
V = np.random.chisquare(df=n2, size=n_samples)
F_H0 = (U / n1) / (V / n2)
# 截断极值以便绘图
data_H0 = F_H0[F_H0 < 8]

# --- 场景 B: 备择假设成立 (H1: σ1 = 2*σ2) ---
# 假设分子对应的总体标准差是分母的 2 倍 (即方差是 4 倍)
ratio_variance = 2**2 # 方差放大4倍
F_H1 = F_H0 * ratio_variance 
data_H1 = F_H1[F_H1 < 15] # 范围会变大

# ==========================================
# 绘图部分 (Visual Verification)
# ==========================================
# 设置全局字体与负号显示
plt.rcParams['font.sans-serif'] = ['SimHei'] 
plt.rcParams['axes.unicode_minus'] = False

# 创建 1x2 画布
fig, axes = plt.subplots(1, 2, figsize=(16, 7))

# 颜色定义 (高对比度)
c_hist = '#A6D8E0'   # 模拟数据的直方图 (浅蓝)
c_theo = '#D93025'   # 理论F分布 (深红实线)
c_signal = '#FF9F43' # 信号数据的直方图 (橙色)

# X轴生成
x_axis = np.linspace(0, 15, 1000)

# --- [左图] 场景 A: 定义验证 (Null Hypothesis) ---
ax1 = axes[0]
sns.histplot(data_H0, bins=60, stat="density", color=c_hist, 
             edgecolor='white', alpha=0.6, label='模拟 F (方差相等)', ax=ax1, zorder=1)

y_theo = stats.f.pdf(x_axis, n1, n2)
ax1.plot(x_axis, y_theo, color=c_theo, linewidth=3, label=f'理论 $F({n1}, {n2})$', zorder=3)

# 装饰 A
# [修正点1] 使用 r'...' 原始字符串，解决 \sigma 转义警告
ax1.set_title(r'场景 A: 定义验证 (方差齐性 $\sigma_1=\sigma_2$)', fontsize=14, fontweight='bold')
ax1.set_xlabel('F 统计量值')
ax1.set_ylabel('概率密度')
ax1.set_xlim(0, 6)
ax1.set_ylim(0, 0.9)
ax1.grid(True, linestyle=':', alpha=0.5)

# 标注 A
ax1.annotate(r'大多数 F 值 $\approx 1$' + '\n(势均力敌)', 
             xy=(0.8, stats.f.pdf(0.8, n1, n2)), xytext=(2.5, 0.6),
             arrowprops=dict(facecolor='black', arrowstyle='->', lw=1.5), 
             fontsize=11, fontweight='bold', alpha=0.8)
ax1.legend(loc='upper right')


# --- [右图] 场景 B: 物理直觉 (Signal vs Noise) ---
ax2 = axes[1]
sns.histplot(data_H1, bins=60, stat="density", color=c_signal, 
             edgecolor='white', alpha=0.5, label='模拟 F (方差不等)', ax=ax2, zorder=1)

# 依然画出标准的零假设分布作为对比
ax2.plot(x_axis, y_theo, color=c_theo, linewidth=3, linestyle='--', label=f'零假设基准 $F({n1}, {n2})$', zorder=3)

# 装饰 B
# [修正点1] 使用 r'...' 原始字符串
ax2.set_title(r'场景 B: 物理直觉 (信号显著 $\sigma_1=2\sigma_2$)', fontsize=14, fontweight='bold')
ax2.set_xlabel('F 统计量值')
ax2.set_ylabel('')
ax2.set_xlim(0, 15)
ax2.grid(True, linestyle=':', alpha=0.5)

# 标注 B
# [修正点2] 移除了 facecolor，因为 arrowstyle='->' 是线条箭头，只能设置 color
ax2.annotate('天平倾斜!\n数据显著右移 (F >> 1)', 
             xy=(4, 0.1), xytext=(7, 0.3),
             arrowprops=dict(arrowstyle='->', color='black', lw=2), 
             fontsize=12, color='black', fontweight='bold',
             bbox=dict(boxstyle='round,pad=0.5', facecolor=c_signal, alpha=0.3))

ax2.text(0.5, 0.6, "红虚线：无差异时的样子 (理论)\n橙色块：有差异时的样子 (现实)\n\n两者分离 $\\to$ 拒绝零假设", 
        transform=ax2.transAxes, ha='center', fontsize=11, 
        bbox=dict(boxstyle='round', facecolor='wheat', alpha=0.3))

ax2.legend(loc='upper right')

# 全局标题
plt.suptitle('F 分布的双重验证：数学定义 vs 物理意义', fontsize=18, y=0.98)

# --- 手动布局调整 ---
plt.subplots_adjust(
    left=0.08, right=0.95, 
    top=0.88, bottom=0.12, 
    wspace=0.15
)

# 保存
output_dir = "Mathematical_statistics/images"
if not os.path.exists(output_dir):
    os.makedirs(output_dir)
save_path = f"{output_dir}/f_distribution_definition_intuition_final.png"
plt.savefig(save_path, dpi=300)

print(f"图片已保存: {save_path}")
plt.show()
```

```python
# ==================================================================================
# F(1, n) 分布在 X=0 处“起飞”的原因 (Why the F(1, n) PDF Spikes at Zero)
# ==================================================================================
# 核心直觉：这是 t 分布的“概率质量”被平方运算强行挤压到原点造成的结果。
# 
# 1. T 分布的聚集性 (The Parent Distribution)
# # - t 分布的特点：绝大部分概率质量都集中在均值 T=0 的附近。
# # - 物理图像：在 t 统计量的坐标轴上，数据点在 T=0 附近堆得最密。
# 
# 2. 平方运算的“挤压效应” (The Squaring Squeeze)
# # - 当对 T 统计量进行平方 (T²) 运算时：
# # - 所有靠近 0 的小数值（例如 ±0.1），平方后会变得更小（例如 0.01）。
# # - 结果：原本在 T≈0 附近的大量概率，被强制挤压到了 F≈0 的这个极小的区间内。
# 
# 3. 密度的爆发 (Density Takeoff)
# # - 由于概率被极度聚集在原点，为了保持曲线下的总面积为 1，
# # - 必然导致 F 分布在 x 趋近于 0 时，其高度（概率密度）急剧上升，趋向无穷。
# # - 视觉体现：这个尖刺就是你在图上看到的，密度函数在左侧“拔地而起”的现象。
# # ----------------------------------------------------------------------------------
```

```python
# ==============================================================================
# 核心笔记：样本均值的分布与标准化 (Sampling Distribution of the Mean)
# ==============================================================================
#
# 1. 核心直觉：平均值的“聚拢效应” (The Clustering Effect)
#    - 现象：单个样本 x_i 可能会离经叛道（偏离 μ 很远）。
#    - 机制：样本均值 X_bar 是大家的“妥协”。极端大的值会被极端小的值抵消。
#    - 结果：X_bar 的波动范围 (σ²/n) 远小于个体的波动 (σ²)。
#    - 规律：样本量 n 越大，分布越“瘦高”，均值越紧密地聚集在真值 μ 身边（精度提升）。
#
# 2. 关键指标：标准误 (Standard Error, SE = σ/√n)
#    - 定义：它是衡量样本均值“测不准程度”的专用尺子。
#    - 区别：σ 是“个体”的波动；σ/√n 是“平均值”的波动。
#    - 代价：如果你想把估计精度提高 10 倍 (让分布变窄 10 倍)，样本量 n 必须扩大 100 倍 (平方根律)。
#
# 3. 标准化操作：U统计量 (The Z-Score Transformation)
#    - 底层逻辑：Z = (观测值 - 期望均值) / 标准差
#      (这是所有正态标准化的通用模板，目的是消除单位，衡量偏离程度)
#
#    - 具体落地：U = (X_bar - μ) / (σ/√n) ~ N(0, 1)
#      * 这里的“观测值”是 X_bar (样本均值)
#      * 这里的“期望均值”是 μ (总体均值)
#      * 这里的“标准差”是 σ/√n (标准误 SE)
#
#    - 物理含义：U 代表“你的样本均值偏离了总体均值几个‘标准误’？”
#      (如果 |U| > 2，说明这种偏离在随机抽样下属于小概率事件)。
#
# 4. 承上启下：通往 t分布 的桥梁
#    - 理想：这里假设 σ (上帝视角的标准差) 是已知的，所以 U 服从完美的标准正态分布 Z。
#    - 现实：如果 σ 未知，只能用 s (样本标准差) 去“冒充”分母，那么 U 就会劣化为 t分布。
# ==============================================================================
```

```python
print('''
步骤,关键操作,核心逻辑 / 统计直觉
1. 抽样,生成 (N, n) 矩阵并求均值,模仿现实：从宽阔的总体(个体)中抽样并取平均，模拟出"样本均值"的分布
2. 对比(左图),总体 vs 均值,聚拢效应：绿色的总体很宽，红色的均值很窄。直观展示 √n 带来的精度提升(方差缩小)
3. 变换,U = (X_bar - μ) / SE,标准化：将有单位的均值转化为无量纲的 Z分数，衡量"偏离了几个标准误"
4. 验证(右图),直方图(橙) + 虚线(黑),归一化：原本在 50 附近的红色分布，经过变换完美重合到了 0 附近的标准正态分布
''')
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
import os

# ==============================================================================
# 核心笔记：样本均值的分布与标准化 (Sampling Distribution of the Mean)
# ==============================================================================
# 1. 核心直觉：平均值的“聚拢效应” (The Clustering Effect)
#    - 现象：样本均值 X_bar 的波动范围 (σ²/n) 远小于个体的波动 (σ²)。
#    - 规律：样本量 n 越大，分布越“瘦高”，均值越紧密地聚集在真值 μ 身边（精度提升）。
#
# 2. 关键指标：标准误 (Standard Error, SE = σ/√n)
#    - 区别：σ 是“个体”的波动；σ/√n 是“平均值”的波动。
#
# 3. 标准化操作：U统计量 (The Z-Score Transformation)
#    - 公式：U = (X_bar - μ) / (σ/√n) ~ N(0, 1)
#    - 物理含义：U 代表“你的样本均值偏离了总体均值几个‘标准误’？”
# ==============================================================================

# ==========================================
# 0. 数据构造 (Construction)
# ==========================================
np.random.seed(2025)
n_sims = 100000

# 设定上帝视角参数 (Population Parameters)
mu = 50      # 总体均值
sigma = 10   # 总体标准差
n = 25       # 样本量 (便于计算, sqrt(25)=5)

# 理论上的标准误 (SE): σ / sqrt(n)
se = sigma / np.sqrt(n) # 10/5 = 2

# 1. 模拟抽样：生成 (n_sims, n) 的矩阵
# 每一行代表一次实验，抽取了 n 个样本
samples = np.random.normal(mu, sigma, (n_sims, n))

# 2. 计算样本均值 X_bar
x_bar = np.mean(samples, axis=1)

# 3. 计算标准化统计量 U
# U = (样本均值 - 总体均值) / 标准误
u_stats = (x_bar - mu) / se

# ==========================================
# 绘图部分 (Visualization)
# ==========================================
plt.rcParams['font.sans-serif'] = ['SimHei']
plt.rcParams['axes.unicode_minus'] = False

fig, axes = plt.subplots(1, 2, figsize=(16, 7))

# 颜色定义
c_pop = '#A3E4D7'   # 总体 (浅绿)
c_mean = '#E74C3C'  # 均值 (红色)
c_std = '#F39C12'   # 标准化 (橙色)
c_ref = '#2C3E50'   # 参考线 (深色)

# -----------------------------------------------------------
# 左图：聚拢效应 (Individual vs Average)
# -----------------------------------------------------------
ax1 = axes[0]
x_range_raw = np.linspace(mu - 4*sigma, mu + 4*sigma, 1000)

# 1. 画背景：总体分布 N(μ, σ)
ax1.plot(x_range_raw, stats.norm.pdf(x_range_raw, mu, sigma), 
         color=c_ref, linestyle='--', alpha=0.5, label=r'个体分布 $N(\mu, \sigma^2)$')
ax1.fill_between(x_range_raw, stats.norm.pdf(x_range_raw, mu, sigma), color=c_pop, alpha=0.3)

# 2. 画前景：样本均值分布 X_bar
sns.histplot(x_bar, bins=60, stat="density", color=c_mean, edgecolor='white', alpha=0.6, 
             label=f'样本均值 $\\bar{{X}}$ (n={n})', ax=ax1)
# 理论曲线 N(μ, σ²/n)
ax1.plot(x_range_raw, stats.norm.pdf(x_range_raw, mu, se), 
         color=c_mean, linewidth=2.5, label=r'理论分布 $N(\mu, \sigma^2/n)$')

ax1.set_title(rf'1. 聚拢效应：平均值比个体更稳定 (SE = $\sigma/\sqrt{{n}}$)', fontsize=14, fontweight='bold')
ax1.set_xlabel('原始数值 (例如: 体重)')
ax1.set_xlim(mu - 3.5*sigma, mu + 3.5*sigma)
ax1.legend(loc='upper right')

# 标注：展示宽度的不同
ax1.annotate('个体波动大\n(宽)', xy=(mu + 2*sigma, 0.01), xytext=(mu + 2*sigma, 0.05),
             arrowprops=dict(arrowstyle='->', color=c_ref), ha='center', color=c_ref)
ax1.annotate('均值波动小\n(窄)', xy=(mu + 2*se, 0.1), xytext=(mu + 3*se, 0.15),
             arrowprops=dict(arrowstyle='->', color=c_mean), ha='center', color=c_mean, fontweight='bold')

# -----------------------------------------------------------
# 右图：标准化 (Standardization)
# -----------------------------------------------------------
ax2 = axes[1]
x_range_z = np.linspace(-4, 4, 1000)

# 1. 模拟的 U 统计量
sns.histplot(u_stats, bins=60, stat="density", color=c_std, edgecolor='white', alpha=0.6, 
             label=r'模拟统计量 $U = \frac{\bar{X}-\mu}{\sigma/\sqrt{n}}$', ax=ax2)

# 2. 理论标准正态 N(0, 1)
ax2.plot(x_range_z, stats.norm.pdf(x_range_z), 
         color=c_ref, linewidth=3, linestyle='--', label='标准正态 $N(0, 1)$')

ax2.set_title('2. 标准化：消除单位，回归 N(0,1)', fontsize=14, fontweight='bold')
ax2.set_xlabel('Z-Score (无量纲)')
ax2.set_xlim(-4, 4)
ax2.legend(loc='upper right')

# 标注
ax2.text(0.5, 0.6, "无论原始参数是多少\n标准化后都变成\n均值0, 方差1", 
         transform=ax2.transAxes, ha='center', 
         bbox=dict(boxstyle='round', facecolor='white', alpha=0.8))

# 全局布局
plt.suptitle('样本均值的分布与标准化验证', fontsize=18, y=0.98)
plt.subplots_adjust(left=0.08, right=0.95, top=0.88, wspace=0.2)

# 保存
output_dir = "Mathematical_statistics/images"
if not os.path.exists(output_dir):
    os.makedirs(output_dir)
save_path = f"{output_dir}/sample_mean_distribution.png"
plt.savefig(save_path, dpi=300)

print(f"图片已保存: {save_path}")
plt.show()
```

```python
# ==============================================================================
# 核心笔记：基于总体均值的卡方分布 (The Foundational Chi-Squared Definition)
# ==============================================================================
# 
# 1. 数学定义：Chi^2 = Sum(Z_i^2) ~ Chi^2(n)
#    - 卡方分布是 n 个相互独立的标准正态变量 (Z) 的平方和。
#    - 这是卡方分布最纯粹、最基础的形态，自由度 n 等于相加项的数量。
# 
# 2. 物理意义：度量“标准化后的总波动”
#    - 公式本质：它计算了所有观测值 (X_i) 偏离“真实中心” (μ) 的总平方距离，并用总体方差 (σ²) 进行标准化。
#    - 作用：用来衡量样本的波动程度是否符合我们对总体的预期。
# 
# 3. 核心洞察：为何需要知道 μ？
#    - 此定义 (自由度 n) 成立的前提是：总体均值 μ 必须是**已知**的。
#    - 现实中 μ 通常未知，我们不得不使用样本均值 X_bar 代替 μ。
#    - 当使用 X_bar 时，会损失一个自由度，卡方分布将变为 Chi^2(n-1)。
#    - 这个 n-1 的卡方分布，才是构建 t分布和F分布时，分母所使用的真实形态。
# ==============================================================================
```

```python
print('''
步骤,关键操作,核心逻辑 / 数学直觉
1. 构造,Z_i^2 \to \sum Z_i^2,拒绝黑盒：回归卡方最纯粹的定义，通过 n 个独立标准正态的平方和来构造分布
2. 观察,自由度 n 的变化 (2 \to 30),形状演变：n 既是自由度也是分布的均值，展示中心如何向右移动
3. 洞察,形态从右偏到对称,见证收敛：证明 Central Limit Theorem (CLT) 适用于和的分布，随着 n 增大，形状趋近正态
4. 验证,直方图 vs 理论线 (2x2),闭环确认：模拟数据完美贴合理论曲线，验证了定义的正确性和参数设置
''')
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import chi2
import seaborn as sns
import os

# ==============================================================================
# 核心笔记：基于总体均值的卡方分布 (The Foundational Chi-Squared Definition)
# ==============================================================================
# 1. 数学定义：Chi^2 = Sum(Z_i^2) ~ Chi^2(n)
#    - 卡方分布是 n 个相互独立的标准正态变量 (Z) 的平方和。
# 2. 物理意义：度量“标准化后的总波动”，用来衡量样本波动是否符合预期。
# 3. 核心洞察：此定义依赖于总体均值 μ 已知，故自由度等于相加项的数量 n。
# ==============================================================================

# ==========================================
# 0. 数据构造 (Construction from Definition)
# ==========================================
np.random.seed(2025)
n_samples = 100000

# 不同的自由度 n，展示形状变化
dfs = [2, 5, 15, 30] 
data_chi2 = {}

for df in dfs:
    # 1. 生成 df 个标准正态分布 Z_i (100000 x df 矩阵)
    norm_samples = np.random.normal(loc=0, scale=1, size=(n_samples, df))
    
    # 2. 核心操作：平方求和 (Sum of Z_i^2)
    chi_simulated = np.sum(norm_samples**2, axis=1)
    
    # 3. 数据清理：为了绘图美观，截断极值（通常在 x=n+4*sqrt(2n) 附近）
    # 我们只保留 99.9% 的数据
    threshold = np.percentile(chi_simulated, 99.9)
    data_chi2[df] = chi_simulated[chi_simulated < threshold]

# ==========================================
# 绘图部分 (Visualization: Shape Evolution)
# ==========================================
plt.rcParams['font.sans-serif'] = ['SimHei']
plt.rcParams['axes.unicode_minus'] = False

# 2x2 子图阵列
fig, axes = plt.subplots(2, 2, figsize=(14, 12))
axes = axes.flatten()

# 颜色设置
c_hist = '#ADD8E6'  # 直方图 (浅蓝)
c_theo = '#8B0000'  # 理论曲线 (深红)

for i, df in enumerate(dfs):
    ax = axes[i]
    sim_data = data_chi2[df]
    
    # 1. 绘制直方图 (模拟数据)
    sns.histplot(sim_data, bins=50, stat="density", color=c_hist, 
                 edgecolor='white', alpha=0.8, label='模拟 $\\sum Z^2$', ax=ax, zorder=1)
    
    # 2. 绘制理论卡方分布 (验证定义)
    # x 轴范围设定，稍微超过均值 n
    x_max = df + 3 * np.sqrt(2 * df) 
    x = np.linspace(0.01, x_max, 500) # 避开 x=0 的奇点
    y_theory = chi2.pdf(x, df)
    
    ax.plot(x, y_theory, color=c_theo, linestyle='-', linewidth=2.5, 
            label=f'理论 $\\chi^2({df})$', zorder=2)
    
    # 3. 装饰与标注
    ax.set_title(f'自由度 $n = {df}$', fontsize=15, fontweight='bold')
    ax.set_xlabel('$\\chi^2$ 统计量值')
    ax.set_ylabel('概率密度')
    ax.legend(loc='upper right')
    ax.grid(True, linestyle=':', alpha=0.6)
    
    # 标注中心点 (均值 = df)
    ax.axvline(df, color='gray', linestyle='--', linewidth=1.5)
    
    if df == 30:
         ax.text(0.65, 0.7, "趋向对称\n(近似正态)", transform=ax.transAxes, 
                 ha='center', fontsize=12, color='darkgreen')

plt.suptitle('卡方分布的构造与形态演变 (基于基础定义)', fontsize=18, y=0.95)
plt.tight_layout(rect=[0, 0.03, 1, 0.9])

# 保存图片
output_dir = "Mathematical_statistics/images"
if not os.path.exists(output_dir):
    os.makedirs(output_dir)
save_path = f"{output_dir}/chi_squared_foundation_n_evolution.png"
plt.savefig(save_path, dpi=300)

print(f"图片已保存: {save_path}")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
import os

# ==============================================================================
# 核心笔记：卡方分布的本质重构 (The Essence of Chi-Squared)
# ==============================================================================
# [自我提醒1] 原子构成：Chi^2(1) 是“原点爆发”的。因为 Z^2 大概率很小，极大概率堆积在 0 附近。
# [自我提醒2] 累加效应：随着 n 增加，均值(n)右移，方差(2n)变大。图像从“贴墙滑梯”变成“对称钟形”。
#
# 1. 构造逻辑：Chi^2 = Sum(Z_i^2)。它是衡量“总变异”的容器。
# 2. 演变逻辑：
#    - n=1: 极度右偏，像指数分布。
#    - n=4: 峰值出现，但依然右偏。
#    - n=10: 逐渐对称。
#    - n=50: 完美正态 (CLT生效)。
# ==============================================================================

# ==========================================
# 0. 数据构造 (Construction)
# ==========================================
np.random.seed(2025)
n_samples = 100000

# 选取具有代表性的自由度节点
dfs = [1, 4, 10, 50] 
sim_data = {}

for n in dfs:
    # 1. 核心构造：生成 n 个标准正态，平方，求和
    # 形状: (n_samples, n) -> sum -> (n_samples,)
    Z = np.random.normal(0, 1, (n_samples, n))
    chi2_val = np.sum(Z**2, axis=1)
    
    # 截断极值以便绘图 (保留 99.5% 的数据)
    limit = np.percentile(chi2_val, 99.5)
    sim_data[n] = chi2_val[chi2_val <= limit]

# ==========================================
# 绘图部分 (Visual Verification)
# ==========================================
plt.rcParams['font.sans-serif'] = ['SimHei'] 
plt.rcParams['axes.unicode_minus'] = False

# 创建 2x2 画布
fig, axes = plt.subplots(2, 2, figsize=(16, 10))
axes = axes.flatten()

# 颜色定义
c_hist = '#82CCDD'   # 模拟直方图 (清爽蓝)
c_theo = '#B71540'   # 理论卡方线 (深红)
c_norm = '#009432'   # 正态近似线 (深绿)
c_mean = '#60a3bc'   # 均值指示线 (灰蓝)

for i, n in enumerate(dfs):
    ax = axes[i]
    data = sim_data[n]
    
    # X轴范围生成 (动态调整)
    x_max = np.max(data)
    # 避开 0 的奇点 (针对 n=1)
    x = np.linspace(0.01 if n==1 else 0, x_max, 1000)
    
    # --- 1. [模拟层] 直方图 ---
    sns.histplot(data, bins=60, stat="density", color=c_hist, 
                 edgecolor='white', alpha=0.6, label=r'模拟 $\sum Z^2$', ax=ax, zorder=1)
    
    # --- 2. [理论层] 卡方分布曲线 ---
    y_chi = stats.chi2.pdf(x, df=n)
    # [修正] 添加 r 前缀，变为 rf'...'
    ax.plot(x, y_chi, color=c_theo, linewidth=3, label=rf'理论 $\chi^2({n})$', zorder=3)
    
    # --- 3. [参照层] 正态近似 (仅在大样本时显示) ---
    if n >= 10:
        # 卡方近似于 N(n, 2n)
        mu_approx = n
        sigma_approx = np.sqrt(2 * n)
        y_norm = stats.norm.pdf(x, mu_approx, sigma_approx)
        
        linestyle = '--' if n < 50 else '-' 
        # [修正] 添加 r 前缀
        label_norm = rf'正态近似 $N({n}, {2*n})$'
        ax.plot(x, y_norm, color=c_norm, linestyle='--', linewidth=2.5, label=label_norm, zorder=2)

    # --- 4. [均值层] 均值指示线 ---
    # [修正] 添加 r 前缀
    ax.axvline(n, color=c_mean, linestyle=':', linewidth=2, label=rf'均值 $\mu={n}$')

    # --- 5. 装饰与教学标注 ---
    ax.set_title(f'自由度 $n={n}$', fontsize=14, fontweight='bold')
    # [修正] 添加 r 前缀
    ax.set_xlabel(r'$\chi^2$ 值')
    if i % 2 == 0: ax.set_ylabel('概率密度')
    
    ax.grid(True, linestyle=':', alpha=0.5)
    ax.legend(loc='upper right', fontsize=10)
    
    # 特殊标注
    if n == 1:
        ax.annotate('原点爆发!\n(Z接近0的概率最大)', xy=(0.2, 1.0), xytext=(1.5, 2.0),
                    arrowprops=dict(facecolor='black', arrowstyle='->'), fontsize=11)
    if n == 50:
        ax.text(0.05, 0.6, "CLT生效：\n形状完全对称\n红线与绿虚线重合", transform=ax.transAxes,
                bbox=dict(boxstyle='round', facecolor='white', alpha=0.8, edgecolor='gray'))

# 全局标题
plt.suptitle('卡方分布的本质：从“原点爆发”到“正态收敛”', fontsize=20, y=0.96)

# 布局优化
plt.subplots_adjust(left=0.08, right=0.95, top=0.88, bottom=0.1, hspace=0.35, wspace=0.2)

# 保存
output_dir = "Mathematical_statistics/images"
if not os.path.exists(output_dir):
    os.makedirs(output_dir)
save_path = f"{output_dir}/chi_squared_evolution_rethink.png"
plt.savefig(save_path, dpi=300)

print(f"图片已保存: {save_path}")
plt.show()
```

```python
### 2.3. 基于样本均值的卡方分布
```

```python
# ==============================================================================
# 核心笔记：样本方差的卡方分布 (Chi-Squared with Sample Mean)
# ==============================================================================
# 
# 1. 公式的变化：从 μ 到 X_bar
#    - 前作：Sum((X - μ)^2) / σ^2 ~ Chi^2(n)。(参照物是固定的真理 μ)
#    - 本作：Sum((X - X_bar)^2) / σ^2 ~ Chi^2(n-1)。(参照物是浮动的样本重心 X_bar)
#
# 2. 自由度的损失 (The Cost of Estimation)
#    - 直觉：样本均值 X_bar 是为了让这组样本的平方和最小而计算出来的。
#    - 后果：它"偷吃"了一部分波动。导致计算出的总变异量比相对于 μ 的变异量要小。
#    - 修正：为了匹配这个变小的量，分布的自由度必须从 n 降为 n-1。
#
# 3. 物理意义：方差的分布
#    - 这个公式 (n-1)S^2 / σ^2 是推断总体方差 σ^2 的基础。
#    - 它告诉我们：如果我们不断抽样并计算样本方差 S^2，这些 S^2 构成的分布形状就是 Chi^2(n-1)。
# ==============================================================================
```

```python
print('''
步骤,关键操作,核心逻辑 / 统计直觉
1. 路线A,Sum((X-μ)^2),上帝视角：计算离"总体均值"的距离。这是基准，拥有完整的 n 个自由度。
2. 路线B,Sum((X-X_bar)^2),凡人视角：计算离"样本均值"的距离。因为样本均值总是位于数据中心，导致总距离变小。
3. 对比,红蓝错位,看见缩水：红色直方图(B)明显位于蓝色直方图(A)的左侧。这证明样本方差低估了真实波动。
4. 验证,理论拟合,自由度修正：红色数据必须配对 Chi^2(n-1) 的曲线才能重合，证明"减1"是修正这一偏差的必要手段。
''')
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
import os

# ==============================================================================
# 核心笔记：样本方差的分布与自由度损失 (The Loss of Degrees of Freedom)
# ==============================================================================
# 1. 核心冲突：参照中心的不同
#    - 路线A (上帝视角)：计算 Sum(X - μ)^2。μ 是客观存在的真值，不随样本变动。
#    - 路线B (凡人视角)：计算 Sum(X - X_bar)^2。X_bar 是计算出来的样本均值，随样本而动。
#
# 2. 数学事实：最小二乘原理
#    - 样本均值 X_bar 的定义特性就是：让这组数据的离差平方和达到最小。
#    - 即：Sum(X - X_bar)^2 < Sum(X - μ)^2 (除非 X_bar 恰好等于 μ)。
#
# 3. 统计后果：自由度献祭
#    - 因为使用了样本均值，计算出的波动总量被"低估"了。
#    - 为了修正这个系统性的低估，理论分布的自由度必须从 n 降为 n-1。
# ==============================================================================

# ==========================================
# 0. 数据构造 (Construction)
# ==========================================
np.random.seed(2025)
n_sims = 100000
n = 6       # 样本量
mu = 50
sigma = 10

# 生成样本矩阵 (N, n)
samples = np.random.normal(mu, sigma, (n_sims, n))

# --- 路线 A: 基于总体均值 mu (无损失) ---
# 自由度: n
chi2_mu = np.sum((samples - mu)**2, axis=1) / (sigma**2)

# --- 路线 B: 基于样本均值 X_bar (有损失) ---
# 自由度: n-1
x_bar = np.mean(samples, axis=1, keepdims=True)
chi2_xbar = np.sum((samples - x_bar)**2, axis=1) / (sigma**2)

# ==========================================
# 绘图部分 (Visualization)
# ==========================================
plt.rcParams['font.sans-serif'] = ['SimHei']
plt.rcParams['axes.unicode_minus'] = False

fig, ax = plt.subplots(figsize=(12, 7))

# 颜色定义
c_mu = '#3498DB'    # 蓝色 (n)
c_xbar = '#E74C3C'  # 红色 (n-1)

# X轴范围
x_axis = np.linspace(0, 20, 1000)

# 1. 绘制 路线A (基于 μ)
sns.histplot(chi2_mu, bins=80, stat="density", color=c_mu, alpha=0.3, element="step",
             label=rf'基于总体均值 $\mu$ (n)', ax=ax)
ax.plot(x_axis, stats.chi2.pdf(x_axis, n), color=c_mu, lw=2.5, ls='--', 
        label=rf'理论 $\chi^2({n})$')

# 2. 绘制 路线B (基于 X_bar)
sns.histplot(chi2_xbar, bins=80, stat="density", color=c_xbar, alpha=0.5, element="step",
             label=rf'基于样本均值 $\bar{{X}}$ (n-1)', ax=ax)
ax.plot(x_axis, stats.chi2.pdf(x_axis, n-1), color=c_xbar, lw=3, 
        label=rf'理论 $\chi^2({n-1})$')

# 3. 装饰与标注
ax.set_title(rf'自由度的代价：从 $\mu$ 到 $\bar{{X}}$ 的分布左移 (样本量 n={n})', fontsize=16, fontweight='bold')
ax.set_xlabel(r'$\chi^2$ 统计量值')
ax.set_ylabel('概率密度')
ax.set_xlim(0, 20)

# 核心标注：左移箭头
peak_mu = n - 2
peak_xbar = (n-1) - 2
ax.annotate('整体左移\n(波动缩水)', 
            xy=(peak_xbar, 0.15), xytext=(peak_mu+2, 0.15),
            arrowprops=dict(color='black', arrowstyle='->', lw=2),
            fontsize=12, fontweight='bold')

# 解释框 (文案已修正)
text_info = (rf"为什么是 n-1 ?" + "\n"
             rf"1. 蓝山用的是真值 $\mu$，波动是纯粹的。" + "\n"
             rf"2. 红山用的是样本均值 $\bar{{X}}$，距离平方和必然变小。" + "\n"
             rf"3. 红色分布只能用 n-1 才能完美拟合。")

ax.text(0.38, 0.6, text_info, transform=ax.transAxes, fontsize=11,
        bbox=dict(boxstyle='round', facecolor='wheat', alpha=0.4))

ax.legend(loc='upper right', fontsize=11)
ax.grid(True, linestyle=':', alpha=0.5)

# 保存
output_dir = "Mathematical_statistics/images"
if not os.path.exists(output_dir):
    os.makedirs(output_dir)
save_path = f"{output_dir}/chi_squared_dof_loss_simulation.png"
plt.savefig(save_path, dpi=300)

print(f"图片已保存: {save_path}")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
import os

# ==============================================================================
# 核心笔记：小样本 vs 大样本 (The Effect of Sample Size)
# ==============================================================================
# 1. 峰值的意义 (The Peak):
#    - 峰值越高，说明数据越"抱团"。
#    - 因为红线(样本均值)的波动被"偷走"了一部分，分布范围变窄(瘦)，
#      为了保持总面积为1，它必须变高。这就是"瘦高"的来源。
#
# 2. 自由度的稀释 (Dilution of DOF):
#    - 小样本 (n=6): 丢失 1 个自由度意味着丢失了 1/6 (16.7%) 的信息。差异巨大。
#    - 大样本 (n=50): 丢失 1 个自由度只意味着丢失 1/50 (2%) 的信息。差异微乎其微。
# ==============================================================================

# ==========================================
# 0. 数据构造 (构造两组对比数据)
# ==========================================
np.random.seed(2025)
n_sims = 100000
mu, sigma = 50, 10

# 场景 1: 小样本
n_small = 6
X_small = np.random.normal(mu, sigma, (n_sims, n_small))
# 上帝视角 (df=n)
chi2_mu_small = np.sum((X_small - mu)**2, axis=1) / sigma**2
# 凡人视角 (df=n-1)
xbar_small = np.mean(X_small, axis=1, keepdims=True)
chi2_xbar_small = np.sum((X_small - xbar_small)**2, axis=1) / sigma**2

# 场景 2: 大样本
n_large = 50
X_large = np.random.normal(mu, sigma, (n_sims, n_large))
# 上帝视角 (df=n)
chi2_mu_large = np.sum((X_large - mu)**2, axis=1) / sigma**2
# 凡人视角 (df=n-1)
xbar_large = np.mean(X_large, axis=1, keepdims=True)
chi2_xbar_large = np.sum((X_large - xbar_large)**2, axis=1) / sigma**2

# ==========================================
# 绘图部分
# ==========================================
plt.rcParams['font.sans-serif'] = ['SimHei']
plt.rcParams['axes.unicode_minus'] = False

fig, axes = plt.subplots(2, 1, figsize=(12, 10))
plt.subplots_adjust(hspace=0.4)

# 颜色
c_mu = '#3498DB'    # 蓝
c_xbar = '#E74C3C'  # 红

# -----------------------------------------------------------
# 子图 1: 小样本 (n=6) - 差异显著
# -----------------------------------------------------------
ax1 = axes[0]
x1 = np.linspace(0, 20, 1000)

# 绘制
sns.histplot(chi2_mu_small, bins=80, stat="density", color=c_mu, alpha=0.3, element="step",
             label=rf'基于总体均值 $\mu$ (df=n)', ax=ax1)
sns.histplot(chi2_xbar_small, bins=80, stat="density", color=c_xbar, alpha=0.5, element="step",
             label=rf'基于样本均值 $\bar{{X}}$ (df=n-1)', ax=ax1)

# 理论线
ax1.plot(x1, stats.chi2.pdf(x1, n_small), color=c_mu, lw=2, ls='--')
ax1.plot(x1, stats.chi2.pdf(x1, n_small-1), color=c_xbar, lw=2)

ax1.set_title(rf'场景A: 小样本 ($n={n_small}$)', fontsize=16, fontweight='bold')
ax1.set_xlim(0, 20)
ax1.set_ylabel('概率密度 (变高!)')

# 标注峰值差异
peak_x = n_small - 3 # 大致位置
ax1.annotate('明显的错位\n(16.7%的差距)', xy=(peak_x, 0.15), xytext=(peak_x+3, 0.15),
             arrowprops=dict(color='black', arrowstyle='->', lw=2), fontsize=12)

# 在图例中明确显示 n 和 n-1
ax1.legend([rf'基准: $\mu$ ({n_small})', rf'实际: $\bar{{X}}$ ({n_small}-1)'], loc='upper right', fontsize=11)


# -----------------------------------------------------------
# 子图 2: 大样本 (n=50) - 差异消失
# -----------------------------------------------------------
ax2 = axes[1]
x2 = np.linspace(20, 90, 1000) # 卡方均值在 n=50 附近

# 绘制
sns.histplot(chi2_mu_large, bins=80, stat="density", color=c_mu, alpha=0.3, element="step",
             label=rf'基于总体均值 $\mu$', ax=ax2)
sns.histplot(chi2_xbar_large, bins=80, stat="density", color=c_xbar, alpha=0.5, element="step",
             label=rf'基于样本均值 $\bar{{X}}$', ax=ax2)

# 理论线
ax2.plot(x2, stats.chi2.pdf(x2, n_large), color=c_mu, lw=2, ls='--')
ax2.plot(x2, stats.chi2.pdf(x2, n_large-1), color=c_xbar, lw=2)

ax2.set_title(rf'场景B: 大样本 ($n={n_large}$)', fontsize=16, fontweight='bold')
ax2.set_xlim(20, 80)
ax2.set_xlabel(r'$\chi^2$ 统计量值')

# 标注重合
ax2.text(0.47, 0.45, "几乎重合！\n因为 $\\bar{X} \\approx \\mu$\n且 1/50 的差距微乎其微", 
         transform=ax2.transAxes, ha='center', fontsize=12,
         bbox=dict(boxstyle='round', facecolor='white', alpha=0.8))

ax2.legend([rf'基准: $\mu$ ({n_large})', rf'实际: $\bar{{X}}$ ({n_large}-1)'], loc='upper right', fontsize=11)

# 保存
output_dir = "Mathematical_statistics/images"
if not os.path.exists(output_dir):
    os.makedirs(output_dir)
save_path = f"{output_dir}/chi_squared_sample_size_effect.png"
plt.savefig(save_path, dpi=300)

print(f"图片已保存: {save_path}")
plt.show()
```

```python
# ==============================================================================
# 核心笔记：峰值移动的物理意义 (The Physics of Peak Shifting)
# ==============================================================================
# 1. 坐标轴的本质：
#    - 横坐标 (X轴) 代表“波动总量” (Sum of Squares)。
#    - 峰值的位置告诉我们：这组数据的“总变异程度”大概集中在什么数值附近。
#
# 2. 左移 (Left Shift) -> 代表“缩水”与“低估”
#    - 场景：从基于真值 μ (n) 切换到 基于样本均值 X_bar (n-1)。
#    - 原因 (最小二乘原理)：X_bar 是数据的几何重心。数学上，数据点离"重心"的
#      距离平方和，一定小于离"靶心(μ)"的距离平方和。
#    - 直觉：因为 X_bar "迎合"了数据，导致算出来的总波动比真实的要小。
#      (这就是为什么我们需要 n-1：因为分子变小了，分母也要变小来补偿)。
#
# 3. 右移 (Right Shift) -> 代表“累加”与“积攒”
#    - 场景：样本量 n 增大 (例如从 n=5 增加到 n=50)。
#    - 原因：卡方分布本质是误差的累加 (Sum of Z^2)。
#    - 直觉：累加的项数越多，总和自然越大。峰值右移是因为"积少成多"。
#
# 4. 变高 (Taller Peak) -> 代表“更集中”
#    - 场景：对比 n 和 n-1 时，红线(n-1) 比 蓝线(n) 更尖锐。
#    - 原因 (面积守恒)：总概率面积必须为 1。
#    - 直觉：因为 X_bar 离数据更近，波动范围变小(变瘦)了；为了维持面积不变，
#      概率密度必须在中心堆积得更高。
# ==============================================================================
```

```python
### 2.4. 样本均值的分布
```

```python
# ==============================================================================
# 核心笔记：t统计量的诞生 (The Birth of the t-Statistic)
# ==============================================================================
# 1. 核心定义：T = (X_bar - μ) / (S/√n) ~ t(n-1)
#    - 这是从 Z统计量 进化而来的形态。
#    - 唯一的改动：用样本标准差 S 替换了未知的总体标准差 σ。
#
# 2. 物理意义：尺子的形变 (The Wobbly Ruler)
#    - Z统计量用的是 σ (钢尺)，刻度恒定。
#    - t统计量用的是 S (皮尺)，每次抽样 S 都在变。
#    - 后果：用一把忽长忽短的尺子去量距离，测量结果(T值)的波动必然更大。
#    - 表现：为了包容这种波动，t分布必须是"厚尾"的(对极端值更宽容)。
#
# 3. 自由度 n-1 的来源
#    - S 的计算依赖于 X_bar。
#    - 既然 S^2 服从自由度为 n-1 的卡方分布，
#    - 那么由 S 构建出的 t分布，其自由度自然也必须是 n-1。
#
# 4. 终极关联 (The Grand Link)
#    - t分布 = 标准正态 / sqrt(卡方/自由度)
#    - 这完美连接了三大分布：正态分布提供中心，卡方分布提供分母的变异性。
# ==============================================================================
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
import os

# ==============================================================================
# 核心笔记：t统计量的诞生与厚尾成因 (The Birth of the t-Statistic)
# ==============================================================================
# 1. 对照实验：
#    - Z = (X_bar - μ) / (σ/√n) : 分母恒定。上帝视角，服从 N(0,1)。
#    - T = (X_bar - μ) / (S/√n) : 分母波动。凡人视角，服从 t(n-1)。
#
# 2. 核心机制：不稳定的尺子
#    - S 是随机变量。当样本量小(n=4)时，S 经常会算得很小 (S << σ)。
#    - 在公式中，S 做分母。一旦 S 变小，整个 T 值就会“爆炸”变大。
#
# 3. 视觉证据：
#    - S 的波动直接导致了 T 值的发散。
#    - 这种由 S 带来的额外波动，在分布图上就体现为“厚尾”(Fat Tails)。
# ==============================================================================

# ==========================================
# 0. 数据构造 (Construction)
# ==========================================
np.random.seed(2025)
n_sims = 50000
n = 4        # 极小样本，放大 S 的不稳定性
mu = 50
sigma = 10

# 生成样本 (N, n)
samples = np.random.normal(mu, sigma, (n_sims, n))

# 计算统计量组件
x_bar = np.mean(samples, axis=1)
# 注意：ddof=1 确保计算的是样本标准差 S (无偏估计)，对应 df=n-1
s_sample = np.std(samples, axis=1, ddof=1) 

# 1. 计算 Z (钢尺)
# 分母是固定的理论标准误
se_true = sigma / np.sqrt(n)
Z_stats = (x_bar - mu) / se_true

# 2. 计算 T (皮尺)
# 分母是浮动的样本标准误 S/√n
se_sample = s_sample / np.sqrt(n)
T_stats = (x_bar - mu) / se_sample

# ==========================================
# 绘图部分 (Visualization)
# ==========================================
plt.rcParams['font.sans-serif'] = ['SimHei']
plt.rcParams['axes.unicode_minus'] = False

fig, axes = plt.subplots(1, 3, figsize=(18, 6))

# 颜色定义
c_z = '#2ECC71'    # 正态/Z (绿)
c_t = '#E74C3C'    # t分布 (红)
c_s = '#F39C12'    # S分布 (橙)

# --- 图1: 尺子的对比 (S vs Sigma) ---
ax1 = axes[0]
sns.histplot(s_sample, bins=50, stat='density', color=c_s, alpha=0.6, label='样本标准差 S', ax=ax1)
ax1.axvline(sigma, color='black', linestyle='--', linewidth=3, label=r'总体标准差 $\sigma$')

ax1.set_title(r'1. 不稳定的尺子: $S$ vs $\sigma$', fontsize=14, fontweight='bold')
ax1.set_xlabel('标准差数值')
ax1.legend()
ax1.text(0.05, 0.8, "S 经常偏小\n导致分母失效", transform=ax1.transAxes, 
         bbox=dict(boxstyle='round', facecolor='white', alpha=0.8))

# --- 图2: 散点关联 (S 对 T 的影响) ---
ax2 = axes[1]
# 为了看清规律，我们只画前 2000 个点
ax2.scatter(s_sample[:2000], T_stats[:2000], alpha=0.4, color='#8E44AD', s=10)
ax2.axvline(sigma, color='black', linestyle='--', alpha=0.5)
ax2.axhline(0, color='gray', linestyle='-', alpha=0.5)

ax2.set_title(r'2. 爆炸机制: S 越小，T 越飞', fontsize=14, fontweight='bold')
ax2.set_xlabel(r'样本标准差 $S$ (分母)')
ax2.set_ylabel(r'统计量 $T$ (结果)')
ax2.set_ylim(-10, 10)
# 标注高风险区
ax2.annotate('危险区!\nS过小导致T极大', xy=(2, 6), xytext=(5, 8),
             arrowprops=dict(facecolor='red', arrowstyle='->'), fontsize=11, color='red')

# --- 图3: 最终分布 (Z vs T) ---
ax3 = axes[2]
x_axis = np.linspace(-6, 6, 1000)

# 画 Z 的直方图 (作为背景对比)
sns.histplot(Z_stats, bins=80, stat='density', color=c_z, alpha=0.3, label='Z 统计量 (正态)', ax=ax3, element="step")
# 画 T 的直方图
sns.histplot(T_stats, bins=80, stat='density', color=c_t, alpha=0.5, label='T 统计量 (t分布)', ax=ax3)

# 理论线
# Z 理论线
ax3.plot(x_axis, stats.norm.pdf(x_axis), color=c_z, ls='--', lw=2)
# T 理论线 (df=n-1 = 3)
ax3.plot(x_axis, stats.t.pdf(x_axis, df=n-1), color=c_t, lw=3)

ax3.set_title(r'3. 最终后果: 厚尾的形成', fontsize=14, fontweight='bold')
ax3.set_xlim(-6, 6)
ax3.legend()
ax3.annotate('厚尾\n(极端值增多)', xy=(3.5, 0.05), xytext=(4, 0.2),
             arrowprops=dict(facecolor=c_t, arrowstyle='->'), color=c_t, fontweight='bold')

# 全局布局
plt.suptitle(f't分布的物理起源：当分母 $S$ 开始波动 (样本量 n={n})', fontsize=18, y=0.98)
plt.subplots_adjust(top=0.85, wspace=0.25, bottom=0.15)

# 保存
output_dir = "Mathematical_statistics/images"
if not os.path.exists(output_dir):
    os.makedirs(output_dir)
save_path = f"{output_dir}/t_statistic_mechanism.png"
plt.savefig(save_path, dpi=300)

print(f"图片已保存: {save_path}")
plt.show()
```

```python
# ==============================================================================
# 三图合一总结：为什么 t分布会有“厚尾”？(The Story of Fat Tails)
# ==============================================================================
# 1. 祸起萧墙 (图1)：样本标准差 S 是不稳定的“皮尺”。在小样本下，它经常会算出
#    比真值 σ 小很多的数值 (即 S 被严重低估)。
#
# 2. 连锁反应 (图2)：因为 S 位于公式的分母 (T = Δ / S)，当 S 偶然变得极小时，
#    整个分数值 T 会瞬间“爆炸”，产生巨大的极端值 (散点图中的喷射状)。
#
# 3. 最终代价 (图3)：这些因分母过小而产生的“人造极端值”，最终堆积在了分布的两端，
#    这就强行把原本苗条的正态分布撑成了“矮胖”的 t分布 (即厚尾现象)。
# ==============================================================================
```

```python
### 2.5. 样本均值方差的正态分布
```

```python
# ==============================================================================
# 核心笔记：样本均值差的分布 (Distribution of the Difference of Means)
# ==============================================================================
#
# 1. 核心公式：(X_bar - Y_bar) ~ N(μ1-μ2, σ1²/n1 + σ2²/n2)
#    - 这是双样本 Z检验 (Two-sample Z-test) 的理论基础。
#
# 2. 直觉陷阱：为什么方差是相加的 (+) ？
#    - 数学原理：Var(A - B) = Var(A) + Var(B) (当 A, B 独立时)。
#    - 物理直觉：不确定性(混乱)只会累积，不会抵消。
#      引入第二个变量 Y，就引入了第二份噪声。无论做加法还是减法，总的噪声(波动)都变大了。
#    - 就像两辆车相向而行(做减法计算距离)，它们相对位置的波动是两人抖动程度之和。
#
# 3. 联合标准误 (Standard Error of Difference)
#    - SE_diff = sqrt(σ1²/n1 + σ2²/n2)
#    - 这是一个"合成"的尺子。它比单独测量 X 或 Y 时都要长，
#      意味着判断"两者是否有差异"时，我们需要更大的安全边际。
# ==============================================================================
```

```python
print('''
步骤,关键操作,核心逻辑 / 统计直觉
1. 构造,D = X_bar - Y_bar,制造差异：从两个总体分别抽样计算均值，然后相减。这是我们要研究的对象。
2. 假设(错),Var_wrong = Var1 - Var2,直觉陷阱：如果我们认为"减法会抵消波动"，得到的分布(黑虚线)非常窄。
3. 假设(对),Var_correct = Var1 + Var2,累积原理：承认"多一个变量多一份乱"，方差叠加，分布(红实线)变宽。
4. 验证,直方图 vs 曲线,视觉判决：宽阔的模拟数据(紫块)撑破了黑色虚线，但被红色实线完美包裹。证明方差是相加的。
''')
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
import os

# ==============================================================================
# 核心笔记：双样本均值差的分布 (Distribution of Difference)
# ==============================================================================
# 1. 核心现象：方差膨胀 (Variance Inflation)
#    - 当我们计算 D = X_bar - Y_bar 时，结果的波动范围比单独的 X_bar 或 Y_bar 都要大。
#
# 2. 直觉对比：加法 vs 减法
#    - 均值做减法 (Location Shift)：中心从 100 和 90 移到了 10。
#    - 方差做加法 (Uncertainty Accumulation)：波动没有抵消，反而叠加了。
#
# 3. 视觉证据：
#    - "错误直觉线"(假设方差相减) 会非常瘦高。
#    - "真实分布"(方差相加) 会明显更矮胖。这证明了混乱是累积的。
# ==============================================================================

# ==========================================
# 0. 数据构造 (Construction)
# ==========================================
np.random.seed(2025)
n_sims = 100000
n1, n2 = 25, 25

# 设定两个总体
# 总体1: 波动大一点
mu1, sigma1 = 100, 15 
# 总体2: 波动小一点
mu2, sigma2 = 90, 10  

# 1. 模拟独立的样本均值
# X_bar ~ N(100, 15^2/25) -> SE1 = 3
# Y_bar ~ N(90, 10^2/25)  -> SE2 = 2
means1 = np.mean(np.random.normal(mu1, sigma1, (n_sims, n1)), axis=1)
means2 = np.mean(np.random.normal(mu2, sigma2, (n_sims, n2)), axis=1)

# 2. 核心操作：计算差值
diff_sim = means1 - means2

# 3. 理论参数计算
mu_diff_theo = mu1 - mu2 # 10
# 正确的方差 (相加)
var_correct = (sigma1**2 / n1) + (sigma2**2 / n2)
se_correct = np.sqrt(var_correct) # sqrt(9 + 4) = sqrt(13) ≈ 3.6

# 错误的方差 (相减) - 仅用于教学展示
var_wrong = abs((sigma1**2 / n1) - (sigma2**2 / n2))
se_wrong = np.sqrt(var_wrong) # sqrt(9 - 4) = sqrt(5) ≈ 2.2

# ==========================================
# 绘图部分 (Visualization)
# ==========================================
plt.rcParams['font.sans-serif'] = ['SimHei'] 
plt.rcParams['axes.unicode_minus'] = False

fig, ax = plt.subplots(figsize=(12, 7))

# 颜色
c_hist = '#a29bfe'    # 模拟直方图 (紫)
c_correct = '#d63031' # 正确理论 (红)
c_wrong = '#2d3436'   # 错误直觉 (灰黑)

x = np.linspace(mu_diff_theo - 4*se_correct, mu_diff_theo + 4*se_correct, 1000)

# 1. 绘制模拟数据的直方图 (现实)
sns.histplot(diff_sim, bins=80, stat="density", color=c_hist, edgecolor='white', alpha=0.6, 
             label=r'模拟差值 $\bar{X} - \bar{Y}$', ax=ax)

# 2. 绘制正确理论线 (方差相加)
y_correct = stats.norm.pdf(x, mu_diff_theo, se_correct)
ax.plot(x, y_correct, color=c_correct, linewidth=3, 
        label=r'正确理论: $\sigma_{diff}^2 = SE_1^2 + SE_2^2$ (矮胖)')

# 3. 绘制错误直觉线 (方差相减)
y_wrong = stats.norm.pdf(x, mu_diff_theo, se_wrong)
ax.plot(x, y_wrong, color=c_wrong, linewidth=2, linestyle='--', 
        label=r'错误直觉: $\sigma_{diff}^2 = |SE_1^2 - SE_2^2|$ (瘦高)')

# 装饰
ax.set_title(rf'双样本均值差的分布：为什么方差是相加的？', fontsize=16, fontweight='bold')
ax.set_xlabel('均值之差')
ax.set_ylabel('概率密度')

# 核心标注
ax.text(0.05, 0.75, "错误直觉(虚线)：\n认为抵消了波动，\n结果分布太窄，\n包不住数据！", 
        transform=ax.transAxes, fontsize=12, color=c_wrong,
        bbox=dict(boxstyle='round', facecolor='white', edgecolor=c_wrong, alpha=0.8))

ax.text(0.75, 0.65, "正确理论(红线)：\n承认波动累积，\n分布变宽，\n完美拟合！", 
        transform=ax.transAxes, fontsize=12, color=c_correct, fontweight='bold',
        bbox=dict(boxstyle='round', facecolor='white', edgecolor=c_correct, alpha=0.8))

ax.legend(loc='upper right', fontsize=11)
ax.grid(True, linestyle=':', alpha=0.5)

# 保存
output_dir = "Mathematical_statistics/images"
if not os.path.exists(output_dir):
    os.makedirs(output_dir)
save_path = f"{output_dir}/difference_of_means_variance_additivity.png"
plt.savefig(save_path, dpi=300)

print(f"图片已保存: {save_path}")
plt.show()
```

```python
### 2.6. 方差相等时样本均值差的t分布(两总体标准差未知但相等)
```

```python
# ==============================================================================
# 核心笔记：合并方差 t检验 (Pooled t-test)
# ==============================================================================
# 1. 核心前提：同方差假设 (Homogeneity of Variance)
#    - 假设两个总体虽然"中心"不同，但"胖瘦"一致 (σ1 = σ2 = σ)。
#    - 这是使用此公式的入场券。如果方差不齐，必须使用 Welch's t-test。
#
# 2. 核心操作：信息的融合 (Pooling)
#    - Sw (Pooled Std Dev) 是 S1 和 S2 的"加权平均"。
#    - 逻辑：既然大家都在估计同一个 σ，那就把所有数据凑一起算，样本量大的话语权大。
#    - 收益：获得了一个比单独 S1 或 S2 都更精准的 σ 估计值。
#
# 3. 自由度红利：df = n1 + n2 - 2
#    - 相比于方差不齐时的 Welch 近似自由度，合并方差能最大限度地保留自由度。
#    - 结果：t分布的尾巴更薄，置信区间更窄，更容易检测出微小的均值差异。
# ==============================================================================
```

```python
print('''
步骤,关键操作,核心逻辑 / 统计直觉
1. 场景,σ1=σ2,前提：上帝告诉我们两个总体的方差是一样的。
2. 左图(A),使用 Sw,信息融合：我们利用这个信息，计算加权平均方差 Sw。结果直方图完美贴合最大自由度的理论红线 t(n1+n2-2)。
3. 右图(B),使用 S1, S2,各管各的：我们忽略这个信息，像做 Z检验一样直接组合标准误。结果直方图比理论红线略微宽一点(厚尾)。
4. 结论,对比差异,效率差异：虽然B方法也没错(它是Welch t的基础)，但在方差相等时，A方法(合并)能获得更窄的分布和更高的检验效率。
''')
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
import os

# ==============================================================================
# 核心笔记：合并方差 vs 直接相加 (Pooled vs Direct/Unpooled)
# ==============================================================================
# 前提：已知两个总体方差相等 (σ1 = σ2)。
#
# 1. 合并思路 (Pooled Approach):
#    - 操作：计算加权平均方差 Sw。
#    - 逻辑：充分利用"方差相等"这一信息，集中数据估计一个更准的 σ。
#    - 结果：获得最大的自由度 (n1+n2-2)，是最优解。
#
# 2. 直接思路 (Direct Approach):
#    - 操作：分别计算 S1 和 S2，然后直接组合标准误 sqrt(S1²/n1 + S2²/n2)。
#    - 逻辑：忽略了"方差相等"的信息，各算各的。
#    - 结果：虽然没错(这是Welch t检验的形式)，但在方差相等时效率较低，
#           有效自由度会小于 n1+n2-2，导致分布比最优解略宽(厚尾)。
# ==============================================================================

# ==========================================
# 0. 数据构造 (Construction)
# ==========================================
np.random.seed(2025)
n_sims = 10000

# 设定参数 (关键：方差必须相等)
sigma = 10
mu1, n1 = 100, 10
mu2, n2 = 105, 15  # 样本量不同，差异会更明显

# 生成样本
X = np.random.normal(mu1, sigma, (n_sims, n1))
Y = np.random.normal(mu2, sigma, (n_sims, n2))

# 计算基础零件
x_bar = np.mean(X, axis=1)
y_bar = np.mean(Y, axis=1)
s1_sq = np.var(X, axis=1, ddof=1)
s2_sq = np.var(Y, axis=1, ddof=1)
theoretical_diff = mu1 - mu2

# --- 路线 A: 合并方差思路 (Pooled) ---
df_total = n1 + n2 - 2
sw_sq = ((n1 - 1) * s1_sq + (n2 - 1) * s2_sq) / df_total
sw = np.sqrt(sw_sq)
se_pooled = sw * np.sqrt(1/n1 + 1/n2)
t_pooled = ((x_bar - y_bar) - theoretical_diff) / se_pooled

# --- 路线 B: 直接相加思路 (Direct/Unpooled) ---
# 类似于 Z 检验或 Welch t 检验的结构
se_direct = np.sqrt(s1_sq/n1 + s2_sq/n2)
t_direct = ((x_bar - y_bar) - theoretical_diff) / se_direct

# ==========================================
# 绘图部分 (Visual Comparison)
# ==========================================
plt.rcParams['font.sans-serif'] = ['SimHei'] 
plt.rcParams['axes.unicode_minus'] = False

fig, axes = plt.subplots(1, 2, figsize=(16, 7))

# 颜色
c_pool = '#0984e3'   # 合并 (蓝)
c_direct = '#6c5ce7' # 直接 (紫)
c_theo = '#d63031'   # 理论曲线 (红)

x_axis = np.linspace(-6, 6, 1000)
# 共同的理论基准：最大自由度的 t 分布
y_theo_best = stats.t.pdf(x_axis, df=df_total)

# --- 左图：合并思路 (Pooled) ---
ax1 = axes[0]
sns.histplot(t_pooled, bins=80, stat="density", color=c_pool, edgecolor='white', alpha=0.6, 
             label='模拟 T (Pooled)', ax=ax1)
ax1.plot(x_axis, y_theo_best, color=c_theo, linewidth=3, linestyle='--',
         label=rf'基准理论 $t({df_total})$')

ax1.set_title(rf'A. 合并方差思路 (最优解)', fontsize=14, fontweight='bold')
ax1.set_xlabel('T 统计量')
ax1.set_ylabel('概率密度')
ax1.set_xlim(-6, 6)
ax1.legend(loc='upper right')
ax1.text(0.05, 0.85, "完美贴合！\n充分利用了信息\n自由度最大化", transform=ax1.transAxes,
         bbox=dict(boxstyle='round', facecolor=c_pool, alpha=0.2))

# --- 右图：直接思路 (Direct) ---
ax2 = axes[1]
sns.histplot(t_direct, bins=80, stat="density", color=c_direct, edgecolor='white', alpha=0.6, 
             label='模拟 T (Direct)', ax=ax2)
# 注意：这里画的是同一条理论线，用来做对比参照
ax2.plot(x_axis, y_theo_best, color=c_theo, linewidth=3, linestyle='--',
         label=rf'基准理论 $t({df_total})$')

ax2.set_title(rf'B. 直接相加思路 (次优解)', fontsize=14, fontweight='bold')
ax2.set_xlabel('T 统计量')
ax2.set_xlim(-6, 6)
ax2.legend(loc='upper right')
ax2.text(0.05, 0.85, "轻微偏差！\n(尾巴比红线胖一点)\n浪费了自由度", transform=ax2.transAxes,
         bbox=dict(boxstyle='round', facecolor=c_direct, alpha=0.2))

# 全局标题
plt.suptitle(f'方差相等时的两种策略对比：信息融合的价值 (n1={n1}, n2={n2})', fontsize=16, y=0.98)

# 保存
output_dir = "Mathematical_statistics/images"
if not os.path.exists(output_dir):
    os.makedirs(output_dir)
save_path = f"{output_dir}/pooled_vs_direct_comparison.png"
plt.savefig(save_path, dpi=300)

print(f"图片已保存: {save_path}")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
import os

# ==============================================================================
# 核心笔记：双样本 t检验的效率对比 (Pooled vs. Direct)
# ==============================================================================
# 目标：通过 QQ 图放大观察，证实 Direct 方法在方差相等时效率较低 (厚尾)。
# ==============================================================================

# ==========================================
# 0. 数据构造 (Construction)
# ==========================================
np.random.seed(2025)
n_sims = 100000 
sigma = 10
mu1, n1 = 100, 10
mu2, n2 = 105, 15 

# 生成样本
X = np.random.normal(mu1, sigma, (n_sims, n1))
Y = np.random.normal(mu2, sigma, (n_sims, n2))

x_bar = np.mean(X, axis=1)
y_bar = np.mean(Y, axis=1)
s1_sq = np.var(X, axis=1, ddof=1)
s2_sq = np.var(Y, axis=1, ddof=1)
theoretical_diff = mu1 - mu2

# --- Pooled (最优解) ---
df_total = n1 + n2 - 2
sw_sq = ((n1 - 1) * s1_sq + (n2 - 1) * s2_sq) / df_total
se_pooled = np.sqrt(sw_sq) * np.sqrt(1/n1 + 1/n2)
t_pooled = ((x_bar - y_bar) - theoretical_diff) / se_pooled

# --- Direct (次优解) ---
se_direct = np.sqrt(s1_sq/n1 + s2_sq/n2)
t_direct = ((x_bar - y_bar) - theoretical_diff) / se_direct

# ==========================================
# 绘图部分 (2x2 Full Verification)
# ==========================================
plt.rcParams['font.sans-serif'] = ['SimHei'] 
plt.rcParams['axes.unicode_minus'] = False

fig, axes = plt.subplots(2, 2, figsize=(16, 12))
axes = axes.flatten()

# 颜色
c_pool = '#0984e3'   # 合并 (蓝)
c_direct = '#6c5ce7' # 直接 (紫)
c_theo = '#d63031'   # 理论基准 (红)

x_axis = np.linspace(-6, 6, 1000)
y_theo_best = stats.t.pdf(x_axis, df=df_total)

# -----------------------------------------------------------
# TOP ROW: 密度对比 (Histograms)
# -----------------------------------------------------------

# A. 左上：Pooled vs 理论
ax1 = axes[0]
sns.histplot(t_pooled, bins=80, stat="density", color=c_pool, edgecolor='white', alpha=0.6, 
             label='模拟 T (Pooled)', ax=ax1)
ax1.plot(x_axis, y_theo_best, color=c_theo, linewidth=3, linestyle='--',
         label=rf'理论 $t({df_total})$')
ax1.set_title(rf'A. 合并方差 (最优解) - 完美拟合', fontsize=14)
ax1.set_xlim(-6, 6)
ax1.set_ylabel('概率密度')
ax1.legend()

# B. 右上：Direct vs 理论
ax2 = axes[1]
sns.histplot(t_direct, bins=80, stat="density", color=c_direct, edgecolor='white', alpha=0.6, 
             label='模拟 T (Direct)', ax=ax2)
ax2.plot(x_axis, y_theo_best, color=c_theo, linewidth=3, linestyle='--',
         label=rf'理论 $t({df_total})$')
ax2.set_title(rf'B. 直接相加 (次优解) - 宏观难察觉', fontsize=14)
ax2.set_xlim(-6, 6)
ax2.legend()
ax2.text(0.2, 0.85, "宏观上，两者难以区分", transform=ax2.transAxes, ha='center',
         bbox=dict(boxstyle='round', facecolor='yellow', alpha=0.3))

# -----------------------------------------------------------
# BOTTOM ROW: QQ 分位数对比 (Tail Diagnostics)
# -----------------------------------------------------------

# 【修正点】：当 fit=False 时，直接用两个变量接收返回值，不要嵌套解包
osm_p, osr_p = stats.probplot(t_pooled, dist="t", sparams=(df_total,), fit=False)
osm_d, osr_d = stats.probplot(t_direct, dist="t", sparams=(df_total,), fit=False)

# C. 左下：Pooled QQ
ax3 = axes[2]
ax3.plot(osm_p, osr_p, '.', color=c_pool, alpha=0.4, markersize=4, label='Pooled Data')
# 绘制基准线
lims = [np.min([osm_p, osm_d]), np.max([osm_p, osm_d])]
ax3.plot(lims, lims, color=c_theo, linewidth=2, linestyle='--', label=rf'理论基准 $t({df_total})$')

ax3.set_title(rf'C. Pooled QQ 图 (完美贴合)', fontsize=14)
ax3.set_xlabel(f'理论分位数 $t({df_total})$')
ax3.set_ylabel('模拟实际值')
ax3.text(0, -4, "紧贴直线，效率最优", color=c_pool, fontweight='bold', ha='center')
ax3.grid(True, linestyle=':', alpha=0.5)

# D. 右下：Direct QQ
ax4 = axes[3]
ax4.plot(osm_d, osr_d, '.', color=c_direct, alpha=0.4, markersize=4, label='Direct Data')
ax4.plot(lims, lims, color=c_theo, linewidth=2, linestyle='--', label=rf'理论基准 $t({df_total})$')

ax4.set_title(rf'D. Direct QQ 图 (尾部偏差)', fontsize=14)
ax4.set_xlabel(f'理论分位数 $t({df_total})$')

# 关键标注：尾部发散
ax4.annotate('明显的厚尾偏差\n(S型发散)', xy=(-3.5, -4.5), xytext=(-5, -2),
             arrowprops=dict(facecolor=c_direct, arrowstyle='->', lw=1.5), fontsize=12, color='red')
ax4.annotate('', xy=(3.5, 4.5), xytext=(5, 2),
             arrowprops=dict(facecolor=c_direct, arrowstyle='->', lw=1.5), fontsize=12, color='red')
ax4.grid(True, linestyle=':', alpha=0.5)

# 全局标题
plt.suptitle(f'方差相等时的 $t$ 检验策略对比：QQ图揭示效率损失 (n1={n1}, n2={n2})', fontsize=18, y=0.95)

# 保存
output_dir = "Mathematical_statistics/images"
if not os.path.exists(output_dir):
    os.makedirs(output_dir)
save_path = f"{output_dir}/pooled_vs_direct_qq_comparison_2x2.png"
plt.savefig(save_path, dpi=300)

print(f"图片已保存: {save_path}")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
from mpl_toolkits.axes_grid1.inset_locator import inset_axes, mark_inset
import os

# ==============================================================================
# 核心笔记：放大镜下的真相 (Magnifying the Deviation)
# ==============================================================================
# 1. 为什么平时看不清？
#    - 大部分数据点挤在中心 (-2, 2) 的直线上，视觉权重过大，掩盖了边缘的偏差。
#
# 2. 我们的手段：人为控制坐标 (Manual Axis Control)
#    - 我们不再看全局，而是把镜头强行推到 X轴 [2.5, 5.0] 的区域。
#    - 在这里，Direct 方法产生的极端值会比理论预期大得多。
#
# 3. 视觉预期：
#    - Pooled (左下): 即使放大，点阵依然紧紧咬住红线。
#    - Direct (右下): 放大后，点阵会明显"飘"在红线上方。这就是效率损失的铁证。
# ==============================================================================

# ==========================================
# 0. 数据构造
# ==========================================
np.random.seed(2025)
n_sims = 100000 
sigma = 10
mu1, n1 = 100, 10
mu2, n2 = 105, 15 

# 生成样本
X = np.random.normal(mu1, sigma, (n_sims, n1))
Y = np.random.normal(mu2, sigma, (n_sims, n2))

# 基础计算
x_bar = np.mean(X, axis=1)
y_bar = np.mean(Y, axis=1)
s1_sq = np.var(X, axis=1, ddof=1)
s2_sq = np.var(Y, axis=1, ddof=1)
theoretical_diff = mu1 - mu2

# --- Pooled (最优解) ---
df_total = n1 + n2 - 2
sw_sq = ((n1 - 1) * s1_sq + (n2 - 1) * s2_sq) / df_total
se_pooled = np.sqrt(sw_sq) * np.sqrt(1/n1 + 1/n2)
t_pooled = ((x_bar - y_bar) - theoretical_diff) / se_pooled

# --- Direct (次优解) ---
se_direct = np.sqrt(s1_sq/n1 + s2_sq/n2)
t_direct = ((x_bar - y_bar) - theoretical_diff) / se_direct

# ==========================================
# 绘图部分 (2x2 with Hyper-Zoom)
# ==========================================
plt.rcParams['font.sans-serif'] = ['SimHei'] 
plt.rcParams['axes.unicode_minus'] = False

fig, axes = plt.subplots(2, 2, figsize=(16, 12))
axes = axes.flatten()

# 颜色
c_pool = '#0984e3'   # 蓝
c_direct = '#6c5ce7' # 紫
c_theo = '#d63031'   # 红

x_axis = np.linspace(-6, 6, 1000)
y_theo_best = stats.t.pdf(x_axis, df=df_total)

# -----------------------------------------------------------
# TOP ROW: 密度对比 (保持不变，作为宏观背景)
# -----------------------------------------------------------
# A. Pooled Hist
ax1 = axes[0]
sns.histplot(t_pooled, bins=80, stat="density", color=c_pool, edgecolor='white', alpha=0.6, ax=ax1)
ax1.plot(x_axis, y_theo_best, color=c_theo, lw=3, ls='--', label=rf'理论 $t({df_total})$')
ax1.set_title(rf'A. 合并方差 (宏观完美)', fontsize=14)
ax1.set_xlim(-6, 6)
ax1.legend()

# B. Direct Hist
ax2 = axes[1]
sns.histplot(t_direct, bins=80, stat="density", color=c_direct, edgecolor='white', alpha=0.6, ax=ax2)
ax2.plot(x_axis, y_theo_best, color=c_theo, lw=3, ls='--', label=rf'理论 $t({df_total})$')
ax2.set_title(rf'B. 直接相加 (宏观近似)', fontsize=14)
ax2.set_xlim(-6, 6)
ax2.legend()

# -----------------------------------------------------------
# BOTTOM ROW: QQ 对比 + 暴力放大镜
# -----------------------------------------------------------

# 计算 QQ 坐标
osm_p, osr_p = stats.probplot(t_pooled, dist="t", sparams=(df_total,), fit=False)
osm_d, osr_d = stats.probplot(t_direct, dist="t", sparams=(df_total,), fit=False)

# 统一的基准线范围
lims = [np.min([osm_p, osm_d]), np.max([osm_p, osm_d])]

# --- C. 左下：Pooled QQ (放大看也是直的) ---
ax3 = axes[2]
ax3.plot(osm_p, osr_p, '.', color=c_pool, alpha=0.2, markersize=3)
ax3.plot(lims, lims, color=c_theo, lw=2, ls='--') # 基准线

ax3.set_title(rf'C. Pooled 细节: 紧咬红线', fontsize=14)
ax3.set_xlabel(f'理论分位数')
ax3.set_ylabel('实际值')
ax3.grid(True, ls=':', alpha=0.5)

# [放大镜 C]
# 这是一个"画中画"，强制放大 x=[2.5, 5] 的区域
axins3 = inset_axes(ax3, width="40%", height="40%", loc='lower right', borderpad=3)
axins3.plot(osm_p, osr_p, '.', color=c_pool, alpha=0.6, markersize=5)
axins3.plot(lims, lims, color=c_theo, lw=3, ls='--')
# 人为控制坐标轴：聚焦尾部
axins3.set_xlim(2.5, 5.0)
axins3.set_ylim(2.5, 5.0) # 这里的 Y 轴范围要根据数据调整，Pooled 应该就在 x 附近
# 添加连线指示
mark_inset(ax3, axins3, loc1=2, loc2=4, fc="none", ec="0.5")
axins3.set_title("尾部特写", fontsize=10)


# --- D. 右下：Direct QQ (放大看就露馅了) ---
ax4 = axes[3]
ax4.plot(osm_d, osr_d, '.', color=c_direct, alpha=0.2, markersize=3)
ax4.plot(lims, lims, color=c_theo, lw=2, ls='--') # 基准线

ax4.set_title(rf'D. Direct 细节: 尾部上翘 (厚尾)', fontsize=14)
ax4.set_xlabel(f'理论分位数')
ax4.grid(True, ls=':', alpha=0.5)

# [放大镜 D] - 这是核心！
axins4 = inset_axes(ax4, width="40%", height="40%", loc='lower right', borderpad=3)
axins4.plot(osm_d, osr_d, '.', color=c_direct, alpha=0.6, markersize=5)
axins4.plot(lims, lims, color=c_theo, lw=3, ls='--') # 理论基准

# ！！！人为控制坐标轴：这里是关键！！！
# 我们聚焦在同样的 x=[2.5, 5.0] 区域
axins4.set_xlim(2.5, 5.0)
# 但 Direct 的 Y 值会比 5.0 大很多（因为厚尾），我们把 Y 轴拉高，展示它"飞出去"的样子
axins4.set_ylim(2.5, 6.5) 

# 添加连线指示
mark_inset(ax4, axins4, loc1=2, loc2=4, fc="none", ec="0.5")
axins4.set_title("尾部特写：飞离红线！", fontsize=10, color='red', fontweight='bold')

# 添加醒目的箭头
axins4.annotate('偏差巨大', xy=(3.5, 4.0), xytext=(2.8, 5.5),
                arrowprops=dict(facecolor='red', arrowstyle='->', lw=2),
                color='red', fontweight='bold')

# 全局标题
plt.suptitle(f'显微镜下的 $t$ 检验：强制放大尾部差异 (n1={n1}, n2={n2})', fontsize=18, y=0.95)

# 保存
output_dir = "Mathematical_statistics/images"
if not os.path.exists(output_dir):
    os.makedirs(output_dir)
save_path = f"{output_dir}/pooled_vs_direct_qq_zoomed.png"
plt.savefig(save_path, dpi=300)

print(f"图片已保存: {save_path}")
plt.show()
```

```python
# ==============================================================================
# 核心笔记：方差比的 F分布 (The F-Distribution for Variance Ratios)
# ==============================================================================
# 
# 1. 数学本质：卡方的碰撞
#    - 分子：样本1的方差 S1² 经过标准化 (除以 σ1²)。
#    - 分母：样本2的方差 S2² 经过标准化 (除以 σ2²)。
#    - F分布 = (Chi_1 / df_1) / (Chi_2 / df_2)。
#
# 2. 物理含义：相对波动的比值
#    - 它衡量了两个样本在各自总体背景下的"波动表现"。
#    - 这是一个纯粹的比率，与数据的单位无关。
#
# 3. 杀手级应用：方差齐性检验 (Homogeneity of Variance Test)
#    - 场景：我们想知道两个总体的波动是否一样 (H0: σ1² = σ2²)。
#    - 魔法：在 H0 成立的条件下，σ1² 和 σ2² 互相抵消。
#    - 结果：统计量简化为 F = S1² / S2²。
#    - 判据：如果 S1² / S2² 显著偏离 1 (落入 F分布的尾部)，则证明方差不等。
# ==============================================================================
```

```python
print('''
步骤,关键操作,核心逻辑 / 统计直觉
1. 场景A(左图),σ1≠σ2,强制修正：原始的方差比(灰)严重偏离理论，只有除以各自的 σ²(蓝) 才能把分布"拉回"到红线(理论F)。
2. 场景B(右图),σ1=σ2,自然契合：因为 σ² 上下抵消，原始的方差比(橙)直接就完美贴合红线。
3. 结论,F检验,上帝隐身：在检验"方差是否相等"时，我们不需要知道方差的具体数值，只需要看 S1²/S2² 是否落在 F分布的合理区间内。
''')
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
import os

# ==============================================================================
# 核心笔记：F分布的双重验证 (Definition vs Application)
# ==============================================================================
# 1. 定义的普适性 (左图):
#    - 即使 σ1 ≠ σ2 (比如一个是10, 一个是30)，只要公式里除以了各自的 σ²，
#      得到的 F值 依然乖乖服从 F(n1-1, n2-1)。
#    - 这一点保证了 F分布 作为理论基石的稳固性。
#
# 2. 应用的便捷性 (右图 - 方差齐性检验原理):
#    - 当我们假设 σ1 = σ2 时 (零假设 H0)，神奇的事情发生了。
#    - 公式中的 σ² 互相抵消。我们不需要知道 σ 到底是多少，
#      直接用 S1²/S2² 就能得到标准的 F分布。
# ==============================================================================

# ==========================================
# 0. 数据构造
# ==========================================
np.random.seed(2025)
n_sims = 100000
n1, n2 = 20, 15 # 自由度 df1=19, df2=14

# --- 场景 A: 混沌世界 (Sigma 不同) ---
sigma_A1, sigma_A2 = 10, 30
# 抽样
X_A = np.random.normal(100, sigma_A1, (n_sims, n1))
Y_A = np.random.normal(100, sigma_A2, (n_sims, n2))
# 算样本方差
s1_sq_A = np.var(X_A, axis=1, ddof=1)
s2_sq_A = np.var(Y_A, axis=1, ddof=1)

# 计算统计量
# 1. 原始比值 (Raw Ratio): S1^2 / S2^2
raw_ratio_A = s1_sq_A / s2_sq_A
# 2. 标准化比值 (True F): (S1^2/σ1^2) / (S2^2/σ2^2)
true_f_A = (s1_sq_A / sigma_A1**2) / (s2_sq_A / sigma_A2**2)


# --- 场景 B: 对称世界 (Sigma 相同) ---
sigma_B = 20 # 两个都是 20
# 抽样
X_B = np.random.normal(100, sigma_B, (n_sims, n1))
Y_B = np.random.normal(100, sigma_B, (n_sims, n2)) # 均值不重要，F只看方差
# 算样本方差
s1_sq_B = np.var(X_B, axis=1, ddof=1)
s2_sq_B = np.var(Y_B, axis=1, ddof=1)

# 计算统计量
# 在 H0 (σ1=σ2) 下，这两个计算方式数学上等价
true_f_B = s1_sq_B / s2_sq_B 


# ==========================================
# 绘图部分
# ==========================================
plt.rcParams['font.sans-serif'] = ['SimHei'] 
plt.rcParams['axes.unicode_minus'] = False

fig, axes = plt.subplots(1, 2, figsize=(16, 7))

# 理论 F 分布曲线
x = np.linspace(0, 5, 1000)
y_theo = stats.f.pdf(x, n1-1, n2-1)

# --- 左图：定义的验证 (σ1 ≠ σ2) ---
ax1 = axes[0]
# 1. 画"错误"的原始比值 (灰色)
sns.histplot(raw_ratio_A, bins=100, stat="density", color='gray', alpha=0.3, element="step",
             label=r'原始比值 $S_1^2/S_2^2$ (未修正)', ax=ax1)
# 2. 画"正确"的F统计量 (蓝色)
sns.histplot(true_f_A, bins=60, stat="density", color='#0984e3', alpha=0.6, edgecolor='white',
             label=r'标准定义 F (修正后)', ax=ax1)
# 3. 理论线
ax1.plot(x, y_theo, color='#d63031', lw=3, label=rf'理论 $F({n1-1}, {n2-1})$')

ax1.set_title(rf'A. 定义验证 ($\sigma_1=10 \neq \sigma_2=30$)', fontsize=14, fontweight='bold')
ax1.set_xlim(0, 4)
ax1.legend()
ax1.annotate('偏离!\n(因为σ不同)', xy=(0.2, 1.5), xytext=(0.5, 2.0),
             arrowprops=dict(facecolor='gray', arrowstyle='->'), color='gray', fontweight='bold')
ax1.annotate('完美回归\n(除以σ后)', xy=(1, 0.6), xytext=(1.5, 1.0),
             arrowprops=dict(facecolor='#0984e3', arrowstyle='->'), color='#0984e3', fontweight='bold')


# --- 右图：检验的应用 (σ1 = σ2) ---
ax2 = axes[1]
# 1. 画直接计算的比值 (橙色)
sns.histplot(true_f_B, bins=60, stat="density", color='#fdcb6e', alpha=0.6, edgecolor='white',
             label=r'直接比值 $S_1^2/S_2^2$', ax=ax2)
# 2. 理论线
ax2.plot(x, y_theo, color='#d63031', lw=3, ls='--', label=rf'理论 $F({n1-1}, {n2-1})$')

ax2.set_title(rf'B. 检验原理 ($\sigma_1 = \sigma_2$ 时 $\sigma$ 抵消)', fontsize=14, fontweight='bold')
ax2.set_xlim(0, 4)
ax2.legend()
# ax2.text(0.5, 0.5, "不需要知道 $\sigma$ 是多少\n比值天生服从 F 分布\n这就是 F检验 的原理", 
        #  transform=ax2.transAxes, ha='center', fontsize=12,
        #  bbox=dict(boxstyle='round', facecolor='white', alpha=0.8))
ax2.text(0.5, 0.5, r"不需要知道 $\sigma$ 是多少\n比值天生服从 F 分布\n这就是 F检验 的原理", 
         transform=ax2.transAxes, ha='center', fontsize=12,
         bbox=dict(boxstyle='round', facecolor='white', alpha=0.8))

# 全局标题
plt.suptitle('F 分布的双重境界：从数学定义到统计检验', fontsize=18, y=0.98)

# 保存
output_dir = "Mathematical_statistics/images"
if not os.path.exists(output_dir):
    os.makedirs(output_dir)
save_path = f"{output_dir}/f_distribution_sigma_cancellation.png"
plt.savefig(save_path, dpi=300)

print(f"图片已保存: {save_path}")
plt.show()
```
