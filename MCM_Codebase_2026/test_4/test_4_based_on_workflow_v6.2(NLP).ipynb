```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.optimize import minimize, differential_evolution, Bounds, NonlinearConstraint
import warnings
import time

# å¿½ç•¥ Scipy å’Œ Seaborn çš„ç¹çè­¦å‘Š
warnings.filterwarnings("ignore")

class NLP_Solver_Capsule:
    def __init__(self, name="NLP_Model", sense='min'):
        """
        [MCM/ICM NLP Solver V6.2]
        ç­–ç•¥: Gradient Descent (Local) + Multi-Start + Differential Evolution (Global)
        ä¿®å¤: æ”¯æŒå…¨å±€ç­–ç•¥çš„éçº¿æ€§çº¦æŸ, æ”¯æŒå‘é‡åŒ–å˜é‡, æ”¯æŒç»“æœå¯¼å‡º
        """
        self.name = name
        self.sense = 1.0 if sense == 'min' else -1.0
        
        # === å˜é‡å­˜å‚¨ ===
        self.var_names = []
        self.var_bounds = [] 
        self.x0 = []
        
        # === é€»è¾‘å­˜å‚¨ ===
        self.objective_func = None
        # V6.2 å…³é”®ä¿®å¤: åŒé‡å­˜å‚¨çº¦æŸ
        # minimize éœ€è¦å­—å…¸æ ¼å¼: {'type':'ineq', 'fun':...}
        # differential_evolution éœ€è¦å¯¹è±¡æ ¼å¼: NonlinearConstraint(...)
        self.constraints = []      
        self.de_constraints = []   
        
        # === ç»“æœå­˜å‚¨ ===
        self.result = None
        self.optimal_x = None
        self.optimal_fun = None
        self.is_solved = False
        self.metadata = {'multi_start_history': []}

    # ======================================================
    # Phase 0: V6.1 è‡ªæè¿°æ¡æ‰‹
    # ======================================================
    def generate_handshake(self, df_dict=None):
        """ç”Ÿæˆå‘ç»™ AI çš„æ¡æ‰‹æŒ‡ä»¤"""
        print(f"\nğŸ¤ === è¯·å¤åˆ¶ä»¥ä¸‹å†…å®¹å‘é€ç»™ AI (Handshake Prompt) ===\n")
        print(f"ã€ç³»ç»Ÿè®¾å®šã€‘\næˆ‘æ­£åœ¨ä½¿ç”¨ `NLP_Solver_Capsule` (V6.2) å¤„ç†éçº¿æ€§è§„åˆ’é—®é¢˜ã€‚")
        print(f"ç¯å¢ƒå·²åŠ è½½è¯¥ç±»ï¼Œè¯·ç›´æ¥ä½¿ç”¨å®ä¾‹ `solver`ã€‚")
        
        print("\nã€å¯ç”¨æ¥å£ (API)ã€‘")
        print("- solver.add_variables(names=['x','y'], bounds=[(0,10), (0,10)], init_guesses=[1,1])")
        print("- solver.add_variable_array(prefix='t', count=50, bounds=(0, 100)) # [V6.2] æ‰¹é‡æ·»åŠ ")
        print("- solver.set_objective(lambda x: x[0]**2 + x[1]) # xæ˜¯å…¨é‡æ‰å¹³æ•°ç»„")
        print("- solver.add_constraint(lambda x: x[0]+x[1]-10, type='ineq') # ineq: f(x)>=0")
        print("- solver.audit(is_convex=False)")
        print("- solver.solve(strategy='multi_start', n_starts=20)")
        print("- solver.analyze_sensitivity(param_idx=0, range_pct=0.2)")
        print("- solver.visualize_landscape(x_idx=0, y_idx=1) # [V6.2] äºŒç»´åœ°è²Œæ‰«æ")
        print("- solver.export_results() # å¯¼å‡ºç»“æœ")
        
        print("\nã€å…³é”®æŒ‡ä»¤ã€‘")
        print("1. å¿…é¡»ä½¿ç”¨ `import numpy as np` ä¸­çš„å‡½æ•° (å¦‚ np.sin)ã€‚")
        print("2. å˜é‡ x æ˜¯æ‰å¹³æ•°ç»„ï¼Œéœ€é€šè¿‡ x[0] è®¿é—®ã€‚")
        print("3. Scipy çº¦æŸç¬¦å·é™·é˜±ï¼šä¸ç­‰å¼ (ineq) å¿…é¡»å†™æˆ `è¡¨è¾¾å¼ >= 0`ã€‚")
        
        if df_dict:
            print("\nã€æ•°æ®æƒ…æŠ¥ã€‘")
            for name, df in df_dict.items():
                print(f"Dataset {name}: {list(df.columns)}")

    # ======================================================
    # Phase 1: å˜é‡å®šä¹‰ (å«å‘é‡åŒ–æ”¯æŒ)
    # ======================================================
    def add_variables(self, names, bounds, init_guesses=None):
        """æ·»åŠ ç¦»æ•£å˜é‡"""
        self.var_names.extend(names)
        self.var_bounds.extend(bounds)
        if init_guesses:
            self.x0.extend(init_guesses)
        else:
            # æ™ºèƒ½åˆå€¼: å–è¾¹ç•Œä¸­ç‚¹ï¼Œé˜²æ­¢ 0 é™·é˜±
            for b in bounds:
                mid = (b[0] + b[1]) / 2 if (b[0] is not None and b[1] is not None) else 0.1
                self.x0.append(mid)
        print(f"âœ… å·²æ·»åŠ  {len(names)} ä¸ªå˜é‡ã€‚æ€»ç»´åº¦: {len(self.var_names)}")

    def add_variable_array(self, prefix, count, bounds, init_val=None):
        """[V6.2] æ‰¹é‡æ·»åŠ å‘é‡å˜é‡ (å¦‚æ—¶é—´æ­¥ t_0...t_n)"""
        names = [f"{prefix}_{i}" for i in range(count)]
        bound_list = [bounds] * count
        init_list = [init_val] * count if init_val is not None else None
        self.add_variables(names, bound_list, init_list)

    # ======================================================
    # Phase 2: é€»è¾‘æ³¨å…¥ (åŒé‡çº¦æŸå­˜å‚¨)
    # ======================================================
    def set_objective(self, func):
        self.objective_func = lambda x: self.sense * func(x)

    def add_constraint(self, func, type='ineq', name="Constraint"):
        """
        æ·»åŠ çº¦æŸ (V6.2 ä¿®å¤ç‰ˆ)
        type: 'ineq' (f(x)>=0) or 'eq' (f(x)==0)
        """
        # 1. ä¸º minimize å­˜å‚¨å­—å…¸
        self.constraints.append({'type': type, 'fun': func})
        
        # 2. [V6.2 ä¿®å¤] ä¸ºå·®åˆ†è¿›åŒ–å­˜å‚¨ NonlinearConstraint
        # differential_evolution ä¸æ”¯æŒ 'ineq' å­—å…¸ï¼Œå¿…é¡»ç”¨å¯¹è±¡
        if type == 'ineq':
            # f(x) >= 0  =>  0 <= f(x) <= inf
            self.de_constraints.append(NonlinearConstraint(func, 0, np.inf))
        elif type == 'eq':
            # f(x) == 0  =>  0 <= f(x) <= 0
            self.de_constraints.append(NonlinearConstraint(func, 0, 0))
            
        print(f"ğŸ”— å·²æ·»åŠ çº¦æŸ: {name}")

    def audit(self, is_convex=False):
        """é€»è¾‘å®¡è®¡"""
        print("\nğŸ›¡ï¸ === NLP é€»è¾‘å®¡è®¡ ===")
        if not is_convex:
            print("â„¹ï¸ [éå‡¸é¢„è­¦] å°†å¯ç”¨ Multi-Start ç­–ç•¥ä»¥é¿å…å±€éƒ¨æœ€ä¼˜ã€‚")
        
        unbounded = sum(1 for b in self.var_bounds if b[0] is None or b[1] is None)
        if unbounded > 0:
            print(f"âš ï¸ [é«˜é£é™©] {unbounded} ä¸ªå˜é‡æ— è¾¹ç•Œã€‚å»ºè®®è®¾ç½®ç‰©ç†è¾¹ç•Œé˜²æ­¢å‘æ•£ã€‚")
        
        if not self.objective_func:
            print("âŒ [é”™è¯¯] æœªè®¾ç½®ç›®æ ‡å‡½æ•°ã€‚")
            return False
            
        return True

    # ======================================================
    # Phase 3: æ±‚è§£ (é›†æˆå…¨å±€ç­–ç•¥)
    # ======================================================
    def solve(self, strategy='multi_start', n_starts=20, max_iter=1000):
        print(f"\nğŸš€ å¼€å§‹æ±‚è§£ (ç­–ç•¥: {strategy})...")
        start_time = time.time()
        
        # æ„é€  Scipy Bounds å¯¹è±¡
        bounds_scipy = Bounds([b[0] if b[0] is not None else -np.inf for b in self.var_bounds],
                              [b[1] if b[1] is not None else np.inf for b in self.var_bounds])
        
        best_res = None
        best_fun = np.inf

        if strategy == 'multi_start':
            # ç­–ç•¥ 1: å¤šèµ·ç‚¹æ¢¯åº¦ä¸‹é™ (æŠ—å±€éƒ¨æœ€ä¼˜)
            success = 0
            for i in range(n_starts):
                # åœ¨è¾¹ç•Œå†…éšæœºæ’’ç‚¹
                rand_x0 = [np.random.uniform(b[0] if b[0] else -10, b[1] if b[1] else 10) 
                           for b in self.var_bounds]
                
                res = minimize(self.objective_func, rand_x0, method='SLSQP', 
                               bounds=bounds_scipy, constraints=self.constraints)
                
                if res.success:
                    success += 1
                    real_fun = res.fun * self.sense
                    self.metadata['multi_start_history'].append(real_fun)
                    if res.fun < best_fun:
                        best_fun = res.fun
                        best_res = res
            
            print(f"   -> å¤šèµ·ç‚¹æ”¶æ•›ç‡: {success}/{n_starts}")
            if len(self.metadata['multi_start_history']) > 1:
                std = np.std(self.metadata['multi_start_history'])
                if std > 1e-4: print(f"âš ï¸ [å¤šå³°è­¦å‘Š] ç»“æœæ ‡å‡†å·® {std:.4f}ï¼Œè§£ä¸ç¨³å®šã€‚")

        elif strategy == 'global':
            # ç­–ç•¥ 2: å·®åˆ†è¿›åŒ– (å¼ºåŠ›å…¨å±€æœç´¢)
            print("   -> å¯åŠ¨å·®åˆ†è¿›åŒ– (Global Search)...")
            # [V6.2 ä¿®å¤] ä¼ å…¥ de_constraintsï¼Œè®©å…¨å±€æœç´¢ä¹Ÿéµå®ˆçº¦æŸï¼
            best_res = differential_evolution(self.objective_func, bounds=self.var_bounds,
                                              constraints=self.de_constraints,
                                              maxiter=max_iter, seed=42)
        else:
            # ç­–ç•¥ 3: å•æ¬¡å±€éƒ¨æœç´¢
            best_res = minimize(self.objective_func, self.x0, method='SLSQP',
                                bounds=bounds_scipy, constraints=self.constraints)

        # ç»“æœå¤„ç†
        if best_res and (best_res.success or strategy=='global'):
            self.result = best_res
            self.optimal_x = best_res.x
            self.optimal_fun = best_res.fun * self.sense
            self.is_solved = True
            print(f"ğŸ’ æœ€ä¼˜å€¼: {self.optimal_fun:,.4f}")
        else:
            print(f"âŒ æ±‚è§£å¤±è´¥: {best_res.message if best_res else 'No solution'}")
        
        print(f"â±ï¸ è€—æ—¶: {time.time() - start_time:.2f}s")

    # ======================================================
    # Phase 4: å¯è§†åŒ–ä¸éªŒè¯ (V6.2 æ–°å¢)
    # ======================================================
    def visualize_landscape(self, idx1=0, idx2=1, points=30):
        """[V6.2] äºŒç»´åœ°è²Œç­‰é«˜çº¿æ‰«æ - ç”¨äºéªŒè¯æ˜¯å¦é™·å…¥å±€éƒ¨æœ€ä¼˜"""
        if not self.is_solved: return
        print(f"\nğŸ” ç”Ÿæˆåœ°è²Œæ‰«æ: {self.var_names[idx1]} vs {self.var_names[idx2]}")
        
        c1, c2 = self.optimal_x[idx1], self.optimal_x[idx2]
        # æ‰«æèŒƒå›´ï¼šæœ€ä¼˜è§£å‘¨å›´ 50%
        r1 = abs(c1)*0.5 if c1!=0 else 1.0
        r2 = abs(c2)*0.5 if c2!=0 else 1.0
        
        x = np.linspace(c1-r1, c1+r1, points)
        y = np.linspace(c2-r2, c2+r2, points)
        X, Y = np.meshgrid(x, y)
        Z = np.zeros_like(X)
        
        base_x = self.optimal_x.copy()
        for i in range(points):
            for j in range(points):
                base_x[idx1], base_x[idx2] = X[i,j], Y[i,j]
                Z[i,j] = self.objective_func(base_x) * self.sense
                
        plt.figure(figsize=(8, 6))
        # ä½¿ç”¨ contourf ç»˜åˆ¶ç­‰é«˜çº¿ï¼Œviridis é…è‰²ç¬¦åˆå­¦æœ¯è§„èŒƒ
        cp = plt.contourf(X, Y, Z, levels=20, cmap='viridis')
        plt.colorbar(cp, label='Objective Value')
        plt.plot(c1, c2, 'r*', markersize=15, label='Optimal Solution')
        plt.xlabel(self.var_names[idx1])
        plt.ylabel(self.var_names[idx2])
        plt.title('Optimization Landscape Scan')
        plt.legend()
        plt.show()

    def analyze_sensitivity(self, param_idx, range_pct=0.2):
        """å•å˜é‡çµæ•åº¦"""
        if not self.is_solved: return
        name = self.var_names[param_idx]
        center = self.optimal_x[param_idx]
        delta = abs(center * range_pct) if center != 0 else 1.0
        vals = np.linspace(center - delta, center + delta, 50)
        objs = []
        
        base_x = self.optimal_x.copy()
        for v in vals:
            base_x[param_idx] = v
            objs.append(self.objective_func(base_x) * self.sense)
            
        plt.figure(figsize=(6, 3))
        plt.plot(vals, objs)
        plt.axvline(center, color='r', linestyle='--')
        plt.title(f'Sensitivity: {name}')
        plt.xlabel('Variable Value')
        plt.ylabel('Objective')
        plt.grid(True, alpha=0.3)
        plt.show()

    def export_results(self):
        """å¯¼å‡ºç»“æœ"""
        if not self.is_solved: return
        df = pd.DataFrame({
            'Variable': self.var_names,
            'Optimal_Value': self.optimal_x,
            'Lower_Bound': [b[0] for b in self.var_bounds],
            'Upper_Bound': [b[1] for b in self.var_bounds]
        })
        filename = f"{self.name}_Solution.xlsx"
        df.to_excel(filename, index=False)
        print(f"\nğŸ’¾ ç»“æœå·²ä¿å­˜: {filename}")
        print("\nğŸ“ LaTeX Table Snippet:")
        print(df.head(5).to_latex(index=False, float_format="%.4f"))

    def get_methodology(self):
        return r"""
        \subsection{Non-Linear Optimization Strategy}
        We formulated the problem as a Non-Linear Programming (NLP) model. 
        To mitigate the risk of local optima inherent in non-convex problems, we employed a **Multi-Start Strategy** combined with the **SLSQP** gradient-based algorithm. 
        Global optimality was further cross-validated using **Differential Evolution** (a heuristic genetic algorithm) under strict constraints.
        Sensitivity analysis was performed by scanning the objective landscape around the optimal solution.
        """
```

## ç¬¬äºŒéƒ¨åˆ†ï¼šæŒ‡æŒ¥å®˜è‡ªç„¶è¯­è¨€å·¥ä½œæµ (Workflow Description)

> **æ ¸å¿ƒå®šä½**ï¼šè¿™æ˜¯ä¸€å¥—ç»™è‡ªå·±çœ‹çš„â€œä½œæˆ˜åœ°å›¾â€ï¼Œå®ƒè§£é‡Šäº†ä¸ºä»€ä¹ˆæˆ‘ä»¬è¦è¿™ä¹ˆåšã€‚åœ¨å‘ AI å‘é€æŒ‡ä»¤å‰ï¼Œè¯·å…ˆå¯¹ç…§æ­¤æµç¨‹å›¾ï¼Œç¡®ä¿æˆ˜ç•¥æ— è¯¯ã€‚

### Phase 0: å¯åŠ¨ä¸æ¡æ‰‹ (Boot & Handshake)
* **åŠ¨ä½œ**ï¼šåŠ è½½ç±»ï¼Œè¯»å–æ•°æ®ï¼ˆå¦‚æœ‰ï¼‰ã€‚
* **å…³é”®ç‚¹**ï¼šè¿è¡Œ `solver.generate_handshake()`ã€‚
    * **æˆ˜ç•¥æ„å›¾**ï¼šæŠŠå¤æ‚çš„ Python å‡½æ•°å®šä¹‰è§„åˆ™ã€ç±»çš„æ–¹æ³•ç­¾åç»Ÿç»Ÿâ€œæ‰”â€ç»™ AIã€‚è®©å®ƒå»å¤„ç†è¯­æ³•ç»†èŠ‚ï¼Œä½ åªè´Ÿè´£ç¡®è®¤å®ƒæ”¶åˆ°äº†â€œæ­¦å™¨â€ã€‚

### Phase 1: å˜é‡ä¸ç‰©ç†è¾¹ç•Œå®šä¹‰ (Variables & Boundaries)
* **åŠ¨ä½œ**ï¼šå‘Šè¯‰ AI æœ‰å“ªäº›å†³ç­–å˜é‡ã€‚
* **é€»è¾‘æ£€æŸ¥ (Critical)**ï¼š**å¿…é¡»è®¾ç½®è¾¹ç•Œï¼**
    * **åŸç†**ï¼šNLP (éçº¿æ€§è§„åˆ’) ä¸­æœªè®¾ç½®è¾¹ç•Œçš„å˜é‡æ˜¯ç¾éš¾ä¹‹æºï¼Œææ˜“å¯¼è‡´ç®—æ³•å‘æ•£æˆ–äº§ç”Ÿç‰©ç†ä¸Šæ— æ„ä¹‰çš„è§£ï¼ˆå¦‚è´Ÿçš„æ—¶é—´ã€æ— é™å¤§çš„è´¨é‡ï¼‰ã€‚
    * **æ“ä½œ**ï¼šå¦‚æœä½ ä¸çŸ¥é“ç¡®åˆ‡è¾¹ç•Œï¼Œä¹Ÿè¦ç»™ä¸€ä¸ª**å®½æ¾çš„ç‰©ç†æé™**ï¼ˆSafe Harborï¼‰ï¼Œä¾‹å¦‚ `[0, 1e9]` æˆ– `[-100, 100]`ã€‚

### Phase 2: æ•°å­¦é€»è¾‘æ³¨å…¥ (Constraint Injection)
* **åŠ¨ä½œ**ï¼šå°†ç›®æ ‡å‡½æ•°å’Œçº¦æŸæ¡ä»¶â€œç¿»è¯‘â€ç»™ AIã€‚
* **éš¾ç‚¹**ï¼šNLP çš„çº¦æŸé€šå¸¸æ˜¯éçº¿æ€§çš„ï¼ˆä¾‹å¦‚ $x^2 + y^2 \le 10$ï¼‰ã€‚
* **é˜²å‘æŒ‡å— (Scipy Trap)**ï¼š
    * Scipy çš„ `ineq` å®šä¹‰æ˜¯ $f(x) \ge 0$ã€‚
    * **å¦‚æœä½ çš„çº¦æŸæ˜¯ $A \le B$**ï¼šå¿…é¡»è½¬åŒ–ä¸º $B - A \ge 0$ã€‚
    * **å¦‚æœä½ çš„çº¦æŸæ˜¯ $A \ge B$**ï¼šå¿…é¡»è½¬åŒ–ä¸º $A - B \ge 0$ã€‚
    * **æŒ‡æŒ¥ç­–ç•¥**ï¼šè¿™ä¸ªè½¬æ¢å¾ˆç¹çï¼Œè®© AI å»å†™ lambda è¡¨è¾¾å¼ï¼Œä½†ä½ åœ¨ `solver.audit()` æˆ–ä»£ç å®¡æŸ¥æ—¶å¿…é¡»**äººå·¥æ ¸å¯¹**æ–¹å‘æ˜¯å¦åäº†ã€‚

### Phase 3: ç­–ç•¥é€‰æ‹©ä¸æ±‚è§£ (Strategy & Solve)
* **æ ¸å¿ƒåŠ¨ä½œ**ï¼šé€‰æ‹© `solve` çš„ç­–ç•¥ã€‚
* **å†³ç­–æ ‘ (Decision Tree)**ï¼š
    1.  **ç®€å•å‡¸é—®é¢˜**ï¼ˆå¦‚çº¿æ€§è§„åˆ’ã€ç®€å•çš„äºŒæ¬¡è§„åˆ’ï¼‰ï¼š
        * $\rightarrow$ ä½¿ç”¨ `'local'` (é€Ÿåº¦æœ€å¿«)ã€‚
    2.  **å¤æ‚éå‡¸é—®é¢˜**ï¼ˆç»å¤§å¤šæ•°ç¾èµ›é¢˜ç›®ï¼Œæœ‰å¤šä¸ªæ³¢å³°æ³¢è°·ï¼‰ï¼š
        * $\rightarrow$ **å¿…é¡»ä½¿ç”¨ `'multi_start'` (é»˜è®¤æ¨è)**ã€‚é€šè¿‡å¤šç‚¹ç©ºé™ï¼Œå¤§æ¦‚ç‡æ‰¾åˆ°å…¨å±€æœ€ä¼˜ã€‚
    3.  **é«˜éš¾åº¦å‘æ•£é—®é¢˜**ï¼ˆå¤šèµ·ç‚¹ç»“æœä¸ä¸€è‡´ï¼Œæ ‡å‡†å·®å¤§ï¼‰ï¼š
        * $\rightarrow$ åˆ‡æ¢ä¸º `'global'` (å·®åˆ†è¿›åŒ– Differential Evolution)ã€‚è¿™æ˜¯é‡æ­¦å™¨ï¼Œé€Ÿåº¦æ…¢ä½†ç¨³å¥ã€‚
* **é€»è¾‘æ£€æŸ¥**ï¼šå¦‚æœ AI ç”Ÿæˆçš„ä»£ç æ²¡æŒ‡å®š `strategy`ï¼Œè¯·**æ‰‹åŠ¨**å°†å…¶æ”¹ä¸º `'multi_start'`ï¼Œä¸è¦å†’é™©ä¿¡ä»»é»˜è®¤çš„å±€éƒ¨ä¼˜åŒ–ã€‚

### Phase 4: ç»“æœéªŒè¯ä¸çµæ•åº¦ (Verification)
* **åŠ¨ä½œ**ï¼šç»˜åˆ¶çµæ•åº¦æ›²çº¿ï¼Œè§‚å¯Ÿåœ°è²Œå›¾ã€‚
* **é€»è¾‘æ£€æŸ¥ (Boundary Check)**ï¼š
    * **è§‚å¯Ÿ**ï¼šæœ€ä¼˜è§£æ˜¯å¦è½åœ¨äº†è¾¹ç•Œä¸Šï¼Ÿ
    * **åˆ¤å®š**ï¼š
        * **åœ¨è¾¹ç•Œä¸Š**ï¼šè¯´æ˜è¯¥çº¦æŸæ˜¯â€œç´§è‡´çš„â€ (Binding)ï¼Œå®ƒæ˜¯ç³»ç»Ÿçš„ç“¶é¢ˆã€‚å¢åŠ è¯¥èµ„æºçš„è¾¹ç•Œèƒ½æ˜¾è‘—æ”¹å–„ç›®æ ‡å‡½æ•°ã€‚
        * **åœ¨å†…éƒ¨**ï¼šè¯´æ˜æ˜¯å†…éƒ¨æå€¼ç‚¹ï¼Œèµ„æºæœ‰å¯Œä½™ã€‚
* **ç‰©ç†æ£€æŸ¥**ï¼š
    * ç»“æœæ˜¯å¦ä¸ºè´Ÿæ•°ï¼ˆå¦‚æœä»£è¡¨è´¨é‡ã€æ—¶é—´ï¼‰ï¼Ÿ
    * æ•°é‡çº§æ˜¯å¦åˆç†ï¼ˆä¾‹å¦‚ç®—å‡ºå«æ˜Ÿé‡ 100 å¨æˆ– 0.1 å…‹ï¼‰ï¼Ÿ
