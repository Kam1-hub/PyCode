```python
import subprocess
import sys

# 库清单：已根据你提供的列表去重并整理
libraries = [
    "xlwings", 
    "xyzservices", "yapf", "yarl", "zict", "zipp", "zope.interface", "zstandard"
]

def update_all():
    print("开始检查并更新库（使用清华源）...")
    # 使用 -m pip 确保调用的是当前环境的 pip
    # --upgrade (-U) 逻辑：无则装，旧则升，新则过
    for lib in libraries:
        try:
            print(f"\n>> 正在处理: {lib}")
            subprocess.check_call([sys.executable, "-m", "pip", "install", "-U", lib])
        except subprocess.CalledProcessError:
            print(f"!! 安装/升级 {lib} 失败，可能是某些库需要特定的系统编译器（如 Twisted）。")
        except Exception as e:
            print(f"!! 发生错误: {e}")

if __name__ == "__main__":
    update_all()
    print("\n所有库处理完毕！")
```
