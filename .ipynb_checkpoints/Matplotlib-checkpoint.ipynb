```python
# 初始化数据
seasons = [1,2,3,4]   # 季度
stock1 = [4,8,2,6]    # 股票1每个季度对应的股价
stock2 = [10,12,5,3]  # 股票2每个季度对应的股价

#转化数据为Numpy数组（可跳过）
seasons = np.array(seasons)
stock1 = np.array(stock1)
stock2 = np.array(stock2)
```

```python
# 画出两个股价的折线图 通过.plot()实现
plt.plot(seasons, stock1)
plt.plot(seasons, stock2)
# plt.savefig("images/pic1_1.png")
plt.show()
```

```python
# 改进折线图
# 改进1: 指定折线图格式 在.plot()中通过第三个变量来指定线段和点的格式
plt.plot(seasons, stock1, "ro--", label="股票代码: abc")
plt.plot(seasons, stock2, "b^--", label="股票代码: def")
# plot第三个参数: fmt = '[marker][line][color]'
# 改进2: 添加标题，横纵坐标轴标注，图标
plt.title("折线图")  # 添加标题
plt.xlabel("季度")  # x轴
plt.ylabel("平均股价")  # y轴 
plt.legend()  # 添加可视化图例 右上角
# plt.savefig("pic1_2.png") # 保存图片
plt.show()
```

```python
# 改进2: 
# 调整横纵坐标轴刻度
plt.plot(seasons, stock1, "ro--", label="股票代码: abc")
plt.plot(seasons, stock2, "b^-", label="股票代码: def")
plt.title("折线图")
plt.xlabel("季度")
plt.ylabel("平均股价")
plt.legend()
# 设置x/y坐标刻度
plt.xticks([1, 2, 3, 4])
plt.yticks(np.arange(2, 13, 1))
# 添加网格与辅助线
plt.grid()
# plt.savefig("images/pic1_3.png")
plt.show()
```

```python
# 改进3: 添加一个绿色方块强调两个股票的股价在第4季度反转
import matplotlib.patches as patches

# 创建一个长方体对象
rect = patches.Rectangle((2.5, 1.5), 2, 5,  # 矩形从左下角起始坐标为(2.5,1.5) 宽度（向右）为2个单位长度 长度（向上）为5个单位长度 
                         linewidth=1, edgecolor='g', # 边框宽度为1 颜色为绿色（g）
                         facecolor='none')

plt.plot(seasons, stock1,
         "ro--", label="股票代码: abc")
plt.plot(seasons, stock2,
         "b^-", label="股票代码: def")
plt.title("折线图")
plt.xlabel("季度")
plt.ylabel("平均股价")

# 添加长方体
plt.gca().add_patch(rect)

plt.legend()
plt.xticks([1, 2, 3, 4])
plt.xticks
plt.yticks(np.arange(2, 13, 1))
plt.grid()
# plt.savefig("images/pic1_4.png")
plt.show()
```

```python
 # 改进4: 改变视图范围至上一张图片的绿色框中
plt.plot(seasons, stock1, "ro--", label="股票代码: abc")
plt.plot(seasons, stock2, "b^-", label="股票代码: def")
plt.title("折线图")
plt.xlabel("季度")
plt.ylabel("平均股价")
plt.legend()

plt.xticks([1, 2, 3, 4])
plt.yticks(np.arange(2, 13, 1))
plt.xlim(2.5, 4.5)  # 改变视图范围至上一张图片的绿色框中
plt.ylim(1.5, 6.5)
plt.grid()
# plt.savefig("images/pic1_5.png")
plt.show()
```

```python
# 例子2: plt.plot()函数中的快捷格式字符字符说明
# 参考: https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.plot.html
# 点符号 + 颜色 + 连线符号 三种元素是自由组合的
plt.figure(figsize=(9,4))
plt.plot(list(range(5)), [5]*5, 'r.-', label="(r.-)红色,小圆点,直线")
plt.plot(list(range(5)), [4]*5, 'go--', label="(go--)绿色,大圆点,虚线")
plt.plot(list(range(5)), [3]*5, 'k+-.', label="(k+-.) 黑色,十字点,点虚线")
plt.plot(list(range(5)), [2]*5, 'c*:', label="(c*:)青色,星号,密集点线")
plt.plot(list(range(5)), [1]*5, 'bs', label="(bs)蓝色,方块,(不连线)")
plt.annotate("更多格式请参考:\n https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.plot.html",
            xy=(0.0, 4.1), fontsize=14)
plt.legend(loc='lower right', prop={'size': 18}) # bbox_to_anchor=(1, -0.1)
# plt.savefig("images/pic2_1.png")
plt.show()
```

```python
# 面向对象(Object-oriented programming)精确语法 每一个Artist都是一个对象
# fig: 整个画面; axes: 2行2列坐标系（储存在numpy数组中） 可以画四个图
fig, axes = plt.subplots(2, 2, figsize=(6, 6))
axes[0, 0].bar(seasons, stock1)  # 指定左上角(0行0列)坐标系 柱状图
axes[0, 1].plot(seasons, stock2, "b^--")  # 指定右上角(0行1列)坐标系 折线图
ax = axes[1, 0]  # 对左下角(1行0列)坐标系代表的变量赋值
# 左下角画散点图(scatterplots)
ax.scatter(seasons, # 季度（横坐标）
           stock2-stock1,  # 股票差价（纵坐标）
           s=[10, 20, 50, 100], # size字符大小
           c=['r', 'b', 'c', 'y']) # color颜色

axes[0, 0].set_title("股票1")  # 左上角(0行0列)标题
axes[0, 1].set_title("股票2")  # 右上角(0行1列)标题
ax.set_ylabel("差价(股票1-股票2)")  # 左下角(1行0列)y轴标注
# plt.savefig("images/pic2_4.png")
plt.show()
#
#fig (整张大画布)
#      /      |      \
#axes[0,0]  axes[0,1]  axes[1,0] ... (具体的子图对象)
#    |          |          |
#  .bar()     .plot()    .scatter()  <-- 各自调用自己的绘图方法
#    |          |          |
#.set_title .set_title  .set_ylabel  <-- 各自设置自己的属性
```

```python
# 改进1: 去掉多余的坐标系、对整个画布对象添加整体大标题/注释
# 面向对象(Object-oriented programming)精确语法
fig, axes = plt.subplots(2, 2, figsize=(6, 6),
                        facecolor="grey",  # 画面背景改为灰色
                        sharex=True, sharey=True)  # 共享xy轴坐标系
axes[0, 0].bar(seasons, stock1)
axes[0, 1].plot(seasons, stock2, "b^--")
ax = axes[1, 0]
ax.plot(seasons, stock2-stock1, "--", color="black")
ax.scatter(seasons, stock2-stock1, 
           s=[10, 20, 50, 100],
           c=['r', 'b', 'c', 'y'])
ax.set_ylabel("差价(股票1-股票2)")
axes[0, 0].set_title("股票1")
axes[0, 1].set_title("股票2")

# 可以删除最后一个坐标系
axes[1, 1].remove()
axes[0, 0].plot(seasons, stock1, 'r+-')
# Figure（整个画布）对象添加整体大标题/注释
fig.suptitle("股票分析图")
fig.supylabel("股价")
fig.supxlabel("季度")
# plt.savefig("images/pic2_5.png", facecolor=fig.get_facecolor()) # 注: 保存图片底色要重新设置
```

```python
# 改进2: 利用右下角画一个3d图画 具体咋实现的？
# 面向对象OOP精确语法
fig, axes = plt.subplots(2, 2, figsize=(6, 6),
                        facecolor="grey",
                        sharex=True, sharey=True)
axes[0, 0].bar(seasons, stock1)
axes[0, 1].plot(seasons, stock2, "b^--")
ax = axes[1, 0]
ax.scatter(seasons, stock2-stock1,
           s=[10, 20, 50, 100],
           c=['r', 'b', 'c', 'y'])
ax.plot(seasons, stock2-stock1, "--", color="black")
ax.set_ylabel("差价(股票1-股票2)")
axes[0, 0].plot(seasons, stock1, 'r+-')
fig.suptitle("股票分析图")
fig.supylabel("股价")
fig.supxlabel("季度")
axes[0, 0].set_title("股票1")
axes[0, 1].set_title("股票2")

# 删除右下角坐标系
axes[1, 1].remove()
# 重新添加右下角坐标系（改变为三维坐标系）
ax = fig.add_subplot(2, 2, 4, 
# 我要在 2行2列 的布局中，针对 第 4 个位置（即右下角）进行操作
# 你可能会问：“之前不是已经有了 axes[1, 1] 代表右下角吗？为什么不能直接写 axes[1, 1].projection = '3d'？”
# 因为坐标系的“维数”是在创建那一刻注定的 所以我们必须调用 fig.add_subplot(...) 并指定 projection='3d'
# 这相当于把原来的 2D 纸拿走，放了一个 3D 盒子在那个位置
                     projection='3d', facecolor="grey") # 赋予三维属性、设置背景色
ax.stem(seasons, stock1, stock2-stock1)
ax.stem(seasons, stock1, stock2-stock1, 
        linefmt='k--', basefmt='k--', 
        bottom=10, orientation='y')
ax.plot_surface(np.array([1,1,4,4]).reshape(2,2),
                np.array([2.5,10,2.5,10]).reshape(2,2),
                np.array([0]*4).reshape(2,2), 
                alpha=0.2, color='red')
ax.plot_surface(np.array([1,1,4,4]).reshape(2,2),
                np.array([10]*4).reshape(2,2),
                np.array([-2.5,8,-2.5,8]).reshape(2,2),
                alpha=0.2, color='black')
ax.set_xlabel("季度(x)")
ax.set_ylabel("股票1(y)")
ax.set_zlabel("差价(z)")
plt.savefig("images/pic2_6.png", facecolor=fig.get_facecolor())
plt.show()
```

```python
# 改进3: 坐标系改变背景颜色，与3D图片中的投影构成对应关系
# 面向对象OOP精确语法
fig, axes = plt.subplots(2, 2, figsize=(6, 6),
                        facecolor="grey",
                        sharex=True, sharey=True,
                        frameon=True)
axes[0, 0].bar(seasons, stock1)
axes[0, 1].plot(seasons, stock2, "b^--")
ax = axes[1, 0]
ax.scatter(seasons, stock2-stock1, 
                   s=[10, 20, 50, 100],
                   c=['r', 'b', 'c', 'y'])
ax.plot(seasons, stock2-stock1, "--", color="black")
ax.set_ylabel("差价(股票1-股票2)")
axes[0, 0].plot(seasons, stock1, 'r+-')
fig.suptitle("股票分析图")
fig.supylabel("股价")
fig.supxlabel("季度")
axes[0, 0].set_title("股票1")
axes[0, 1].set_title("股票2")

axes[1, 1].remove()
ax = fig.add_subplot(2, 2, 4, 
                     projection='3d', facecolor="grey")
ax.stem(seasons, stock1, stock2-stock1)
ax.stem(seasons, stock1, stock2-stock1, 
        linefmt='k--', basefmt='k--', 
        bottom=10, orientation='y')
ax.plot_surface(np.array([1,1,4,4]).reshape(2,2),
                np.array([2.5,10,2.5,10]).reshape(2,2),
                np.array([0]*4).reshape(2,2), 
                alpha=0.2, color='red')
ax.plot_surface(np.array([1,1,4,4]).reshape(2,2),
                np.array([10]*4).reshape(2,2),
                np.array([-2.5,8,-2.5,8]).reshape(2,2),
                alpha=0.2, color='black')
ax.set_xlabel("季度(x)")
ax.set_ylabel("股票1(y)")
ax.set_zlabel("差价(z)")


# 改变坐标系的背景颜色（在画图后改变属性，OOP）
axes[1, 0].set_facecolor('grey')
axes[1, 0].patch.set_alpha(0.2)
axes[0, 0].set_facecolor('red')
axes[0, 0].patch.set_alpha(0.2)
plt.tight_layout() # '内容紧凑' 

# plt.savefig("images/pic2_7.png", facecolor=fig.get_facecolor())
plt.show()
```

```python
# 例1. 同时打印数学+中文
# 设置字体(主要字体/数学公式字符集)
rc = {
    "font.family": "Times New Roman",  # 全局默认字体设为新罗马 (英文衬线体)
    "mathtext.fontset": "stix",        # 数学公式使用 stix 字体 (类似 LaTeX 风格)
}
mpl.rcParams.update(rc)  # 更新配置

fig, ax = plt.subplots()

# 1. 设置 X 轴标签 (混合内容)
# r'' 表示原生字符串，防止转义
# $...$ 之间的内容会被 Matplotlib 识别为数学公式
# \mathrm{} 表示公式中的非斜体文本
ax.set_xlabel(r'密度$\mathrm{kg/m}^3$', 
              fontname='Microsoft YaHei', fontsize=20) # 指定标签使用微软雅黑

# 2. 在图上添加文本 (Text 1)
# 演示：微软雅黑中文 + Times New Roman 英文
ax.text(0.2, 0.8, r'微软雅黑 $\mathrm{Times New Roman}$',
        fontname='Microsoft YaHei', fontsize=20)

# 3. 在图上添加文本 (Text 2)
# 演示：宋体中文 + Times New Roman 英文
ax.text(0.2, 0.6, r'荣耀字体 $\mathrm{Times New Roman}$',
        fontname='HONOR Sans CN', fontsize=30)

plt.ylim(0.5, 0.9)
plt.show()
```

```python
# 常见工作流2：代码复用（函数）
# 三个“三角函数”
x = np.linspace(0, 10, 100)
y1 = np.cos(x)
y2 = np.sin(x)
y3 = np.tanh(x) # tanh函数
# 画时间序列曲线
# 输入/输出都包含坐标轴变量
def plot_time_series(x, y, fmt, lab="", ax=None):
    if ax is None:
        fig, ax = plt.subplot()
    ax.plot(x, y, fmt, label=lab)
    # x轴固有格式
    ax.set_xlabel("time")
    ax.xaxis.set_major_locator(plt.MultipleLocator(np.pi / 2))
    ax.xaxis.set_minor_locator(plt.MultipleLocator(np.pi / 4))
    labs = ax.xaxis.get_ticklabels()
    tick_positions = np.linspace(0, 2*np.pi, len(labs), endpoint=False)
    ax.set_xticks(tick_positions)
    ax.set_xticklabels([r"{:.2f}$\pi$".format(i/2) for i, l in enumerate(labs)])
    # 改之前的：ax.xaxis.set_ticklabels([r"{:.2f}$\pi$".format(i/2) for i, l in enumerate(labs)])
    # set_ticklabels() 应该只与固定的刻度数量一起使用，即在 set_ticks() 之后使用
    return ax

# 两个坐标轴
fig, axes = plt.subplots(2, 1, figsize=(6, 3),
                         sharex=True, facecolor="white")
# 在第一个坐标周画两条线
plot_time_series(x, y1, 'b-', r'$y=sin(x)$', ax=axes[0])
plot_time_series(x, y2, 'r:', r'$y=cos(x)$', ax=axes[0])
# 在第二个坐标周画一条线
plot_time_series(x, y3, 'g--', ax=axes[1])

# plt.savefig("images/pic3_2.png", facecolor=fig.get_facecolor())
plt.show()
```

```python
# 拓展：很多格式可以后期再函数外处理
def plot_time_series(x, y, fmt, lab="", ax=None):
    if ax is None:
        fig, ax = plt.subplot()
    ax.plot(x, y, fmt, label=lab)
    # x轴固有格式
    ax.set_xlabel("time")
    ax.xaxis.set_major_locator(plt.MultipleLocator(np.pi / 2))
    ax.xaxis.set_minor_locator(plt.MultipleLocator(np.pi / 4))
    labs = ax.xaxis.get_ticklabels()
    ax.xaxis.set_ticklabels([r"{:.2f}$\pi$".format(i/2) for i, l in enumerate(labs)])
    return ax

fig, axes = plt.subplots(2, 1, figsize=(6, 3), 
                         sharex=True, facecolor="white")
plot_time_series(x, y1, 'b-', r'$y=sin(x)$', axes[0])
plot_time_series(x, y2, 'r:', r'$y=cos(x)$', axes[0])
plot_time_series(x, y3, 'g--', ax=axes[1])

# 后期调整格式/添加内容/美化...
axes[0].set_xlabel("") # 删去横坐标的注释
axes[0].legend(loc="upper right") # 调整图示位置
axes[1].legend(loc="upper right")
axes[0].grid() # 加一个参考线
axes[1].grid()

# plt.savefig("images/pic3_3.png", facecolor=fig.get_facecolor())
plt.show()
```

```python
# 后期调整
fig, axes = plt.subplots(2, 1, figsize=(6, 3),
                         sharex=True, facecolor="white")
plot_time_series(x, y1, 'b-', r'$y=sin(x)$', axes[0])
plot_time_series(x, y2, 'r:', r'$y=cos(x)$', axes[0])
plot_time_series(x, y3, 'g--', ax=axes[1])
# 调整格式/添加内容/美化...
axes[0].set_xlabel("")
# 补充图示
axes[1].lines[0].set_label(r"$y=tanh(x)$") # 漏脸图示 再补上
axes[0].legend(loc="upper right")
axes[1].legend(loc="upper right")
axes[0].grid()
axes[1].grid()

plt.savefig("images/pic3_4.png", facecolor=fig.get_facecolor())
plt.show()
```
