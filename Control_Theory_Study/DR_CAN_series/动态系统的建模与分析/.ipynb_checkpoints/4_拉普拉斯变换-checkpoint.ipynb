Âç∑ÁßØÂÆöÁêÜ (13:17 - 13:59)‚ÄúËøôÈáåÈù¢ËøòÊúâ‰∏Ä‰∏™ÈùûÂ∏∏ÈáçË¶ÅÁöÑÊÄßË¥®ÔºåÂ∞±ÊòØÂç∑ÁßØÁöÑÊãâÊôÆÊãâÊñØÂèòÊç¢„ÄÇ$$\mathcal{L}\{f(t) * g(t)\} = F(s) \cdot G(s)$$ËøôÊòØ‰∏Ä‰∏™ÈùûÂ∏∏ÈáçË¶ÅÁöÑÊÄßË¥®ÔºåÂõ†‰∏∫ÂÆÉÊääÂç∑ÁßØÁöÑËøêÁÆóÂèòÊàê‰∫Ü‰πòÊ≥ïÁöÑËøêÁÆóÔºåÂ§ßÂ§ßÂú∞ÁÆÄÂåñ‰∫ÜËøêÁÆóÂíåÂàÜÊûêÁöÑÂ§çÊùÇÁ®ãÂ∫¶„ÄÇ‚Äù

## 13:17 - 13:59ÔºöÂÖ≥ÈîÆÊÄßË¥®ÔºöÂç∑ÁßØÂÆöÁêÜ
* **ÂÆöÁêÜÊèèËø∞**ÔºöËØÅÊòéÊó∂ÂüüÂç∑ÁßØÁ≠â‰∫éÈ¢ëÂüü‰πòÊ≥ï„ÄÇ
* **ÂÖ¨Âºè**Ôºö
  $$\mathcal{L}\{f(t) * g(t)\} = F(s)G(s)$$

‚ÄúËøôÈáåÈù¢ËøòÊúâ‰∏Ä‰∏™ÈùûÂ∏∏ÈáçË¶ÅÁöÑÊÄßË¥®ÔºåÂ∞±ÊòØÂç∑ÁßØÁöÑÊãâÊôÆÊãâÊñØÂèòÊç¢„ÄÇ$$\mathcal{L}\{f(t) * g(t)\} = F(s) \cdot G(s)$$ËøôÊòØ‰∏Ä‰∏™ÈùûÂ∏∏ÈáçË¶ÅÁöÑÊÄßË¥®ÔºåÂõ†‰∏∫ÂÆÉÊääÂç∑ÁßØÁöÑËøêÁÆóÂèòÊàê‰∫Ü‰πòÊ≥ïÁöÑËøêÁÆóÔºåÂ§ßÂ§ßÂú∞ÁÆÄÂåñ‰∫ÜËøêÁÆóÂíåÂàÜÊûêÁöÑÂ§çÊùÇÁ®ãÂ∫¶„ÄÇ‚Äù

---

## 14:00 - 16:11ÔºöÊÄªÁªìÔºö‰º†ÈÄíÂáΩÊï∞ÔºàTransfer FunctionÔºâÁöÑËØûÁîü
* **ÂõûÂΩíÊ°à‰æã**ÔºöÂà©Áî®‰∏äËø∞ÊÄßË¥®ÔºåÂ∞ÜÁîµË∑ØÂæÆÂàÜÊñπÁ®ãÁõ¥Êé•ËΩ¨Âåñ‰∏∫ $S$ Âüü‰ª£Êï∞Âºè„ÄÇ
* **Ê†∏ÂøÉÂÆö‰πâ**Ôºö**ËæìÂá∫‰∏éËæìÂÖ•ÁöÑÊãâÊôÆÊãâÊñØÂèòÊç¢‰πãÊØî**Âç≥‰∏∫‰º†ÈÄíÂáΩÊï∞ $H(s)$„ÄÇ

‚ÄúÊàë‰ª¨Áé∞Âú®ÂõûËøáÂ§¥Êù•ÁúãÊúÄÂºÄÂßãÁöÑËøô‰∏™‰æãÂ≠ê„ÄÇÂ¶ÇÊûúÊàë‰ª¨ÊääËøô‰∏™ÂæÆÂàÜÊñπÁ®ã‰∏§ËæπËøõË°åÊãâÊôÆÊãâÊñØÂèòÊç¢ÁöÑËØùÔºå‰Ω†Â∞±‰ºöÂèëÁé∞Ôºö$s E(s) = L s^2 I(s) + R s I(s) + \frac{1}{C} I(s)$ÊääÂÆÉÊï¥ÁêÜ‰∏Ä‰∏ãÔºåÊàë‰ª¨Â∞±ÂèØ‰ª•ÂæóÂà∞ $I(s)$Ôºå‰πüÂ∞±ÊòØÁîµÊµÅÂä®ÊÄÅÊñπÁ®ãÁöÑÊãâÊôÆÊãâÊñØÂèòÊç¢Á≠â‰∫éÔºö$$I(s) = \frac{s}{Ls^2 + Rs + \frac{1}{C}} E(s)$$Êàë‰ª¨ÁúãÂà∞Êää‰∏Ä‰∏™ÂæÆÂàÜÊñπÁ®ãÂú®Ëøô‰∏™Âú∞ÊñπÂ∞±ÂèòÊàê‰∫Ü‰∏Ä‰∏™‰ª£Êï∞ÊñπÁ®ã‰∫Ü„ÄÇÂÜôÊàêÊ°ÜÂõæÁöÑËØùÔºåËæìÂÖ•ÊòØ $E(s)$Ôºå‰πò‰ª•‰∏≠Èó¥ÁöÑËøô‰∏ÄÈÉ®ÂàÜÔºåÂ∞±Á≠â‰∫éËæìÂá∫ $I(s)$„ÄÇ‰∏≠Èó¥ÁöÑËøô‰∏™Ê°ÜÊ°ÜÔºåÊàë‰ª¨Â∞±ÊääÂÆÉÂè´ÂÅö‰º†ÈÄíÂáΩÊï∞ÔºàTransfer FunctionÔºâ„ÄÇËøô‰∏™ÈùûÂ∏∏ÈùûÂ∏∏ÁöÑÈáçË¶Å„ÄÇ‚Äù

# Ê†∏ÂøÉÊ¶ÇÂøµÔºöÊãâÊôÆÊãâÊñØÂèòÊç¢ (Laplace Transform)

ÊãâÊôÆÊãâÊñØÂèòÊç¢ÊòØÊéßÂà∂ÁêÜËÆ∫ÁöÑÂü∫Á°ÄÔºåÂÖ∂Ê†∏ÂøÉ‰ª∑ÂÄºÂú®‰∫éÂ∞ÜÂ§çÊùÇÁöÑ**Êó∂ÂüüÂæÆÂàÜÊñπÁ®ã**ËΩ¨Âåñ‰∏∫ÁÆÄÂçïÁöÑ**Â§çÈ¢ëÂüü‰ª£Êï∞ÊñπÁ®ã**„ÄÇ

---

## 1. Êï∞Â≠¶ÂÆö‰πâ
ÂØπ‰∫é‰∏Ä‰∏™Êó∂Èó¥ÂáΩÊï∞ $f(t)$ÔºåÂÖ∂ÊãâÊôÆÊãâÊñØÂèòÊç¢ $F(s)$ ÂÆö‰πâ‰∏∫Ôºö

$$F(s) = \mathcal{L}\{f(t)\} = \int_{0}^{\infty} f(t) e^{-st} dt$$

* **ÂâçÊèêÂÅáËÆæ**Ôºö‰ªÖÁ†îÁ©∂ $t \ge 0$ ÁöÑÊÉÖÂÜµÔºå‰∏çËÄÉËôë $t < 0$ ÁöÑÁ≥ªÁªüÂéÜÂè≤„ÄÇ
* **Â§çÈ¢ëÁéáÂèÇÊï∞**Ôºö$s = \sigma + j\omega$ ÊòØ‰∏Ä‰∏™Â§çÊï∞„ÄÇ

---

## 2. Áâ©ÁêÜ‰∏éÁõ¥ËßâÁêÜËß£
* **Âá†‰ΩïËßÜËßí**ÔºöÊó∂Âüü $f(t)$ ÊòØ‰∫åÁª¥ÂõæÂÉèÔºåËÄå $s$ ÂüüÊòØ‰∏Ä‰∏™Âπ≥Èù¢ÔºåÂèòÊç¢ÂêéÁöÑ $F(s)$ ÊòØ $s$ Âπ≥Èù¢‰∏äÁöÑ‰∏âÁª¥Êõ≤Èù¢„ÄÇ
* **‰∏éÂÇÖÈáåÂè∂ÂèòÊç¢ÁöÑÂÖ≥Á≥ª**ÔºöÂΩìÂÆûÈÉ® $\sigma = 0$ Êó∂ÔºåÊãâÊôÆÊãâÊñØÂèòÊç¢ÈÄÄÂåñ‰∏∫**ÂÇÖÈáåÂè∂ÂèòÊç¢ÔºàFTÔºâ**„ÄÇ
* **Êú¨Ë¥®Áõ¥Ëßâ**ÔºöÂèØ‰ª•ÁêÜËß£‰∏∫ÊãâÊôÆÊãâÊñØÂèòÊç¢ÊòØÂú®È¢ëÂüüÂàÜÊûêÁöÑÂü∫Á°Ä‰∏äÂ¢ûÂä†‰∫ÜÊî∂ÊïõÂõ†Â≠ê $e^{-\sigma t}$Ôºå‰ΩøÂÖ∂ËÉΩÂ§üÂ§ÑÁêÜÊõ¥Â§öÁ±ªÂûãÁöÑÈùûÂπ≥Á®≥‰ø°Âè∑„ÄÇ

---

## üìù ÂÖ∏ÂûãÂáΩÊï∞ÁöÑÊãâÂºèÊé®ÂØº

### 1. ÊåáÊï∞ÂáΩÊï∞Ôºö$f(t) = e^{-at}$
**Êé®ÂØºÊ≠•È™§**Ôºö
1.  **‰ª£ÂÖ•ÂÆö‰πâ**Ôºö$\int_{0}^{\infty} e^{-at} e^{-st} dt = \int_{0}^{\infty} e^{-(s+a)t} dt$
2.  **Ê±ÇÂéüÂáΩÊï∞**Ôºö$\left[ -\frac{1}{s+a} e^{-(s+a)t} \right]_{0}^{\infty}$
3.  **ÂèñÊûÅÈôê**ÔºöÂΩì $t \to \infty$ Êó∂Ôºå$e^{-(s+a)t} \to 0$ÔºàÂÅáËÆæ $s+a$ ÂÆûÈÉ®Â§ß‰∫é 0Ôºâ„ÄÇ
* **ÁªìËÆ∫ÂàÜÊûê**Ôºö$\mathcal{L}\{e^{-at}\} = \frac{1}{s+a}$

### 2. Ê≠£Âº¶ÂáΩÊï∞Ôºö$f(t) = \sin(at)$
* **ÂâçÊèêÊé®ÂØºÔºàÊ¨ßÊãâÂÖ¨ÂºèÔºâ**Ôºö$\sin(at) = \frac{e^{jat} - e^{-jat}}{2j}$
* **Êé®ÂØºÈÄªËæë**ÔºöÂà©Áî®ÊãâÂºèÂèòÊç¢ÁöÑ**Á∫øÊÄßÊÄßË¥®**ÔºåÂ∞ÜÂÖ∂ÊãÜÂàÜ‰∏∫‰∏§‰∏™ÊåáÊï∞ÂáΩÊï∞ÁöÑÂèòÊç¢‰πãÂ∑Æ„ÄÇ
* **ÁªìËÆ∫ÂàÜÊûê**Ôºö$\mathcal{L}\{\sin(at)\} = \frac{a}{s^2 + a^2}$

---

## ‚öôÔ∏è Ê†∏ÂøÉËøêÁÆóÊÄßË¥®Ë°®

| ÊÄßË¥®ÂêçÁß∞ | Êó∂ÂüüÂΩ¢Âºè $f(t)$ | $s$ ÂüüÂΩ¢Âºè $F(s)$ | Áâ©ÁêÜ/Â∑•Á®ãÁõ¥Ëßâ |
| :--- | :--- | :--- | :--- |
| **Á∫øÊÄß** | $a f(t) + b g(t)$ | $a F(s) + b G(s)$ | Á≥ªÁªüÁ¨¶ÂêàÂè†Âä†ÂéüÁêÜÔºåÂèØÊãÜÂàÜÂàÜÊûê„ÄÇ |
| **‰∏ÄÈò∂ÂæÆÂàÜ** | $f'(t)$ | $s F(s) - f(0)$ | Ê±ÇÂØºÂèò‰πòÊ≥ï„ÄÇ$s$ Âõ†Â≠ê‰ª£Ë°®ÂæÆÂàÜÁÆóÂ≠ê„ÄÇ |
| **‰∫åÈò∂ÂæÆÂàÜ** | $f''(t)$ | $s^2 F(s) - s f(0) - f'(0)$ | ÂàùÂßãÊù°‰ª∂‰∏∫ 0 Êó∂Ôºå‰∫åÊ¨°Ê±ÇÂØºÂØπÂ∫î $s^2$„ÄÇ |
| **ÁßØÂàÜ** | $\int f(\tau) d\tau$ | $\frac{1}{s} F(s)$ | ÁßØÂàÜÂèòÈô§Ê≥ï„ÄÇ$\frac{1}{s}$ ‰ª£Ë°®Á≥ªÁªüÁöÑÁ¥ØÁßØÊïàÂ∫î„ÄÇ |
| **Âç∑ÁßØ** | $f(t) * g(t)$ | $F(s) \cdot G(s)$ | Êó∂ÂüüÊûÅÂÖ∂Â§çÊùÇÁöÑÂç∑ÁßØÂú®È¢ëÂüüÁÆÄÂåñ‰∏∫‰πòÊ≥ï„ÄÇ |

---

## üèóÔ∏è Ê°à‰æãÂàÜÊûêÔºö‰ªéÂæÆÂàÜÊñπÁ®ãÂà∞‰º†ÈÄíÂáΩÊï∞
ËßÜÈ¢ë‰ª• **RLC ÁîµË∑ØÁ≥ªÁªü** ‰∏∫‰æãÂ±ïÁ§∫‰∫ÜÂÆåÊï¥Âª∫Ê®°ËøáÁ®ã„ÄÇ

### 1. Áâ©ÁêÜÂª∫Ê®° (Êó∂Âüü)
Ê†πÊçÆÂü∫Â∞îÈúçÂ§´ÂÆöÂæãÔºåÁîµË∑ØÁ≥ªÁªüÁöÑÂä®ÊÄÅÊñπÁ®ã‰∏∫Ôºö
$$e' = L i'' + R i' + \frac{1}{C} i$$
ÂÖ∂‰∏≠ËæìÂÖ•‰∏∫ÁîµÂéã $e'(t)$ÔºåËæìÂá∫‰∏∫ÁîµÊµÅ $i(t)$„ÄÇÂú®Êó∂Âüü‰∏≠ÔºåËæìÂá∫ÊòØËæìÂÖ•‰∏éÁ≥ªÁªüÁâπÂæÅÂáΩÊï∞ $g(t)$ ÁöÑÂç∑ÁßØ„ÄÇ

### 2. Êï∞Â≠óÂåñËΩ¨Âåñ ($s$ Âüü)
ÂÅáËÆæÂàùÂßãÊù°‰ª∂‰∏∫ 0ÔºåÂØπÁ≠âÂºè‰∏§ËæπËøõË°åÊãâÊôÆÊãâÊñØÂèòÊç¢Ôºö
$$s E(s) = (L s^2 + R s + \frac{1}{C}) I(s)$$

### 3. ‰º†ÈÄíÂáΩÊï∞ (Transfer Function) ÂÆö‰πâ
‰º†ÈÄíÂáΩÊï∞ $G(s)$ ÂÆö‰πâ‰∏∫ÔºöÂú®Èõ∂ÂàùÂßãÊù°‰ª∂‰∏ãÔºåÁ≥ªÁªü**ËæìÂá∫**ÁöÑÊãâÂºèÂèòÊç¢‰∏é**ËæìÂÖ•**ÁöÑÊãâÂºèÂèòÊç¢‰πãÊØî„ÄÇ
$$G(s) = \frac{I(s)}{E(s)} = \frac{s}{Ls^2 + Rs + \frac{1}{C}}$$

> **ËÄÅÂ∏àÂº∫Ë∞ÉÁöÑÁõ¥ËßâÁêÜËß£**Ôºö
> ‰º†ÈÄíÂáΩÊï∞ $G(s)$ ÂÆåÂÖ®ÂàªÁîª‰∫ÜÁ≥ªÁªüÁöÑÂõ∫ÊúâÁâπÊÄßÔºàÁî± $L, R, C$ ÂÜ≥ÂÆöÔºâÔºå‰∏éÂÖ∑‰ΩìÁöÑËæìÂÖ•‰ø°Âè∑Êó†ÂÖ≥„ÄÇÂú® $s$ ÂüüÔºåÊàë‰ª¨Âè™ÈúÄË¶ÅÁî®ËæìÂÖ•‰πò‰ª• $G(s)$ Â∞±ËÉΩÂæóÂà∞ËæìÂá∫ÔºåÈÅøÂÖç‰∫ÜÂèØÊÄïÁöÑÂç∑ÁßØËÆ°ÁÆó„ÄÇ

---

## ‚ö†Ô∏è ÂùëÁÇπÈ¢ÑÊµã‰∏éÂêéÁª≠ÂÖ≥ËÅî
* **ÈöúÁ¢çÁÇπ**ÔºöÂ¶ÇÊûú‰∏çÁêÜËß£ $s$ Âõ†Â≠ê‰∏éÂæÆÂàÜÁöÑÂÖ≥Á≥ªÔºåÂêéÁª≠Â≠¶‰π†‚ÄúÊ†πËΩ®Ëøπ‚ÄùÊó∂Â∞ÜÊó†Ê≥ïÁêÜËß£‰∏∫‰ªÄ‰πàÊîπÂèòÂ¢ûÁõä‰ºöÂΩ±ÂìçÁ≥ªÁªüÁöÑÁ®≥ÂÆöÊÄß„ÄÇ
* **ÂÖ≥ËÅîÁÇπ**Ôºö
    1.  **Á®≥ÂÆöÊÄß**Ôºö‰º†ÈÄíÂáΩÊï∞ÂàÜÊØçÁöÑÊ†πÔºàÊûÅÁÇπÔºâÂÜ≥ÂÆö‰∫ÜÁ≥ªÁªüÊòØÂê¶Êî∂Êïõ„ÄÇ
    2.  **È¢ëÂüüÂàÜÊûê**ÔºöÂ∞Ü $s$ ÊõøÊç¢‰∏∫ $j\omega$ÔºåÂç≥ÂèØÁõ¥Êé•ËøõÂÖ•**Ê≥¢Âæ∑Âõæ (Bode Plot)** ÂàÜÊûêÈò∂ÊÆµ„ÄÇ

# ÁîµË∑ØÂä®ÊÄÅ‰ªøÁúü‰ª£Á†Å

```python
import numpy as np
import matplotlib.pyplot as plt
import control
from IPython.display import display, Latex
import warnings

warnings.filterwarnings("ignore", category=FutureWarning, module="control")

# ==========================================
# 1. ÊâãÂä®‰øÆÊîπÂèÇÊï∞Âå∫Âüü (Êï∞Â≠óÂÆûÈ™åÂÆ§ÊéßÂà∂Âè∞)
# ==========================================
R = 1.0   # ÁîµÈòª (Œ©) - Â¢ûÂ§ß R ‰ºöÊòæËëóÂ¢ûÂä†Á≥ªÁªüÈòªÂ∞º
L = 0.5   # ÁîµÊÑü (H)
C = 0.1   # ÁîµÂÆπ (F)
# ==========================================

# ËÆæÁΩÆÁªòÂõæÈ£éÊ†º‰∏é‰∏≠ÊñáÊîØÊåÅ
plt.rcParams['font.sans-serif'] = ['SimHei'] # ËÆæÁΩÆ‰∏≠ÊñáÂ≠ó‰Ωì
plt.rcParams['axes.unicode_minus'] = False  # Ê≠£Â∏∏ÊòæÁ§∫Ë¥üÂè∑

# 2. Âª∫Á´ã‰º†ÈÄíÂáΩÊï∞ (Ê†πÊçÆ DR_CAN ËßÜÈ¢ëÊé®ÂØº)
# Êó∂ÂüüÂæÆÂàÜÊñπÁ®ã: e' = Li'' + Ri' + (1/C)i 
# ÂèòÊç¢‰∏∫ s Âüü: sE(s) = (Ls^2 + Rs + 1/C)I(s)
# ‰º†ÈÄíÂáΩÊï∞: G(s) = I(s)/E(s) = Cs / (LCs^2 + RCs + 1)
num = [C, 0]
den = [L * C, R * C, 1]
sys = control.TransferFunction(num, den)

# 3. ËÆ°ÁÆó‰∏éÁªòÂà∂Èò∂Ë∑ÉÂìçÂ∫î (Êó∂ÂüüÂàÜÊûê)
t, y = control.step_response(sys)

plt.figure(figsize=(12, 5))

# Â≠êÂõæ 1: Èò∂Ë∑ÉÂìçÂ∫îÊ≥¢ÂΩ¢
plt.subplot(1, 2, 1)
plt.plot(t, y, 'b', lw=2, label=f'R={R}Œ©, L={L}H, C={C}F')
plt.axhline(0, color='black', lw=1, ls='--')
plt.ylim(-0.3, 0.4) 
plt.xlim(0, 7)
plt.title('RLC ÁîµË∑ØÈò∂Ë∑ÉÂìçÂ∫î (Êó∂ÂüüÊ≥¢ÂΩ¢)', fontsize=12)
plt.xlabel('Êó∂Èó¥ (s)')
plt.ylabel('ÁîµÊµÅ (A)')
plt.grid(True, alpha=0.3)
plt.legend()

# 4. Ëé∑ÂèñÂπ∂‰ºòÂåñÊûÅÁÇπÊòæÁ§∫
try:
    poles = sys.poles()
except AttributeError:
    poles = control.poles(sys)

# Ê†ºÂºèÂåñÂáΩÊï∞ÔºöÂéªÈô§ÂÜó‰ΩôËôöÈÉ®Âπ∂‰øùÁïô 4 ‰ΩçÂ∞èÊï∞
def format_pole(p):
    real, imag = round(p.real, 4), round(p.imag, 4)
    if abs(imag) < 1e-5: return f"{real}"
    return f"{real} {'+' if imag > 0 else '-'} {abs(imag)}j"

clean_poles = [format_pole(p) for p in poles]

# 5. ÁªòÂà∂ s Âπ≥Èù¢ÊûÅÁÇπÂàÜÂ∏ÉÂõæ (È¢ëÂüüÂàÜÊûê)
plt.subplot(1, 2, 2)
# ‰ΩøÁî® control Â∫ìËá™Â∏¶ÁöÑÊûÅÁÇπÂàÜÂ∏ÉÂõæÔºåÂπ∂Â¢ûÂº∫ÊòæÁ§∫ÊïàÊûú
control.pzmap(sys, plot=True, title='ÊûÅÁÇπÂàÜÂ∏ÉÂõæ (s-Plane)')
plt.grid(True, which='both', linestyle='--', alpha=0.5)
plt.axhline(0, color='black', lw=1)
plt.axvline(0, color='black', lw=1)

plt.tight_layout()
plt.show()

# 6. ‰∏ì‰∏öÂåñÁªìÊûúËæìÂá∫ (LaTeX Ê∏≤Êüì)
print("-" * 50)
print(f">>> Êï∞Â≠óÂåñÂàÜÊûêÊä•Âëä <<<")
print(f"Ê†ºÂºèÂåñÊûÅÁÇπÊñáÊú¨: [{', '.join(clean_poles)}]")

# ‰ΩøÁî® LaTeX Êï∞Â≠¶ÂÖ¨ÂºèÂ±ïÁ§∫ÊûÅÁÇπÔºåÊñπ‰æøÁõ¥Êé•Â≠òÂÖ•Á¨îËÆ∞
latex_expr = r"s_{1,2} = " + ", ".join([str(round(p.real, 4)) + (f" \pm {round(abs(p.imag), 4)}j" if abs(p.imag) > 1e-5 else "") for p in [poles[0]]])
if abs(poles[0].imag) < 1e-5: # Â§ÑÁêÜ‰∏§‰∏™‰∏çÁõ∏Á≠âÁöÑÂÆûÊ†πÊÉÖÂÜµ
    latex_expr = r"s_{1,2} = \{" + ", ".join([str(round(p.real, 4)) for p in poles]) + r"\}"

display(Latex(f"ÊûÅÁÇπÊï∞Â≠¶Ë°®ËææÂºèÔºö"))
display(Latex(f"$$ {latex_expr} $$"))
```

# üéØ ÊéßÂà∂ÁêÜËÆ∫ÔºöÊûÅÁÇπ‰ΩçÁΩÆ‰∏éÁ≥ªÁªüË°å‰∏∫ÁöÑÊ∑±Â∫¶Ëß£Êûê

Âú®ÊéßÂà∂ÁêÜËÆ∫‰∏≠ÔºåÊûÅÁÇπÂú®Â§çÂπ≥Èù¢Ôºà$s$ Âπ≥Èù¢Ôºâ‰∏≠ÁöÑ‰ΩçÁΩÆÊòØÂÜ≥ÂÆöÁ≥ªÁªüÂä®ÊÄÅÁâπÊÄßÁöÑ‚ÄúÁÅµÈ≠Ç‚Äù„ÄÇË¶ÅÁêÜËß£ÊûÅÁÇπÂ¶Ç‰ΩïÊìçÊéßÁ≥ªÁªüÔºåÊúÄÁõ¥ËßÇÁöÑÊñπÊ≥ïÊòØÂ∞ÜÂ§çÈ¢ëÁéá $s$ ‰ª£ÂÖ•Êó∂Âüü‰ø°Âè∑ÁöÑÂü∫Êú¨ÂΩ¢ÂºèËøõË°åÊï∞Â≠¶ÂàÜËß£„ÄÇ

---

## 1. Êï∞Â≠¶Â∫ïÂ±ÇÈÄªËæëÔºö$e^{st}$ ÁöÑÊãÜËß£

ÂΩìÊàë‰ª¨ËØ¥‰∏Ä‰∏™Á≥ªÁªüÁöÑÊûÅÁÇπÊòØ $s = \sigma + j\omega$ Êó∂ÔºåÂÆÉÂú®Êó∂Âüü‰∏≠ÂØπÂ∫îÁöÑÂü∫Êú¨ÂìçÂ∫îÂΩ¢ÂºèÊòØ $e^{st}$„ÄÇÂà©Áî®ÊåáÊï∞ËøêÁÆóÊ≥ïÂàôÔºåÊàë‰ª¨ÂèØ‰ª•Â∞ÜÂÖ∂ÊãÜËß£‰∏∫Ôºö

$$e^{st} = e^{(\sigma + j\omega)t} = \underbrace{e^{\sigma t}}_{\text{ÂπÖÂ∫¶ÊéßÂà∂ÔºàÂÆûÈÉ®Ôºâ}} \cdot \underbrace{e^{j\omega t}}_{\text{Áõ∏‰ΩçÊéßÂà∂ÔºàËôöÈÉ®Ôºâ}}$$

ÈÄöËøáËøô‰∏™ÂÖ¨ÂºèÔºåÊàë‰ª¨ÂèØ‰ª•ÁúãÂà∞ $\sigma$ Âíå $\omega$ ÂàÜÂ∑•ÊòéÁ°ÆÂú∞ÊéßÂà∂ÁùÄÊ≥¢ÂΩ¢ÁöÑ‚ÄúËÉñÁò¶‚Äù‰∏é‚ÄúÁñèÂØÜ‚Äù„ÄÇ



---

## 2. ÂÆûÈÉ® $\sigma$ÔºöË°∞ÂáèÂõ†Â≠ê‰∏éÁ®≥ÂÆöÊÄß

$\sigma$ÔºàÂÆûÈÉ®ÔºâÂÜ≥ÂÆö‰∫Ü‰ø°Âè∑ËÉΩÈáèÈöèÊó∂Èó¥ÁöÑÂèòÂåñË∂ãÂäø„ÄÇ

* **Áâ©ÁêÜÊú¨Ë¥®**ÔºöÂú® $e^{\sigma t}$ ‰∏≠Ôºå$\sigma$ ‰Ωç‰∫éÊåáÊï∞‰ΩçÁΩÆ„ÄÇ
* **Êî∂ÊïõÔºàÁ®≥ÂÆöÔºâ**ÔºöÂΩì $\sigma < 0$ÔºàÂ∑¶ÂçäÂπ≥Èù¢ÔºâÊó∂Ôºå$e^{\sigma t}$ ÈöèÊó∂Èó¥Ë∂ã‰∫é $0$„ÄÇËøôÂÉèÊòØ‰∏Ä‰∏™Êî∂Áº©ÁöÑ‚ÄúÂåÖÁªúÁ∫ø‚ÄùÔºåÂØºËá¥ÂìçÂ∫îÈÄêÊ∏êÊ∂àÂ§±„ÄÇ
* **ÂèëÊï£Ôºà‰∏çÁ®≥ÂÆöÔºâ**ÔºöÂΩì $\sigma > 0$ÔºàÂè≥ÂçäÂπ≥Èù¢ÔºâÊó∂ÔºåÂìçÂ∫îÂëàÊåáÊï∞Á∫ßÂ¢ûÈïø„ÄÇ
* **‰∏¥ÁïåÁä∂ÊÄÅ**ÔºöÂΩì $\sigma = 0$ÔºàËôöËΩ¥‰∏äÔºâÊó∂ÔºåÂπÖÂ∫¶ÊÅíÂÆö„ÄÇ
* **Â∑•Á®ãÁõ¥Ëßâ**Ôºö$\sigma$ ÊèèËø∞‰∫ÜÁ≥ªÁªüÊ∂àËÄóËÉΩÈáèÁöÑËÉΩÂäõÔºàÂ¶ÇÁîµÈòªÊ∂àËÄóÁîµËç∑Ôºâ„ÄÇ$\sigma$ ÁöÑÁªùÂØπÂÄºË∂äÂ§ßÔºåË°∞ÂáèË∂äÂø´ÔºåÁ≥ªÁªüË∂äÊó©ÊÅ¢Â§çÈùôÊ≠¢„ÄÇÂõ†Ê≠§ÔºåÂÆÉË¢´Áß∞‰∏∫**Ë°∞ÂáèÂõ†Â≠ê**ÔºàAttenuation FactorÔºâ„ÄÇ

---

## 3. ËôöÈÉ® $\omega$ÔºöËßíÈ¢ëÁéá‰∏éÈúáËç°ÊÄß

$\omega$ÔºàËôöÈÉ®ÔºâÂÜ≥ÂÆö‰∫Ü‰ø°Âè∑ÁöÑÈúáËç°È¢ëÁéáÔºåË¢´Áß∞‰∏∫**ËßíÈ¢ëÁéá**ÔºàAngular FrequencyÔºâ„ÄÇ

* **Áâ©ÁêÜÊú¨Ë¥®**ÔºöÊ†πÊçÆÊ¨ßÊãâÂÖ¨ÂºèÔºå$e^{j\omega t} = \cos(\omega t) + j\sin(\omega t)$„ÄÇ
* Ëøô‰ª£Ë°®‰∏Ä‰∏™Âú®Âçï‰ΩçÂúÜ‰∏äÊóãËΩ¨ÁöÑÁü¢Èáè„ÄÇÂÆÉ‰∏çÊîπÂèò‰ø°Âè∑ÁöÑÂ§ßÂ∞èÔºåÂè™ÊîπÂèòÊñπÂêëÔºàÁõ∏‰ΩçÔºâ„ÄÇ
* $\omega$ Ë∂äÂ§ßÔºåÊóãËΩ¨Ë∂äÂø´ÔºåÊó∂ÂüüÊ≥¢ÂΩ¢ÈúáËç°Ë∂äÈ¢ëÁπÅ„ÄÇ
* **Â∑•Á®ãÁõ¥Ëßâ**Ôºö$\omega$ ÊèèËø∞‰∫ÜÁ≥ªÁªüËÉΩÈáè‰∫§Êç¢ÁöÑÈÄüÂ∫¶ÔºàÂ¶ÇÁîµÊÑü‰∏éÁîµÂÆπ„ÄÅÂä®ËÉΩ‰∏éÂäøËÉΩ‰πãÈó¥ÁöÑÂæÄÂ§çËΩ¨ÁßªÔºâ„ÄÇ

---

## 4. ÂçèÂêå‰ΩúÁî®Ôºö‰∏âÁª¥ËßÜËßí‰∏ãÁöÑ‚ÄúËÆæËÆ°Â∏à‚Äù

Êàë‰ª¨ÂèØ‰ª•Êää $\sigma$ Âíå $\omega$ Áúã‰ΩúÊòØÂàÜÂ∑•ÊòéÁ°ÆÁöÑ‰∏§‰Ωç‚ÄúËÆæËÆ°Â∏à‚ÄùÔºö

| ÂèÇÊï∞ | ËßíËâ≤ | ÂÜ≥ÂÆö‰∫Ü‰ªÄ‰πàÔºü | Êó∂ÂüüÁâπÂæÅ |
| :--- | :--- | :--- | :--- |
| **$\sigma$ (ÂÆûÈÉ®)** | ÂπÖÂ∫¶ËÆæËÆ° | Á≥ªÁªüÁöÑ**Á®≥ÂÆöÊÄß** | ÂÜ≥ÂÆöÊ≥¢ÂΩ¢ÊòØÊ∂àÂ§±ÔºàË°∞ÂáèÔºâ„ÄÅÊÅíÂÆö„ÄÅËøòÊòØÁàÜÁÇ∏ÔºàÂèëÊï£Ôºâ„ÄÇ |
| **$\omega$ (ËôöÈÉ®)** | È¢ëÁéáËÆæËÆ° | Á≥ªÁªüÁöÑ**ÈúáËç°ÊÄß** | ÂÜ≥ÂÆöÊ≥¢ÂΩ¢ÊØèÁßíÈíüÈúáËç°ÔºàË∑≥Âä®ÔºâÁöÑÊ¨°Êï∞„ÄÇ |

**ÂêàÊàêÊïàÊûúÔºöÈúáËç°Ë°∞ÂáèÊ≥¢ÔºàDamped Sine WaveÔºâ**
ÂΩìÊûÅÁÇπ‰∏∫ÂÖ±ËΩ≠Â§çÊï∞Êó∂Ôºå$\sigma$ ÂÜ≥ÂÆöÂåÖÁªúÁ∫øÁöÑÊî∂Áº©ÈÄüÂ∫¶ÔºåËÄå $\omega$ ÂÜ≥ÂÆöÂÜÖÈÉ®Ê≥¢Âä®ÁöÑÁñèÂØÜ„ÄÇ

## 5. ÁªºÂêàÂà§ÂÆöÔºö$s$ Âπ≥Èù¢‰ΩçÁΩÆ‰∏éÊ≥¢ÂΩ¢Êò†Â∞Ñ



| ÊûÅÁÇπ‰ΩçÁΩÆ | Êó∂ÂüüÊ≥¢ÂΩ¢ÁâπÂæÅ | Áâ©ÁêÜÁä∂ÊÄÅÊèèËø∞ |
| :--- | :--- | :--- |
| **Ë¥üÂÆûËΩ¥‰∏ä** ($\omega=0, \sigma<0$) | Âπ≥ÊªëË°∞ÂáèÔºåÊó†ÈúáËç° | ËøáÈòªÂ∞º (Overdamped) Êàñ ‰∏¥ÁïåÈòªÂ∞º |
| **Â∑¶ÂçäÂπ≥Èù¢ÂÖ±ËΩ≠Â§çÊï∞** | ÈúáËç°Ë°∞ÂáèÔºåÊúÄÁªàÊî∂Êïõ | Ê¨†ÈòªÂ∞º (Underdamped) |
| **ËôöËΩ¥‰∏äÂÖ±ËΩ≠Â§çÊï∞** ($\sigma=0$) | Á≠âÂπÖÈúáËç°Ôºå‰∏çÊî∂Êïõ‰πü‰∏çÂèëÊï£ | Êó†ÈòªÂ∞º (Undamped) |
| **Âè≥ÂçäÂπ≥Èù¢** ($\sigma>0$) | Ê≥¢ÂΩ¢‰∏çÊñ≠Êâ©Â§ßÔºåÊúÄÁªàÂèëÊï£ | ‰∏çÁ®≥ÂÆöÁ≥ªÁªü (Unstable) |

## üéì ÊÄªÁªì‰∏éÊ≥®ÊÑè‰∫ãÈ°π

### Áâ©ÁêÜÁõ¥ËßâÔºö
* **Â∑¶ÁßªÁ≥ªÁªüÊõ¥Á®≥ÂÆö**ÔºöÊûÅÁÇπÂêëÂ∑¶Âπ≥ÁßªÔºå$\sigma$ ÁöÑÁªùÂØπÂÄºÂèòÂ§ßÔºåÁ≥ªÁªüËØØÂ∑ÆÊ∂àÂ§±Êõ¥Âø´„ÄÇ
* **‰∏ªÂØºÊûÅÁÇπ**ÔºöÂú®Â§öÊûÅÁÇπÁ≥ªÁªü‰∏≠ÔºåË∑ùÁ¶ªËôöËΩ¥ÊúÄËøëÁöÑÊûÅÁÇπÊ∂àÂ§±ÊúÄÊÖ¢ÔºåÂÆÉ‰∏ªÂØº‰∫ÜÊàë‰ª¨ÊúÄÁªàÁúãÂà∞ÁöÑÊ≥¢ÂΩ¢ÁâπÂæÅ„ÄÇ
* **ÂèÇÊï∞ÂÖ≥ËÅî**ÔºöÂú® RLC ÁîµË∑Ø‰∏≠ÔºåÂáèÂ∞èÁîµÈòª $R$ ‰ºö‰ΩøÊûÅÁÇπ‰ªéÂÆûËΩ¥‚ÄúÁÇ∏ÂºÄ‚ÄùËøõÂÖ•Â§çÂπ≥Èù¢ÔºåÁ≥ªÁªüÁî±ÂçïË∞ÉË°∞ÂáèËΩ¨‰∏∫ÈúáËç°ÂìçÂ∫î„ÄÇ

### ‚ö†Ô∏è Á¨¶Âè∑ËØ¥ÊòéÔºö
* **$\omega$ (Â∞èÂÜô)**ÔºöÊ®°ÊãüËßíÈ¢ëÁéáÔºàrad/sÔºâÔºåËøûÁª≠Á≥ªÁªü $s$ Âπ≥Èù¢ÁöÑÊ†áÂáÜÂçï‰Ωç„ÄÇ
* **$\Omega$ (Â§ßÂÜô)**ÔºöÂ∏∏Áî®‰∫éÁ¶ªÊï£Á≥ªÁªüÔºà$z$ ÂèòÊç¢ÔºâÊàñÊï∞Â≠ó‰ø°Âè∑Â§ÑÁêÜÔºåÂå∫ÂàÜ‰∏çÂêåÁöÑÈ¢ëÁéáËΩ¥„ÄÇ
* Âú®ÂΩìÂâçÁöÑÊãâÊôÆÊãâÊñØÂèòÊç¢Â≠¶‰π†‰∏≠ÔºåÂÆÉ‰ª¨ÈÄöÂ∏∏ÈÉΩÊåáÂêëÂ§çÂπ≥Èù¢ÁöÑËôöËΩ¥ÂùêÊ†á„ÄÇ

# üßÆ ËÆ°ÁÆó RLC ÁîµË∑ØÁöÑÊûÅÁÇπÔºöÂª∫Ê®°‰∏éÂàÜÊûê

ËÆ°ÁÆó RLC ÁîµË∑ØÁöÑÊûÅÁÇπÊòØÊéßÂà∂ÁêÜËÆ∫‰∏≠‚ÄúÂª∫Ê®°‰∏éÂàÜÊûê‚ÄùÁöÑÊ†∏ÂøÉÊ≠•È™§„ÄÇÊ†πÊçÆËßÜÈ¢ëÊé®ÂØºÂíå‰Ω†ÁöÑÂ≠¶‰π†Â∑•‰ΩúÊµÅÔºåÊàë‰ª¨ÂèØ‰ª•Â∞ÜËÆ°ÁÆóËøáÁ®ãÂàÜ‰∏∫‰ª•‰∏ãÂõõ‰∏™ÈÄªËæëÈò∂ÊÆµÔºö

---

### 1. Âª∫Á´ãÊó∂ÂüüÂæÆÂàÜÊñπÁ®ã
È¶ñÂÖàÔºåÊ†πÊçÆ**Âü∫Â∞îÈúçÂ§´ÁîµÂéãÂÆöÂæãÔºàKVLÔºâ**ÔºåËßÜÈ¢ë‰∏≠Âª∫Á´ãÁöÑÁîµË∑ØÂä®ÊÄÅÊñπÁ®ã‰∏∫Ôºö

$$e'(t) = L i''(t) + R i'(t) + \frac{1}{C} i(t)$$



[Image of an RLC series circuit diagram]


ÂÖ∂‰∏≠Ôºö
* $e'(t)$ ‰∏∫ËæìÂÖ•ÁîµÂéãÁöÑÂØºÊï∞
* $i(t)$ ‰∏∫ÂõûË∑ØÁîµÊµÅ
* $L, R, C$ ÂàÜÂà´‰∏∫ÁîµÊÑü„ÄÅÁîµÈòªÂíåÁîµÂÆπ

---

### 2. ËøõË°åÊãâÊôÆÊãâÊñØÂèòÊç¢
ÂÅáËÆæÂàùÂßãÊù°‰ª∂‰∏∫Èõ∂Ôºà$i(0)=0, i'(0)=0$ÔºâÔºåÂØπÂæÆÂàÜÊñπÁ®ã‰∏§ËæπÂêåÊó∂ËøõË°åÊãâÊôÆÊãâÊñØÂèòÊç¢Ôºö
* $i''(t) \to s^2 I(s)$
* $i'(t) \to s I(s)$
* $i(t) \to I(s)$

ÂèòÊç¢ÂêéÁöÑ‰ª£Êï∞ÊñπÁ®ã‰∏∫Ôºö
$$sE(s) = (L s^2 + R s + \frac{1}{C}) I(s)$$

---

### 3. ÊûÑÂª∫‰º†ÈÄíÂáΩÊï∞‰∏éÁâπÂæÅÊñπÁ®ã
ÊûÅÁÇπÁöÑÂÆö‰πâÊòØ**‰º†ÈÄíÂáΩÊï∞ÂàÜÊØçÂ§öÈ°πÂºèÁöÑÊ†π**„ÄÇ

‰º†ÈÄíÂáΩÊï∞ $G(s)$ ‰∏∫Ôºö
$$G(s) = \frac{I(s)}{E(s)} = \frac{s}{L s^2 + R s + \frac{1}{C}}$$

‰ª§ÂàÜÊØçÁ≠â‰∫éÈõ∂ÔºåÂæóÂà∞Á≥ªÁªüÁöÑ**ÁâπÂæÅÊñπÁ®ã**Ôºö
$$L s^2 + R s + \frac{1}{C} = 0$$

---

### 4. Ê±ÇËß£ÊûÅÁÇπ
‰ΩøÁî®Ê±ÇÊ†πÂÖ¨Âºè $s = \frac{-b \pm \sqrt{b^2 - 4ac}}{2a}$ÔºåËÆ°ÁÆóÁâπÂæÅÊñπÁ®ãÁöÑ‰∏§‰∏™Ê†πÔºàÂç≥ÊûÅÁÇπ $s_{1,2}$ÔºâÔºö

$$s_{1,2} = \frac{-R \pm \sqrt{R^2 - \frac{4L}{C}}}{2L}$$

Êï¥ÁêÜÂêéÂèØÂæóÊûÅÁÇπË°®ËææÂºèÔºö
$$s_{1,2} = -\frac{R}{2L} \pm \sqrt{\left(\frac{R}{2L}\right)^2 - \frac{1}{LC}}$$



---

## üìä ÊûÅÁÇπÁªìÊûúÁöÑÁâ©ÁêÜÊÑè‰πâ

Ê†πÊçÆÊ†πÂè∑‰∏ãÁöÑÂà§Âà´Âºè $\Delta = R^2 - \frac{4L}{C}$ ÁöÑÊ≠£Ë¥üÔºåÁ≥ªÁªü‰ºöÂá∫Áé∞‰∏çÂêåÁöÑÂìçÂ∫îÁä∂ÊÄÅÔºö

1.  **ËøáÈòªÂ∞º (Overdamped)**Ôºö
    * **Êù°‰ª∂**ÔºöËã• $R^2 > \frac{4L}{C}$
    * **ÊûÅÁÇπ**Ôºö‰∏§‰∏™‰∏çÂêåÁöÑË¥üÂÆûÊï∞„ÄÇ
    * **Ë°®Áé∞**ÔºöÂìçÂ∫îÂπ≥ÊªëË°∞ÂáèÔºåÊó†ÈúáËç°„ÄÇ
2.  **‰∏¥ÁïåÈòªÂ∞º (Critically Damped)**Ôºö
    * **Êù°‰ª∂**ÔºöËã• $R^2 = \frac{4L}{C}$
    * **ÊûÅÁÇπ**Ôºö‰∏§‰∏™ÈáçÂêàÁöÑË¥üÂÆûÊï∞„ÄÇ
3.  **Ê¨†ÈòªÂ∞º (Underdamped)**Ôºö
    * **Êù°‰ª∂**ÔºöËã• $R^2 < \frac{4L}{C}$
    * **ÊûÅÁÇπ**Ôºö‰∏ÄÂØπÂÖ±ËΩ≠Â§çÊï∞„ÄÇ
    * **Ë°®Áé∞**ÔºöÂìçÂ∫î‰ºöÂá∫Áé∞Ë°∞ÂáèÁöÑÈúáËç°ÔºåËôöÈÉ®Â§ßÂ∞èÂÜ≥ÂÆö‰∫ÜÈúáËç°È¢ëÁéá„ÄÇ


# Ê†∏ÂøÉÂÖ¨ÂºèÈ™åËØÅ

```python
import sympy
from sympy import symbols, exp, sin, laplace_transform, Equality
# Ê†∏ÂøÉ‰øÆÂ§çÔºö‰ªé IPython.display ÂØºÂÖ• display
from IPython.display import display, Latex

# ÂàùÂßãÂåñ SymPy ÁöÑÊâìÂç∞Á≥ªÁªüÔºå‰ΩøÂÖ∂‰ºòÂÖà‰ΩøÁî® LaTeX
sympy.init_printing(use_latex=True)

# 1. Á¨¶Âè∑ÂÆö‰πâ
# real=True Âíå positive=True Â∏ÆÂä© SymPy Â§ÑÁêÜÊãâÊôÆÊãâÊñØÁßØÂàÜÁöÑÊî∂ÊïõÊù°‰ª∂
t, s = symbols('t s', real=True, positive=True)
a = symbols('a', real=True, positive=True)

def verify_math_derivations():
    """
    Âà©Áî® SymPy È™åËØÅËßÜÈ¢ë‰∏≠ÁöÑÊãâÊôÆÊãâÊñØÂèòÊç¢ÊâãÁÆóÁªìÊûú
    """
    print(">>> ÂºÄÂßãÁ¨¶Âè∑È™åËØÅÔºöÊãâÊôÆÊãâÊñØÂèòÊç¢Ê†∏ÂøÉÂÖ¨Âºè <<<\n")

    # --- Ê°à‰æã 1ÔºöÈ™åËØÅÊåáÊï∞ÂáΩÊï∞ L{e^(-at)} = 1 / (s + a) ---
    #
    f_exp = exp(-a * t)
    # laplace_transform ËøîÂõû: (ÁªìÊûú, Êî∂ÊïõÂπ≥Èù¢, Êî∂ÊïõÊù°‰ª∂)
    F_exp = laplace_transform(f_exp, t, s)[0] 
    
    display(Latex(r"**1. È™åËØÅÊåáÊï∞ÂáΩÊï∞Ôºö**"))
    display(Equality(sympy.Symbol(r"\mathcal{L}\{e^{-at}\}"), F_exp))

    # --- Ê°à‰æã 2ÔºöÈ™åËØÅÊ≠£Âº¶ÂáΩÊï∞ L{sin(at)} = a / (s^2 + a^2) ---
    #
    f_sin = sin(a * t)
    F_sin = laplace_transform(f_sin, t, s)[0]
    
    display(Latex(r"**2. È™åËØÅÊ≠£Âº¶ÂáΩÊï∞Ôºö**"))
    display(Equality(sympy.Symbol(r"\mathcal{L}\{\sin(at)\}"), F_sin))

    # --- Ê°à‰æã 3ÔºöÈ™åËØÅÂæÆÂàÜÊÄßË¥® L{f'(t)} = sF(s) - f(0) ---
    # ËßÜÈ¢ëÊé®ÂØº RLC ÂæÆÂàÜÊñπÁ®ãÁöÑÂÖ≥ÈîÆ
    f = sympy.Function('f')(t)
    F_diff = laplace_transform(f.diff(t), t, s, noconds=True)
    
    display(Latex(r"**3. È™åËØÅ‰∏ÄÈò∂ÂæÆÂàÜÊÄßË¥®Ôºö**"))
    display(Equality(sympy.Symbol(r"\mathcal{L}\{f'(t)\}"), F_diff))

# ËøêË°åÈ™åËØÅÈÄªËæë
verify_math_derivations()
```

# Âç∑ÁßØËøáÁ®ãÊºîÁ§∫‰ª£Á†Å

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
from IPython.display import HTML
import warnings
# ÂøΩÁï•Êù•Ëá™ matplotlib.animation Ê®°ÂùóÁöÑÁâπÂÆö UserWarning
warnings.filterwarnings('ignore', message='Animation was deleted without rendering anything')

# 1. ÂáÜÂ§á‰ø°Âè∑
dt = 0.05
t = np.arange(-2, 10, dt)
tau = np.arange(-2, 10, dt)

# ÂÆö‰πâ f(tau): Á≥ªÁªüËÑâÂÜ≤ÂìçÂ∫î (e^-t)
f_tau = np.where(tau >= 0, np.exp(-tau), 0)
# ÂÆö‰πâ g(tau): ËæìÂÖ•‰ø°Âè∑ (Èò∂Ë∑É‰ø°Âè∑)
g_tau = np.where(tau >= 0, 1, 0)

# 2. ËÆæÁΩÆÁªòÂõæÁîªÂ∏É
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(10, 8))
plt.subplots_adjust(hspace=0.4)

def update(t_val):
    ax1.clear()
    ax2.clear()
    
    # ÁøªËΩ¨Âπ∂ÁßªÂä®ÂêéÁöÑ‰ø°Âè∑ g(t - tau)
    g_shifted = np.where((t_val - tau) >= 0, 1, 0)
    # ÈáçÂè†ÈÉ®ÂàÜ (‰πòÁßØ)
    product = f_tau * g_shifted
    # ÂΩìÂâçÂç∑ÁßØÂÄº (Èù¢ÁßØÁßØÂàÜ)
    conv_val = np.sum(product) * dt
    
    # --- Â≠êÂõæ 1: Â±ïÁ§∫ÊªëÂä®‰∏é‰πòÁßØËøáÁ®ã ---
    ax1.plot(tau, f_tau, 'b', label='Á≥ªÁªüÁâπÊÄß $f(\\tau) = e^{-\\tau}$')
    ax1.plot(tau, g_shifted, 'g--', label=f'ÁßªÂä®‰ø°Âè∑ $g({t_val:.1f} - \\tau)$')
    ax1.fill_between(tau, product, color='red', alpha=0.3, label='ÈáçÂè†‰πòÁßØÂå∫ (ÁßØÂàÜÈù¢ÁßØ)')
    ax1.set_title('Âç∑ÁßØÁâ©ÁêÜËøáÁ®ãÔºöÁøªËΩ¨‰∏éÊªëÂä® (Time Domain)', fontsize=12)
    ax1.set_xlim(-2, 8)
    ax1.set_ylim(-0.2, 1.2)
    ax1.legend(loc='upper right')
    ax1.grid(True, alpha=0.3)
    
    # --- Â≠êÂõæ 2: Âä®ÊÄÅÁªòÂà∂ÁªìÊûúÊõ≤Á∫ø ---
    t_history = np.arange(0, t_val + dt, dt)
    # Ëß£ÊûêËß£Ôºö1 - e^-t
    res_history = 1 - np.exp(-t_history)
    
    ax2.plot(t_history, res_history, 'r', lw=2, label='Âç∑ÁßØÁªìÊûú $h(t) = \\int f(\\tau)g(t-\\tau)d\\tau$')
    ax2.scatter(t_val, conv_val, color='red') # ÂΩìÂâçÁÇπ
    ax2.set_title('ÁªìÊûúËΩ®ËøπÔºöÈáçÂè†Èù¢ÁßØÁöÑÁ¥ØÁßØ', fontsize=12)
    ax2.set_xlim(-2, 8)
    ax2.set_ylim(-0.2, 1.2)
    ax2.set_xlabel('Êó∂Èó¥ (t)')
    ax2.legend(loc='lower right')
    ax2.grid(True, alpha=0.3)

# 3. ÁîüÊàêÂä®Áîª
ani = FuncAnimation(fig, update, frames=np.arange(0, 7, 0.2), interval=100)
plt.close() # Èò≤Ê≠¢ÊòæÁ§∫ÈùôÊÄÅÈáçÂ§çÂõæ

# Âú® Jupyter ‰∏≠‰ª• HTML5 ËßÜÈ¢ëÊ†ºÂºèËæìÂá∫
from IPython.display import HTML
HTML(ani.to_jshtml())
```

## üìñ Âç∑ÁßØÂä®ÁîªÊ∑±Â∫¶Ëß£ÊûêÔºöÁâ©ÁêÜÁõ¥Ëßâ‰∏éÊï∞Â≠¶Êú¨Ë¥®

ËøôÊÆµÂä®ÁîªÁõ¥ËßÇÂú∞Â±ïÁ§∫‰∫Ü Êó∂ÂüüÂç∑ÁßØ (Time-Domain Convolution) ÁöÑÂä®ÊÄÅÂÖ®ËøáÁ®ã„ÄÇÂú®ÊéßÂà∂ÁêÜËÆ∫‰∏≠ÔºåÁêÜËß£Âç∑ÁßØÊòØÁêÜËß£‚ÄúÁ≥ªÁªüÂ¶Ç‰ΩïÂìçÂ∫îÂ§ñÈÉ®ËæìÂÖ•‚ÄùÁöÑÂÖ≥ÈîÆ„ÄÇ

1. ‰ø°Âè∑ÂÆö‰πâ‰∏éÁâ©ÁêÜËÉåÊôØ
* ËæìÂÖ•‰ø°Âè∑ $g(\tau)$ÔºöÂçï‰ΩçÈò∂Ë∑É‰ø°Âè∑ (Unit Step Signal)
  * Êï∞Â≠¶Ë°®ËææÔºö$g(t) = 1$ (ÂΩì $t \ge 0$)„ÄÇ
  * Áâ©ÁêÜÊÑè‰πâÔºö‰ª£Ë°®Âú® $t=0$ Êó∂ÂàªÔºåÁ≥ªÁªüÊé•ÂÖ•‰∫Ü‰∏Ä‰∏™ÊÅíÂÆöÁöÑÂ§ñÈÉ®Âà∫ÊøÄÔºà‰æãÂ¶ÇÔºöÂêà‰∏äÂºÄÂÖ≥ÔºåÊé•ÈÄö‰∏Ä‰∏™ 1V ÁöÑÁõ¥ÊµÅÁîµÊ∫êÔºâ„ÄÇ
* Á≥ªÁªüÁâπÊÄß $f(\tau)$ÔºöËÑâÂÜ≤ÂìçÂ∫î (Impulse Response)
  * Êï∞Â≠¶Ë°®ËææÔºö$f(t) = e^{-t}$„ÄÇ
  * Áâ©ÁêÜÊÑè‰πâÔºö‰ª£Ë°®Á≥ªÁªüÊú¨Ë∫´ÁöÑ‚ÄúÈÅó‰º†Âü∫Âõ†‚ÄùÊàñ‚ÄúËÆ∞ÂøÜ‚Äù„ÄÇ$e^{-t}$ ÁöÑË°∞ÂáèÂΩ¢ÂºèÊÑèÂë≥ÁùÄÁ≥ªÁªüÂØπË∂ä‰πÖËøúÁöÑËæìÂÖ•ÔºåËÆ∞ÂøÜË∂äÊ®°Á≥ä„ÄÇ

2. Âç∑ÁßØËøáÁ®ãÁöÑ‚ÄúÂõõÊ≠•Êõ≤‚Äù

Âä®Áîª‰∏≠ÁöÑÂ≠êÂõæ 1 Â±ïÁ§∫‰∫ÜÂç∑ÁßØÁßØÂàÜÂÖ¨Âºè $(f * g)(t) = \int_{0}^{t} f(\tau)g(t-\tau)d\tau$ ÁöÑÂá†‰ΩïÂÆûÁé∞Ôºö

* ÁøªËΩ¨ (Flip)ÔºöÂÖ¨Âºè‰∏≠ÁöÑ $-\tau$ Â∞ÜËæìÂÖ•‰ø°Âè∑ $g(\tau)$ ËøõË°å‰∫ÜÈïúÂÉèÁøªËΩ¨„ÄÇÂéüÊú¨ÂêëÂè≥Âª∂‰º∏ÁöÑÈò∂Ë∑É‰ø°Âè∑ÔºåÁé∞Âú®ÂèòÊàê‰∫ÜÂêëÂ∑¶Áúã„ÄÇ
* Âπ≥Áßª (Shift)ÔºöÈöèÁùÄÊó∂Èó¥ $t$ÔºàÂä®Áîª‰∏≠ÁöÑ t_valÔºâ‰∏çÊñ≠Â¢ûÂä†ÔºåÁªøËâ≤ËôöÁ∫ø‰ª£Ë°®ÁöÑÁøªËΩ¨‰ø°Âè∑‰∏çÊñ≠ÂêëÂè≥ÊªëÂä®„ÄÇËøô‰ª£Ë°®Á≥ªÁªüÂú® $t$ Êó∂ÂàªÔºå‚ÄúÂõûÁúã‚Äù‰ªé $0$ Âà∞ $t$ Êó∂ÂàªÁöÑÊâÄÊúâÂéÜÂè≤ËæìÂÖ•„ÄÇ
* Áõ∏‰πò (Multiply)ÔºöÁ≥ªÁªüÁâπÊÄßÔºàËìùËâ≤ÂÆûÁ∫øÔºâ‰∏éÂΩìÂâçÊó∂ÂàªË¶ÜÁõñÁöÑËæìÂÖ•‰ø°Âè∑ÔºàÁªøËâ≤ËôöÁ∫øÔºâÁõ∏‰πò„ÄÇ
* ÁßØÂàÜ (Integrate)ÔºöÁ∫¢Ëâ≤Èò¥ÂΩ±Âå∫ÂüüÁöÑÈù¢ÁßØÂ∞±ÊòØÁõ∏‰πòÂêéÁöÑÁßØÂàÜÁªìÊûú„ÄÇËøô‰∏™Èù¢ÁßØÂÄºÁõ¥Êé•ÂÜ≥ÂÆö‰∫ÜÁ≥ªÁªüÂú®ÂΩìÂâçÊó∂Âàª $t$ ÁöÑËæìÂá∫Â§ßÂ∞è„ÄÇ

3. ÁªìÊûúÂàÜÊûêÔºöÈò∂Ë∑ÉÂìçÂ∫î (Step Response)

Â≠êÂõæ 2 Â±ïÁ§∫‰∫ÜÂç∑ÁßØÁöÑÊúÄÁªà‰∫ßÂá∫‚Äî‚ÄîÈò∂Ë∑ÉÂìçÂ∫îÊõ≤Á∫øÔºö
* Âä®ÊÄÅÁ¥ØÁßØÔºöÈöèÁùÄÁªøËâ≤‰ø°Âè∑ÂêëÂè≥ÊªëÂä®ÔºåÁ∫¢Ëâ≤Èò¥ÂΩ±Èù¢ÁßØË∂äÊù•Ë∂äÂ§ßÔºåÂõ†Ê≠§‰∏ãÊñπÁöÑÁ∫¢Á∫øÔºàËæìÂá∫ÁªìÊûúÔºâ‰∏çÊñ≠‰∏äÂçá„ÄÇ
* Ë∂ã‰∫éÈ•±ÂíåÔºöÁî±‰∫éÁ≥ªÁªüÁâπÊÄß $e^{-\tau}$ Âú®‰∏çÊñ≠Ë°∞ÂáèÔºåÊñ∞Â¢ûÈù¢ÁßØÁöÑË¥°ÁåÆË∂äÊù•Ë∂äÂ∞èÔºåÊõ≤Á∫øÊúÄÁªà‰ºöÊó†ÈôêÊé•Ëøë‰∫é $1$„ÄÇ
* Êï∞Â≠¶È™åËØÅÔºöËØ•ËøáÁ®ãÁöÑËß£ÊûêËß£‰∏∫ $h(t) = \int_{0}^{t} 1 \cdot e^{-\tau} d\tau = 1 - e^{-t}$„ÄÇ

## üí° ‰∏∫‰ªÄ‰πàÊàë‰ª¨Ë¶ÅÂ≠¶‰π†ÊãâÊôÆÊãâÊñØÂèòÊç¢Ôºü

ËßÇÂØüÂä®ÁîªÂèØ‰ª•ÂèëÁé∞ÔºåÊó∂ÂüüÂç∑ÁßØÊ∂âÂèäÂ§çÊùÇÁöÑÂõæÂΩ¢ÊªëÂä®ÂíåÈù¢ÁßØÁßØÂàÜÔºåÊâãÂä®ËÆ°ÁÆóÊûÅÂÖ∂ÁπÅÁêê„ÄÇËÄåÊ†πÊçÆÂç∑ÁßØÂÆöÁêÜÔºö

Êó∂ÂüüÁöÑÂç∑ÁßØ = È¢ëÂüüÁöÑ‰πòÊ≥ï

Âú® $s$ Âüü‰∏≠ÔºåÊàë‰ª¨Âè™ÈúÄË¶ÅÂ∞ÜÁ≥ªÁªüÁöÑ‰º†ÈÄíÂáΩÊï∞ $F(s) = \frac{1}{s+1}$ ‰∏éËæìÂÖ•ÁöÑÊãâÊ∞èÂèòÊç¢ $G(s) = \frac{1}{s}$ Áõ∏‰πòÔºåÂç≥ÂèØÁõ¥Êé•ÂæóÂà∞ËæìÂá∫ÁöÑË°®ËææÂºèÔºåÂΩªÂ∫ïË∑≥Ëøá‰∫ÜËøô‰∏™‚ÄúÊªëÂä®ÁßØÂàÜ‚ÄùÁöÑÁóõËã¶ËøáÁ®ã„ÄÇËøôÂ∞±ÊòØ DR_CAN ËÄÅÂ∏àÂº∫Ë∞ÉÁöÑÊãâÊôÆÊãâÊñØÂèòÊç¢ÁöÑÂ∑•Á®ãÂ®ÅÂäõ„ÄÇ

## üìñ ËøêË°åËøôÊÆµ‰ª£Á†ÅÂêéÔºåËØ∑ÂÖ≥Ê≥®Ôºö
* Á∫¢Ëâ≤Èò¥ÂΩ±ÈÉ®ÂàÜÔºöÈöèÁùÄÁªøËâ≤ËôöÁ∫øÔºàËæìÂÖ•‰ø°Âè∑ÔºâÂêëÂè≥Êé®ÁßªÔºåÁ∫¢Ëâ≤Èò¥ÂΩ±Èù¢ÁßØË∂äÊù•Ë∂äÂ§ßÔºåËøôÁõ¥ËßÇËß£Èáä‰∫Ü‰∏∫‰ªÄ‰πàÁªìÊûúÊõ≤Á∫øÔºà‰∏ãÊñπÁ∫¢Á∫øÔºâÂú®‰∏çÊñ≠‰∏äÂçá„ÄÇ
* Áâ©ÁêÜÊÑè‰πâÁöÑÂÜÖÂåñÔºö‰Ω†‰ºöÂèëÁé∞Ôºå$e^{-t}$ Ë∂äÂæÄÂêéÁöÑÈÉ®ÂàÜÂØπÂΩìÂâçÁöÑË¥°ÁåÆË∂äÂ∞è„ÄÇËøôÂØπÂ∫î‰∫Ü DR_CAN ËÄÅÂ∏àËÆ≤ÁöÑÔºöÁ≥ªÁªüÂØπË∂ä‰πÖËøúÁöÑËæìÂÖ•‚ÄúËÆ∞ÂøÜ‚ÄùË∂äÊ®°Á≥ä„ÄÇ
* ÂØπÊØîÈ¢ëÂüüÔºöÂú® $s$ ÂüüÔºåËøô‰∏™ËøáÁ®ãË¢´ÁÆÄÂåñÊàê‰∫Ü‰∏Ä‰∏™Áû¨Èó¥ÂÆåÊàêÁöÑ‰ª£Êï∞‰πòÊ≥ïÔºåËøô‰πüÂ∞±ÊòØ‰∏∫‰ªÄ‰πàÂ∑•Á®ãÂ∏àÊõ¥ÊÑøÊÑèÂú®Â§çÈ¢ëÂüüÊÄùËÄÉÈóÆÈ¢ò„ÄÇ

# ÊûÅÁÇπ‰ΩçÁΩÆ‰∏éÁ≥ªÁªüË°å‰∏∫ÂÖ≥ËÅîÂàÜÊûê

```python
import numpy as np
import matplotlib.pyplot as plt
import control
import warnings
from IPython.display import display, Latex

# 1. ÁéØÂ¢ÉÈÖçÁΩÆ
warnings.filterwarnings('ignore', category=FutureWarning)
plt.rcParams['font.sans-serif'] = ['SimHei'] 
plt.rcParams['axes.unicode_minus'] = False

def analyze_pole_stability(R, L, C, case_name):
    """
    ÊèêÂèñÂàÜÊØçÂ§öÈ°πÂºèÔºåËÆ°ÁÆóÊûÅÁÇπÔºåÂπ∂ËßÇÂØüÊó∂ÂüüË°®Áé∞
    """
    # Âª∫Á´ã RLC ‰º†ÈÄíÂáΩÊï∞: G(s) = Cs / (LCs^2 + RCs + 1)
    # ÂàÜÊØçÂç≥‰∏∫ÁâπÂæÅÊñπÁ®ã: LCs^2 + RCs + 1 = 0
    num = [C, 0]
    den = [L * C, R * C, 1]
    sys = control.TransferFunction(num, den)
    
    # --- ‰ªªÂä° 1ÔºöÊèêÂèñÁâπÂæÅÊñπÁ®ãÂπ∂ËÆ°ÁÆóÊûÅÁÇπ ---
    poles = sys.poles()
    
    # --- ‰ªªÂä° 2ÔºöËÆ°ÁÆóÈò∂Ë∑ÉÂìçÂ∫î ---
    t, y = control.step_response(sys, T=np.linspace(0, 10, 1000))
    
    # --- ÁªòÂõæÈÄªËæë ---
    fig, (ax_pz, ax_step) = plt.subplots(1, 2, figsize=(14, 5))
    
    # A. ÊûÅÁÇπÂú® s Âπ≥Èù¢ÁöÑ‰ΩçÁΩÆ
    control.pzmap(sys, plot=True, ax=ax_pz, title=f'{case_name} - ÊûÅÁÇπ‰ΩçÁΩÆ')
    ax_pz.grid(True, linestyle='--', alpha=0.5)
    
    # B. ÁîµÊµÅ i(t) ÁöÑÊó∂ÂüüÁâ©ÁêÜË°®Áé∞
    ax_step.plot(t, y, lw=2)
    ax_step.axhline(0, color='black', lw=1, ls='--')
    ax_step.set_title(f'{case_name} - ÁîµÊµÅ i(t) ÂìçÂ∫î')
    ax_step.set_xlabel('Êó∂Èó¥ (s)')
    ax_step.set_ylabel('ÁîµÊµÅ (A)')
    ax_step.grid(True, alpha=0.3)
    
    plt.tight_layout()
    plt.show()
    
    # C. Êï∞Â≠¶ÂåñÁªìÊûúÂ±ïÁ§∫
    p_text = ", ".join([f"{p.real:.2f} + {p.imag:.2f}j" if abs(p.imag) > 1e-4 else f"{p.real:.2f}" for p in poles])
    display(Latex(f"**{case_name} ÁâπÂæÅÊ†πÔºàÊûÅÁÇπÔºâÔºö** $s_{{1,2}} = \\{{ {p_text} \\}}$"))

# --- ÂØπÊØîÂÆûÈ™å ---

# Ê°à‰æã AÔºöÊûÅÁÇπÂú®Â∑¶ÂçäÂπ≥Èù¢ (R=1, L=0.5, C=0.1) -> ÊúâÈòªÂ∞ºÁ®≥ÂÆö
analyze_pole_stability(R=1.0, L=0.5, C=0.1, case_name="ÂÆûÈ™å AÔºöÊûÅÁÇπÂú®Â∑¶ÂçäÂπ≥Èù¢")

# Ê°à‰æã BÔºöÊûÅÁÇπÂú®ËôöËΩ¥‰∏ä (R=0, L=0.5, C=0.1) -> Êó†ÈòªÂ∞ºÁ≠âÂπÖÈúáËç°
analyze_pole_stability(R=0, L=0.5, C=0.1, case_name="ÂÆûÈ™å BÔºöÊûÅÁÇπÂú®ËôöËΩ¥‰∏ä")
```

1. Â∑¶ÂçäÂπ≥Èù¢ÁöÑÊûÅÁÇπ (Stable Decay)Ôºö
* ÁâπÂæÅÔºöÊûÅÁÇπÂÆûÈÉ® $\sigma < 0$„ÄÇ
* Áâ©ÁêÜË°®Áé∞ÔºöÂç≥‰ΩøÁ≥ªÁªü‰∫ßÁîüÈúáËç°ÔºåÊ≥¢ÂΩ¢ÁöÑÊåØÂπÖ‰πü‰ºöÈöèÁùÄÊó∂Èó¥ÂëàÊåáÊï∞Á∫ßË°∞ÂáèËá≥Èõ∂„ÄÇ
* Â∑•Á®ãÁõ¥ËßâÔºö‰ª£Ë°®ÁîµË∑Ø‰∏≠ÊúâÁîµÈòª $R$ Ê≠£Âú®Ê∂àËÄóËÉΩÈáèÔºåÁ≥ªÁªüÊòØÊ∏êËøëÁ®≥ÂÆöÁöÑ„ÄÇ

2. ËôöËΩ¥‰∏äÁöÑÊûÅÁÇπ (Marginally Stable / Undamped)Ôºö
* ÁâπÂæÅÔºöÊûÅÁÇπÂÆûÈÉ® $\sigma = 0$ÔºåÂç≥ÊûÅÁÇπ‰∏∫‰∏ÄÂØπÁ∫ØËôöÊï∞ $\pm j\omega$„ÄÇ

* Â∑•Á®ãÁõ¥ËßâÔºö‰ª£Ë°®ÁêÜÊÉ≥Áä∂ÊÄÅ‰∏ãÊ≤°ÊúâÁîµÈòªÔºà$R=0$ÔºâÔºåËÉΩÈáèÂú®ÁîµÊÑüÂíåÁîµÂÆπ‰πãÈó¥Êó†ÈôêÂæ™ÁéØÔºå‰πüÂ∞±ÊòØÂ§Ñ‰∫é‚Äú‰∏¥ÁïåÁ®≥ÂÆö‚ÄùÁä∂ÊÄÅ„ÄÇ
