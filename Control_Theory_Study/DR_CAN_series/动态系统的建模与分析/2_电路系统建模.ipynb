## 一、 基础电气元件 (Basic Elements)

在视频的前半部分，老师通过手写板书复习了电路建模的基础元件及其数学关系：

| 元件名称 | 物理量/符号 | 单位 | 数学关系 / 特性 |
| :--- | :--- | :--- | :--- |
| **电量** | $q$ | 库仑 (C) | 电荷的积累量 |
| **电流** | $i$ | 安培 (A) | $i = \frac{dq}{dt}$（电荷随时间的流速） |
| **电压** | $e$ | 伏特 (V) | 两点间的电位差 |
| **电阻 (R)** | $R$ | 欧姆 ($\Omega$) | $e_R = iR$（欧姆定律）；电压方向与电流一致 |
| **电容 (C)** | $C$ | 法拉 (F) | $q = Ce_C \implies e_C = \frac{1}{C} \int_{0}^{t} i dt$（储能元件） |
| **电感 (L)** | $L$ | 亨利 (H) | $e_L = L \frac{di}{dt}$（抵抗电流变化的感应电动势） |

## 二、 基尔霍夫定律 (Kirchhoff's Laws)

板书详细定义了电路分析的两大支柱：

* **基尔霍夫电流定律 (KCL)**：
    * **定义**：所有进入某节点的电流总和等于离开该节点的电流总和。
    * **公式**：$\sum i_{in} = \sum i_{out}$。
* **基尔霍夫电压定律 (KVL)**：
    * **定义**：沿着闭合回路的所有元件两端电压的代数和等于零。
    * **注意点**：定义电流方向为正方向。电阻电压为正 ($+$)，电源电压由负极到正极则为负 ($-$)。

## 三、 案例一：串联 RLC 电路建模

* **电路描述**：输入电压源 $e_i$，电感 $L$，电容 $C$，电阻 $R$ 串联成单一回路。
* **KVL 方程**：
    $$e_L + e_C + e_R - e_i = 0$$
* **代入元件特性**：
    $$L \frac{di}{dt} + \frac{1}{C} \int i dt + iR = e_i$$
* **消去积分项（两边求导）**：
    $$L \frac{d^2i}{dt^2} + R \frac{di}{dt} + \frac{1}{C}i = \frac{de_i}{dt}$$

## 四、 案例二：多回路复杂电路建模 (重点)

* **已知参数**：$L=2H, C=1/4F, R_1=1\Omega, R_2=3\Omega$。
* **目标**：建立输出电压 $e_o$ 与输入电压 $e_i$ 的微分方程。
* **定义电流**：左侧回路电流 $i_1$，右侧回路电流 $i_2$。
* **关键点**：流过中间电容的电流为 $(i_1 - i_2)$。
* **列出方程组**：
    * **Loop 1**: $e_L + e_C - e_i = 0 \implies 2\dot{i}_1 + 4\int(i_1-i_2)dt = e_i$
    * **Loop 2**: $e_{R1} + e_{R2} - e_C = 0 \implies i_2 + 3i_2 - 4\int(i_1-i_2)dt = 0$
* **消元过程**：
    1.  由 Loop 2 得：$4i_2 = 4\int(i_1-i_2)dt \implies i_2 = \int(i_1-i_2)dt$。
    2.  求导得：$\dot{i}_2 = i_1 - i_2 \implies i_1 = \dot{i}_2 + i_2$。
    3.  再次求导得：$\dot{i}_1 = \ddot{i}_2 + \dot{i}_2$。
    4.  代入 Loop 1 并转换至 $e_o$：已知 $e_o = R_2 i_2 = 3i_2 \implies i_2 = \frac{e_o}{3}$。
* **最终结果**：
    $$2\ddot{e}_o + 2\dot{e}_o + 4e_o = 3e_i$$
