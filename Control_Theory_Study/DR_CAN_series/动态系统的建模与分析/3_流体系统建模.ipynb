# 📝 控制理论学习笔记：流体系统建模

这是一份关于基础流体系统动态建模的详细记录，涵盖了从物理概念到微分方程推导的全过程。

---

## 1. 案例引入与建模目标 (00:00 - 00:26)



* **已知条件**：
    * 进口处流量：$q_{in}$
    * 容器横截面积：$A$
    * 出口流阻：$R$
* **目标**：推导液面高度 $h$ 的动态微分方程。

---

## 2. 基本元素与压强概念 (00:27 - 02:18)

在不可压缩均质流体（密度 $\rho$ 为常数）系统中，核心物理量如下：

| 物理量 | 符号 | 单位 | 说明 |
| :--- | :--- | :--- | :--- |
| **密度** | $\rho$ | $kg/m^3$ | 均质流体常数 |
| **流量** | $q$ | $m^3/s$ | 单位时间流过的体积 |
| **体积** | $V$ | $m^3$ | $V = A \times h$ |
| **高度** | $h$ | $m$ | 液面位置 |
| **压强** | $P$ | $N/m^2$ (Pa) | 压力概念 |

### 压强的三个核心概念
1.  **静压 (Hydrostatic Pressure)**：由流体重力产生。
    
    $$P_{hydro} = \frac{mg}{A} = \frac{\rho V g}{A} = \frac{\rho A h g}{A} = \rho gh$$
2.  **绝对压强 (Absolute Pressure)**：包含大气压强。
    $$P_{abs} = P_a + \rho gh$$
3.  **表压 (Gauge Pressure)**：测量压力与参考压强（通常为大气压）的差值。
    $$P_{gauge} = P_{abs} - P_a = \rho gh$$

---

## 3. 流阻与理想压源 (02:19 - 03:37)

* **流阻 ($R$)**：流体流经管道、连接处时受到的阻碍。
    * **计算公式**：$P_1 - P_2 = \rho q R$
    * **电路类比**：类似于欧姆定律 $E_1 - E_2 = iR$。
    * **物理意义**：$\rho q$ 代表质量流率 ($kg/s$)。压差越大，单位时间内流过的质量越多。
* **理想压源**：如压缩机或泵。
    * **特性**：经过压源后，压力线性增加：$P_2 = P_1 + P_s$。

---

## 4. 质量守恒与动态方程推导 (03:38 - 05:35)



### 质量守恒定律
容器内流体质量的变化率等于流入质量与流出质量之差：
$$\frac{dm}{dt} = \dot{m}_{in} - \dot{m}_{out}$$

### 转化为高度变化率
1.  两边除以密度 $\rho$（因为 $m/\rho = V$）：
    $$\frac{dV}{dt} = q_{in} - q_{out}$$
2.  代入 $V = Ah$（$A$ 为常数）：
    $$A \frac{dh}{dt} = q_{in} - q_{out} \implies \frac{dh}{dt} = \frac{1}{A}(q_{in} - q_{out})$$

### 转化为压力变化率
已知 $P = P_a + \rho gh$，对时间 $t$ 求导：
$$\frac{dP}{dt} = \rho g \frac{dh}{dt}$$
代入高度变化公式：
$$\frac{dP}{dt} = \frac{\rho g}{A}(q_{in} - q_{out})$$

---

## 5. 综合求解案例 (05:36 - 08:08)

### 求解过程
1.  **应用质量守恒**：
    $$A \frac{dh}{dt} = q_{in} - q_{out}$$
2.  **建立出口流量方程**：
    在流阻 $R$ 两端，压差为 $P_1 - P_a$。
    已知 $P_1 = P_a + \rho gh$，则：
    $$\rho q_{out} R = P_1 - P_a = (P_a + \rho gh) - P_a = \rho gh$$
    解得：$q_{out} = \frac{gh}{R}$
3.  **合成最终动态方程**：
    将 $q_{out}$ 代入步骤 1：
    $$A \frac{dh}{dt} = q_{in} - \frac{gh}{R}$$
    进一步化简得到液面变化的动态方程：
    $$\frac{dh}{dt} = \frac{q_{in}}{A} - \frac{gh}{AR}$$

---
**提示**：这是一个典型的一阶线性系统。
